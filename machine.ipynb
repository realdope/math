**Information theory**: information $\blue{h(x)=-\log_2 p(x)}$ is equivalent to surprise and uncertainty, which holds an inverse relationship with probability.
If $x,y$ are independent, then $h(x,y)=h(x)+h(y)$.

- **Discrete entropy**: $\blue{\H[x]=-\sum_xp(x)\log_2 p(x)}=\E[h(x)]$ is average information. Its unit is bits.

  - **8-state proof**: Suppose $x$ has 8 possible states (A-H).
  If uniform $p(x)=\/{1}{8}$ then $\H[x]=-\sum_x\/{1}{8}(-3)=3$ bits binary coding 001,010,011,100,101,110,111.
  If non-uniform distribution $[\/{1}{2},\/{1}{4},\/{1}{8},\/{1}{16},\/{1}{64},\/{1}{64},\/{1}{64},\/{1}{64}]$ then $\H[x]=\/{1}{2}(1)+\/{1}{4}(2)+\/{1}{8}(3)+\/{1}{16}(4)+\/{4}{64}(6)=2$ bits using variable coding 0,10,110,1110,111110.
  Entropy is the absolute lower bound on average number of bits to transmit the state of a variable.

  - **Thermodynamics**: In Maxwell-Boltzmann statistics, gas molecules are distinguishable.
  Number of ways to distribute $N$ particles into bins of size $n_1+...+n_k=N$ is called multiplicity $W=\/{N!}{\prod_in_i!}$.
  Using Stirling approximation ($\ln n!=n\ln n-n$)
  we have
  $\ln W=\ln(\/{N!}{\prod_in_i!})$
  $=\ln N!-\sum_i\ln n_i!$
  $‚ČąN\ln N-N-\sum_in_i\ln n_i+\sum_in_i$
  $=N\ln N-\sum_in_i\ln n_i$
  $=\sum_in_i\ln N-\sum_in_i\ln n_i$
  $=-\sum_in_i\ln(\/{n_i}{N})$.
  Then we divide by $N$ to find entropy per particle
  $\lim\limits_{N‚Üí‚ąě}\/{1}{N}\ln W=-\sum_ip_i\ln(p_i)$
  $\dg{‚áí}\red{H=-\sum_ip_i\ln(p_i)}$

  \t{subject to}&p‚ČĹ0,\v{1}^‚ä§p=1

  $\dg{‚áí}L=\sum_ip_i\ln(p_i)+őĹ(\v{1}^‚ä§p-1)$
  $\dg{‚áí}g(őĹ)=\inf_p(\sum_ip_i\ln(p_i)+őĹ\sum_ip_i-őĹ)$
  $=-őĹ+\inf_p\sum_i(őĹp_i+p_i\ln(p_i))$.
  Minimize
  $‚ąā_{p_i}=őĹ+\ln(p_i)+1=0$
  $\dg{‚áí}p_i^*=e^{-őĹ-1}\ ‚ąÄi$

- **Differential entropy**: Mean value theorem says the area under PDF for the $i$-th bin has mean value $p(x_i)$ such that the probability is $p(x_i)őĒ$ satisfying $\sum_ip(x_i)őĒ=1$.
We quantize
$H_őĒ=-\sum_ip(x_i)őĒ[\ln p(x_i)+\ln őĒ]$

**Kullback-Leibler Divergence**: Suppose $p(x)$ is the true probability of 'A-Z', but $q(x)$ is what the model assumes. KL divergence is the measure of inefficiency that arise from mismatch between the model $q(x)$ and reality $p(x)$.
$\red{\KL(p\|q)=‚ąęp(x)\ln\/{p(x)}{q(x)}\ dx}$
$=-‚ąęp(x)\ln q(x)\ dx-(-‚ąęp(x)\ln p(x)\ dx)$ matches reality $p$ to model $q$.

- $\blue{\KL(p\|q)\neq\KL(q\|p)}$:

  - **Forward KL**: $\KL(p\|q)=‚ąęp(x)\ln\/{p(x)}{q(x)}\ dx$ is expectation wrt true distribution $p(x)$.
  In regions where $q(x)‚Čą0$ and $p(x)>0$, KL blows up. Loss minimization forces $q(x)$ to stretch out and avoid zeros in all regions where true distribution has probability mass.
  This **zero-avoiding** behavior overestimates variance.
  If $p(x)$ is a 2-peak GMM, $q(x)$ tries to position its peak at the mean of the two modes (**mean seeking**).

  - **Reverse KL**: $\KL(q\|p)=‚ąęq(x)\ln\/{q(x)}{p(x)}\ dx$ wants to avoid placing probability mass where true distribution $p(x)‚Čą0$, therefore it is **zero-forcing** for most of its distribution thereby creating a narrow distribution.
  If $p(x)$ is a 2-peak GMM, $q(x)$ tries to position its peak at one mode and zero everywhere else (**mode seeking**).

- $\blue{\KL(p\|q)‚Č•0}$.
Proof: Let $f(y)=-\ln\/{q(x)}{p(x)}$.
Then by Jensen's inequality
$\KL(p\|q)=\E_p[f(y)]‚Č•f(\E_p[y])$
$=-\ln ‚ąęp(x)\/{q(x)}{p(x)}\ dx$
$=-\ln ‚ąęq(x)\ dx$
$=0$.

- **Density estimation**: $\KL(p\|q)=-‚ąęp(x)\ln\/{q(x|őł)}{p(x)}\ dx$
$=-‚ąęp(x)\ln q(x|őł)\ dx+‚ąęp(x)\ln p(x)\ dx$.
We want to tweak $őł$ to make our model $q(x|őł)$ as close as possible to unknown true distribution $p(x)$.
Using empirical approximation: $‚ąęp(x)f(x)\ dx‚Čą\/{1}{N}\sum_if(x_i)$
$\dg{‚áí}\red{\KL(p\|q)‚Čą\/{1}{N}\sum_i[-\ln q(x_i|őł)+\ln p(x_i)]}$.
Therefore optimization
$\arg\min_őł\KL(p\|q)$
$‚Čą\arg\min_őł\sum_i-\ln q(x_i|őł)$
$‚Čą\arg\max_őł\sum_i\ln q(x_i|őł)$ is exactly MLE and log-likelihood!

- **Mutual information**:
$\red{\I[x,y]:=\KL(p(x,y)\|p(x)p(y))}$
$=-\iint p(x,y)\ln\/{p(x)p(y)}{p(x,y)}\ dx\ dy$

**Non-parametric**: Histogram with $K$ of $N$ points falling into bin $i$ that has bin size $V$ has probability $\/{K}{N}$ and continuous density $p=\/{K}{NV}$.
Suppose points are drawn from unknown true density $p(x)$ we approximate $P=‚ąę_Vp(x)\ dx‚Čąp(x)V$ for small $V$ in the neighborhood of $x$ following binomial expectation $K=NP$.
Therefore $\red{p(x)=\/{K}{NV}}$.
Fixing $K$ to determine $V$ is K-nearest-neighbors.
Fixing $V$ to determine $K$ is kernel method.

- **Kernel density estimation** is histogram method with overlapping bins positioned at data points (Parzen window) instead of a fixed grid. Let **kernel function** be unit hypercube
$\red{k(u)=\BC 1&|u_i|‚Č§1/2\\0&\t{otherwise}\EC}$ (Boxcar kernel), and each bin is a cube of size $V=h^D$.
Then
$K=\sum_ik(\/{x-x_i}{h})$ is the number of points inside a bin centered at $x$,
and is also number of bins centered at $x_i$ that contain $x$.
The density estimate is
$\blue{p(x)=\/{1}{N}\sum_i\/{1}{h^D}k(\/{x-x_i}{h})}$.

  - **Gaussian kernel**: $\red{k(u)=\Normal(0,1)}$ then $\blue{p(x)=\/{1}{N}\sum_i\/{1}{(2ŌÄh^2)^{1/2}}\e{-\/{\n{x-x_i}^2}{2h^2}}}$

  - Kernel functions $k(u)$ satisfy $k(u)‚Č•0,\ ‚ąęk(u)\ du=1$.

  - At dense clusters, small bins stack and describe a contour for the probability density. However sparse regions are overfit to noisy spikes in an ocean of zeros that mathematically assume zero probability of finding new data points. Using large bin size would wash out details of dense clusters.

- **K-nearest-neighbor**: In density estimation, grow a sphere at $x$ until it contains $K$ points, then probability density $p(x)=\/{K}{NV}$ is inversely proportional to sphere size $V$.

  - Suppose $N_k$ of $N$ points are in class $C_k$.
  To classify point $x$, we grow a sphere centered at $x$ until it contains $K$ points irrespective of class and $K_k$ of which belong to class $C_k$.
  $\BC \t{likelihood}&\t{marginal}&\t{prior}\\
  p(x|C_k)=\/{K_k}{N_kV}&p(x)=\/{K}{NV}&p(C_k)=\/{N_k}{N}
  \EC$
  $\dg{‚áí}p(C_k|x)=\/{p(x|C_k)p(C_k)}{p(x)}=\/{K_k}{K}$ posterior.
  Assign to majority class.

- Gram-Schmidt orthogonalization of $X=\BM\v{1}&x\EM‚ąą‚ĄĚ^{N√ó(p+1)}$ finds $Z=\BM z_0&...&z_p\EM‚ąą‚ĄĚ^{N√ó(p+1)}$ as orthogonal basis for the column space of $X=Zőď$.

**Shrinkage**: shrinkage methods reduce variance by shrinking the coefficients toward zero. Because regression is not unit invariant, the model will pull intercept toward zero if it is included in ridge penalty thus resulting in different slopes for Celsius vs Kelvin. Standard practice is to center $x$ and $y$ so that $\bar{x}=\bar{y}=0$ and $\hat{ő≤}_0=0$. Let $X‚ąą‚ĄĚ^{N√óp}$ and $\hat{ő≤}‚ąą‚ĄĚ^p$.

- **Ridge**: $\red{\hat{ő≤}^\t{ridge}=\arg\min_ő≤(\n{y-Xő≤}^2+őĽ\n{ő≤}^2)}$
$=\blue{(X^‚ä§X+őĽI)^{-1}X^‚ä§y}$. Let $X=Uő£V^‚ä§$ with thin $U‚ąą‚ĄĚ^{N√óD}$. Then $X^‚ä§X=Vő£^2V^‚ä§$

Along each axis $u_ju_j^‚ä§y$ is projection of $y$ onto unit vector $u_j$. Ridge regression is anisotropic and axis-dependent. Along each axis $\/{ŌÉ_j^2}{ŌÉ_j^2+őĽ}u_ju_j^‚ä§y$ applies shrinkage according to singular values.

  - **PCA**: Let $X^‚ä§X=Vő£^2V^‚ä§$, then code $Z_M=XV_M=U_Mő£_M‚ąą‚ĄĚ^{N√óM}$. The first principal component is the first column $z_1=Xv_1‚ąą‚ĄĚ^N$ with sample variance $\V[z_1]=\/{1}{N}z_1^‚ä§z_1$
  $=\/{1}{N}v_1^‚ä§X^‚ä§Xv_1$
  $=\/{ŌÉ_1^2}{N}v_1^‚ä§v_1$
  $\dg{‚áí}\blue{\V[z_1]=\/{ŌÉ_1^2}{N}}$.
  Then $ő£^2$ in $\/{ŌÉ_j^2}{ŌÉ_j^2+őĽ}$ are just variances of the data along principal axes.
  Higher variance axes are easy to estimate slopes - ridge leaves this alone. Low variance slopes can flip wildly to effect of noise - ridge aggressively attunates the coordinates along this axis toward zero (robust).

  - **Degrees of freedom**: Anisotropic shrinkage pulls each parameter toward zero and reduces freedom.
  $\BA

  =\tr[X(X^‚ä§X+őĽI)^{-1}X^‚ä§]

  is the effective degrees of freedom of the ridge regression fit.

- **Lasso**: $\red{\hat{ő≤}^\t{lasso}=\arg\min_ő≤(\n{y-Xő≤}_2^2+őĽ\n{ő≤}_1)}$ uses penalty $\sum_j|ő≤_j|$ instead of $\sum_jő≤_j^2$. Then in 2007 coordinate descent became the gold standard algorithm for lasso and elastic net.

- **Elastic net**: $őĽ\sum_{j=1}^p(őĪő≤_j^2+(1-őĪ)|ő≤_j|)$ has sharp corners while $\sum_j|ő≤_j|^q$ with q=1.x has more rounded corners.

**Least angle regression (LAR)**: Standardize $X‚ąą‚ĄĚ^{N√óp}$ (zero mean, unit norm). Center $y$ (zero mean). Set $ő≤=0$. Calculate residual $r_0=y-\bar{y}$. Find column $x_j$ most correlated with $r_0$ and add to active predictor set $\m{A}_1=\{x_j\}$.

- **Step-$k$** has prediction $\blue{\hat{y}_k=X_{\m{A}_k}ő≤_{\m{A}_k}}$, residual $\red{r_k=y-X_{\m{A}_k}ő≤_{\m{A}_k}}$, and direction $\red{őī_k=(X_{\m{A}_k}^‚ä§X_{\m{A}_k})^{-1}X_{\m{A}_k}^‚ä§r_k}$.

  - $\BA
  \Corr(x_j,r_k)=\/{\V[x_j,r_k]}{\sqrt{\V[x_j]\V[r_k]}}=\/{\/{1}{N}x_j^‚ä§r_k}{\sqrt{\/{1}{N}\V[r_k]}}=\/{1}{\sqrt{N\V[r_k]}}x_j^‚ä§r_k
  \dg{‚áí}\blue{c_k=X_{\m{A}_k}^‚ä§r_k}
  \EA$
  represents $\Corr(X_{\m{A}_k},r_k)$.

  - $őī_k‚ąą‚ĄĚ^{|\m{A}_k|}$ is the direction of coefficient update.
  As $x_j$ enters $A_k$, coefficient $ő≤_j$ has the same sign as its correlation with residual $x_j^‚ä§r_k$.
  $\BA
  \dg{‚áí}&őī_k=(X_{\m{A}_k}^‚ä§X_{\m{A}_k})^{-1}\brown{X_{\m{A}_k}^‚ä§r_k}\dg{‚áí}őī_{kj}=\/{\brown{x_j^‚ä§r_k}}{x_j^‚ä§x_j}\dg{‚áí}ő≤_j(őĪ)=őĪőī_{kj}=őĪ\/{\brown{x_j^‚ä§r_k}}{x_j^‚ä§x_j}
  \EA$

- **Updates**: During step-$k$ gradually increase step size $őĪ=0‚Üí1$ and grow active coefficients $\red{ő≤_{\m{A}_k}(őĪ)=ő≤_{\m{A}_k}+őĪőī_k}$ on top of the model states at the start of step-$k$.
$\BA
\dg{‚áí}&r(őĪ)=y-X_{\m{A}_k}ő≤_{\m{A}_k}(őĪ)=y-X_{\m{A}_k}(ő≤_{\m{A}_k}+őĪőī_k)=r_k-őĪX_{\m{A}_k}őī_k
\dg{‚áí}\brown{r(őĪ)=r_k-őĪX_{\m{A}_k}őī_k}\\
\dg{‚áí}&c(őĪ)=X_{\m{A}_k}^‚ä§r(őĪ)=X_{\m{A}_k}^‚ä§(r_k-őĪX_{\m{A}_k}őī_k)=c_k-őĪX_{\m{A}_k}^‚ä§X_{\m{A}_k}őī_k=c_k-őĪX_{\m{A}_k}^‚ä§X_{\m{A}_k}(X_{\m{A}_k}^‚ä§X_{\m{A}_k})^{-1}X_{\m{A}_k}^‚ä§r_k\dg{‚áí}\brown{c(őĪ)=(1-őĪ)c_k}\\
\dg{‚áí}&\hat{y}(őĪ)=X_{\m{A}_k}ő≤_{\m{A}_k}(őĪ)=X_{\m{A}_k}(ő≤_{\m{A}_k}+őĪőī_k)=\hat{y}_k+őĪX_{\m{A}_k}őī_k\dg{‚áí}\brown{\hat{y}(őĪ)=\hat{y}_k+őĪu_k}
\EA$

  - As coefficients move away from zero, the residual shrinks by $őĪX_{\m{A}_k}őī_k$ and the correlation of all active predictors with the residual are tied to $c(őĪ)$ and decrease at the same rate $(1-őĪ)$.

  - Equiangular direction $\blue{u_k=X_{\m{A}_k}őī_k}‚ąą‚ĄĚ^N$ is the direction of prediction update that ensures the correlation of all active predictors with the residual decrease at the same rate. Geometrically it sits exactly in the middle of all active predictors, and creates the least possible angle that is equal among all of them. For example if k=3, then $u_k$ is the vector right down the center of the cone formed by columns $x_1,x_2,x_3$.

  - $őĪ‚ąą[0,1]$. When $őĪ=1$ we have the OLS coefficients on the active set $X_{\m{A}_k}$
  $\BA
  ő≤_{\m{A}_k}+őī_k=ő≤_{\m{A}_k}+(X_{\m{A}_k}^‚ä§X_{\m{A}_k})^{-1}X_{\m{A}_k}^‚ä§r_k=ő≤_{\m{A}_k}+(X_{\m{A}_k}^‚ä§X_{\m{A}_k})^{-1}X_{\m{A}_k}^‚ä§(y-X_{\m{A}_k}ő≤_{\m{A}_k})=(X_{\m{A}_k}^‚ä§X_{\m{A}_k})^{-1}X_{\m{A}_k}^‚ä§y\dg{‚áí}\green{ő≤_{\m{A}_k}(1)=\hat{ő≤}_{\m{A}_k}^\t{ls}}
  \EA$

- **Next step**: As $őĪ$ increases and $c(őĪ)$ decreases, another column $x_j'$ satisfying $|x_j'^‚ä§r(őĪ)|=c(őĪ)$ is added to $\m{A}_{k+1}$ to start next step. As such LAR always takes $p$ steps to get to the full OLS coefficients.

  - Along the way LAR gives a sparse model with 1, 2, ..., p predictors with non-zero coefficients smoothly growing toward their OLS values.

- **LAR-Lasso** adds an additional rule: when $ő≤_j$ shrinks to exact zero, $x_j$ is removed from $\m{A}_{k+1}$ to start the next step.
LAR-Lasso takes more than $p$ steps to get the full OLS coefficients, and along the way traces the entire continuous path of lasso-regularized coefficients with $őĽ=‚ąě‚Üí0$. Any $őĽ>\max_j|x_j^‚ä§y|$ just gives $ő≤=0$.

  - Lasso ojective $\min_ő≤(\/{1}{2}\n{y-Xő≤}_2^2+őĽ\n{ő≤}_1)$ using subgradient $‚ąá\n{ő≤}_1$ has KKT stationarity conditions
  $\BA
  ‚ąá_ő≤=(y-Xő≤)^‚ä§(-X)+őĽ‚ąá\n{ő≤}_1^‚ä§=0\dg{‚áí}X^‚ä§r=őĽ‚ąá\n{ő≤}_1
  \dg{‚áí}\BC
  ő≤_j\neq0&‚ąā_j\n{ő≤}_1=\t{sign}(ő≤_j)&\brown{x_j^‚ä§r=őĽ‚čÖ\t{sign}(ő≤_j)}\\
  ő≤_j=0&‚ąā_j\n{ő≤}_1‚ąą[-1,1]&\brown{|x_j^‚ä§r|‚Č§őĽ}
  \EC\\
  \EA$.
  LAR-lasso active predictors ($ő≤_j\neq0$) have tied correlation $\brown{x_j^‚ä§r(őĪ)=c(őĪ)‚čÖ\t{sign}_j}$ while inactive predictors ($ő≤_j=0$) satisfy $\brown{|x_j'^‚ä§r(őĪ)|‚Č§c(őĪ)}$.
  LAR-lasso satisfies KKT stationarity as long as $\t{sign}_j=\t{sign}(ő≤_j)$, which only fails during $ő≤_j$ zero-crossing.
  Kicking out $x_j$ from $\m{A}_k$ prevents the failure of alignment with KKT stationarity.

  - LAR-lasso has same order of computation as a least squares fit with QR decomposition of $X$.

- **Degrees of freedom**: At step-$k$ the degrees of freedom is exactly $k$.

**Principal components regression**: Let $X=Uő£V^‚ä§‚ąą‚ĄĚ^{N√óp}$ with thin $U‚ąą‚ĄĚ^{N√óp}$, then principal components $Z_M=XV_M$ where $M‚Č§p$ already has orthogonal columns and doesn't need deflation.

$\blue{\hat{ő≤}_M^\t{pcr}=V_M\hat{őł}}=\sum_{j=1}^M\hat{őł}_jv_j$.

# PRML: 4 Linear Models for Classification

**Discriminant function** takes an input $x‚ąą‚ĄĚ^D$ and assigns to a class $C_1,...,C_K$.
Binary problem has target $t‚ąą\{1_{C_1},0_{C_2}\}$ while $K>2$ use one-hot $t‚ąą\{0,1\}^K$.
Linear regression has prediction function $y(x)=w^‚ä§x+w_0$.
Classification has **activation function** $y(x)=f(w^‚ä§x+w_0)$ mapping linear output to a probability (sigmoid).


- **Binary linear discriminant** has a decision hyperplane $y(x)=w^‚ä§x+w_0=0$ as class boundary. A hyperplane is defined to be
$\red{\{x|w^‚ä§x=-w_0\}=\{x|x‚ąąx_0+w^\perp\}}$ where
$x_0=-\/{w_0}{\n{w}}\/{w}{\n{w}}$ is the projection of origin. Distance from hyperplane to any point $x_1$ is the distance from $x_0$ to projection of $x_1$ onto $w$, which is
$\n{\/{w^‚ä§x_1}{\n{w}}\/{w}{\n{w}}-x_0}=\/{y(x_1)}{\n{w}}$.

  - Naively extending to $K>2$ multiclass by combining multiple binary classifiers.
  **One-vs-rest** trains $K$ classifiers independently.
  **One-vs-one** trains $K(K-1)/2$ classifiers independently.
  There will be ambiguous regions in the input space with points claimed by multiple classifiers and points claimed by no classifier.

- **$K$-class linear discriminant** trains $K$ discriminant functions simultaneously
$\red{y_k(x)=w_k^‚ä§x+w_{k0}}$
with decision region
$\red{R_k=\{x|y_k(x)‚Č•y_j(x)\ ‚ąÄj\}}$ and decision boundary
$y_k(x)=y_j(x)$
$\dg{‚áí}(w_k-w_j)^‚ä§x+(w_{k0}-w_{j0})=0$.

  - The decision region is convex.
  Proof: $R_k=‚čā_j\{x|(w_j-w_k)^‚ä§x‚Č§w_{k0}-w_{j0}\}$ is an intersection of halfspaces.

**OLS linear discriminant**: $\BA
W=\BM w_{10}&...&w_{K0}\\w_1&...&w_K\EM‚ąą‚ĄĚ^{(D+1)√óK},X=\BM\v{1}&x\EM‚ąą‚ĄĚ^{N√ó(D+1)}\dg{‚áí}XW=\BM y_1(x_1)&...&y_K(x_1)\\\vdots&\ddots&\vdots\\y_1(x_N)&...&y_K(x_N)\EM‚ąą‚ĄĚ^{N√óK}
\EA$ is trained against $T‚ąą\{0,1\}^{N√óK}$ by minimizing RSS error $\red{\/{1}{2}\n{XW-T}_F^2}=\/{1}{2}\tr[(XW-T)^‚ä§(XW-T)]$.

- $\red{W^*=(X^‚ä§X)^{-1}X^‚ä§T}$ where every row $y(x_i)^‚ä§=(XW)_i‚ąą‚ĄĚ^K$ guarantees $\v{1}^‚ä§y(x_i)=1$ but not $y(x_i)‚ČĹ0$ and is therefore not a probability distribution.

- OLS coefficients represents the normal of the separating hyperplane. On a 2D scatterplot of $x_1,x_2$ axes, regression predicts $x_2$ based on $x_1$.
Binary classification uses $x_1$ and $x_2$ to predict a third axis where $t=0$ is lower and $t=1$ is higher.
The boundary is the waterline at half-height.

- **Too correct problem** (Figure 4.4):
OLS tilts decision boundary to outlier training points placed far from boundary because it assumes $t_i‚ąľ\Normal(y(x_i),ŌÉ^2)‚ąą(-‚ąě,‚ąě)$ full range against actual $t_i‚ąą\{0,1\}$ discrete range.
Outlier $x_i$ makes large predictions far from $\{0,1\}$.
For example $y(x_i)=5$ has penalty $\n{5-1}^2$ (too correct), thus tilts the boundary.

- **Masking problem** (Figure 4.5): $K=3$ parallel clusters of data has targets (1,0,0), (0,1,0), and (0,0,1). $y_1(x)$ and $y_3(x)$ easily predicts the two outer classes. Middle column goes from 0 to 1 to 0, but affine $y_2(x)$ cannot bend and therefore fits to a middling constant, say, $y_2(x)=0.33$, and is easily masked by $y_1(x)$ and $y_2(x)$ in $\arg\max_ky_k(x)$.

**Fisher linear discriminant**: $\red{y(x_i)=w^‚ä§x_i+w_0}$ projects points onto axis $w$ against a threshold placed at $-w_0$.

- **Scatter matrices**: Covariance matrix is a sum of within-class and between-class scatters. Let $X‚ąą‚ĄĚ^{N√óD}$
$\BA
\dg{‚áí}&\V[X]=(X-\v{1}\bar{x}^‚ä§)^‚ä§(X-\v{1}\bar{x}^‚ä§)=\sum_i(x_i-\bar{x})(x_i-\bar{x})^‚ä§=\sum_k\sum_{i‚ąąC_k}(x_i-\bar{x})(x_i-\bar{x})^‚ä§\\

&=\sum_k\sum_{i‚ąąC_k}[(x_i-\bar{x}_k)(x_i-\bar{x}_k)^‚ä§+(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^‚ä§]=\green{\sum_k\sum_{i‚ąąC_k}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^‚ä§}+\yello{\sum_kN_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^‚ä§}\\
\dg{‚áí}&\V[X]=\green{S_W}+\yello{S_B}\dg{‚áí}\BC
\green{S_W=\sum_k\sum_{i‚ąąC_k}(x_i-\bar{x})(x_i-\bar{x})^‚ä§}=\sum_k(N_k-1)\hat{ő£}_k\dg{ - \t{within-class scatter}}\\
\yello{S_B=\sum_kN_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^‚ä§}\dg{ - \t{between-class scatter}}
\EC
\EA$

- **Binary scatter**: Let $k=\{1,2\}$ and $\bar{x}_k=\/{1}{N_k}\sum_{i‚ąąC_k}x_i$. Let $y_i=w^‚ä§x_i$ be data points projected onto axis $w$.
$\BA
\dg{‚áí}&\dg{\t{between-class }}(\bar{y}_1-\bar{y}_2)^2=[w^‚ä§(\bar{x}_1-\bar{x}_2)]^2=w^‚ä§[\yello{(\bar{x}_1-\bar{x}_2)(\bar{x}_1-\bar{x}_2)^‚ä§}]w=w^‚ä§\yello{S_{12}}w\dg{ - \t{$S_{12}$ and $S_B$ are related as below}}\\
&\dg{‚áí}\bar{x}=\/{N_1}{N}\bar{x}_1+\/{N_2}{N}\bar{x}_2
\dg{‚áí}\bar{x}_1-\bar{x}=\bar{x}_1-(\/{N_1}{N}\bar{x}_1+\/{N_2}{N}\bar{x}_2)\dg{‚áí}\brown{\bar{x}_1-\bar{x}}=\/{N_2}{N}(\bar{x}_1-\bar{x}_2)\dg{‚áí}\grass{\bar{x}_2-\bar{x}}=-\/{N_1}{N}(\bar{x}_1-\bar{x}_2)\\
&\dg{‚áí}\yello{S_B}=N_1(\brown{\bar{x}_1-\bar{x}})(\brown{\bar{x}_1-\bar{x}})^‚ä§+N_2(\grass{\bar{x}_2-\bar{x}})(\grass{\bar{x}_2-\bar{x}})^‚ä§=(N_1\/{N_2^2}{N^2}+N_2\/{N_1^2}{N^2})(\bar{x}_1-\bar{x}_2)(\bar{x}_1-\bar{x}_2)^‚ä§\dg{‚áí}\yello{S_B=\/{N_1N_2}{N}S_{12}}\\
\dg{‚áí}&\dg{\t{within-class }}s_k^2=\sum_{i‚ąąC_k}(y_i-\bar{y}_k)^2=\sum_{i‚ąąC_k}[w^‚ä§(x_i-\bar{x}_k)]^2=w^‚ä§[\green{\sum_{i‚ąąC_k}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^‚ä§}]w=w^‚ä§\green{S_{k}}w\\

\EA$

- **Fisher criterion** is $\red{J(w):=\/{w^‚ä§\yello{S_B}w}{w^‚ä§\green{S_W}w}}$ is the ratio of projected scatter along axis $w$.
We pick axis $w^*=\arg\sup_wJ(w)$ that maximizes $\/{(\bar{y}_2-\bar{y}_1)^2}{s_1^2+s_2^2}=\/{w^‚ä§\yello{S_{12}}w}{w^‚ä§[\green{S_1+S_2}]w}$.

- **Dimensionality/Rank**: $K$ points can only make up to $K-1$ linearly independent vectors.
$K$ centroids can only span up to $K-1$ dimensions.
When $D‚ČęK$ the additional orthogonal dimensions provide zero discriminitive value.
$S_B=\sum_kN_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^‚ä§‚ąą‚ĄĚ^{D√óD}$ is a sum of $K$ rank-1 matrices.
$\rk(S_B)=K$ but $\bar{x}$ used 1 degree of freedom. Only $K-1$ of those matrices are linearly independent.
$S_W^{-1}S_B$ has at most $K-1$ non-zero eigenvalues.
Therefore $1< D'‚Č§K-1$.

- **Whitening and rank reduction**: Conceptually $\red{y_i=W^‚ä§x_i=V^‚ä§S_W^{-1/2}x_i}$.
$\BA
\dg{‚áí}&\red{W=S_W^{-1/2}V}\dg{‚áí}\brown{S_W^{-1}S_B}W=Wőõ\dg{‚áí}S_W^{-1}S_B(S_W^{-1/2}V)=(S_W^{-1/2}V)őõ\dg{‚áí}\brown{S_W^{-1/2}S_BS_W^{-1/2}}V=Vőõ\dg{‚áí}\tilde{S}_W^{-1}\tilde{S}_BV=Vőõ\\
&\BC

\tilde{S}_B=S_W^{-1/2}S_BS_W^{-1/2}\dg{ - \t{$V$ is orthogonal matrix containing eigenvectors of whitened $\tilde{S}_B$}}\\
\tilde{\V}[X]v=(I+\tilde{S}_B)v=(1+őĽ)v\dg{ - \t{whitened $\tilde{\V}[X]$ and $\tilde{S}_B$ share the same eigenvectors in $V$}}
\EC\\
\dg{‚áí}&\tilde{x}_i=S_W^{-1/2}x_i\dg{‚áí}\tilde{S}_w=\sum_k\sum_{i‚ąąC_k}(\tilde{x}_i-\tilde{\bar{x}}_k)(\tilde{x}_i-\tilde{\bar{x}})^‚ä§=\sum_k\sum_{i‚ąąC_k}S_W^{-1/2}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^‚ä§S_W^{-1/2}\\

&\dg{‚áí\t{ OR }}\V[x_i|C_k]=S_W\dg{‚áí}\V[\tilde{x}_i|C_k]=\V[S_W^{-1/2}x_i|C_k]=S_W^{-1/2}\V[x_i|C_k]S_W^{-1/2‚ä§}=S_W^{-1/2}S_WS_W^{-1/2}=I

  - **Whitening / Sphering**: We start with $K$ classes of points in standard space, each grouped in ellipsoid cluster whose shape is defined by their within-class covariance.
  $S_W$ is the summation (therefore average) of $K$ cluster shapes.
  $\tilde{x}_i=S_W^{-1/2}x_i$ brings all points to a new basis where $\tilde{S}_W=I$ is a unit-variance sphere instead of an ellipsoid.
  Each transformed cluster on average looks like a unit-norm ball (**centroid**).

  - **Rank reduction**: PCA code $z_i=V_M^‚ä§x_i$ is coordinates of $x_i$ along the first $M$ eigenvectors of $\V[X]$ representing $M$ highest variance directions in $X$.
  The $D'$ eigenvectors of $\tilde{S}_B$ contained in $V‚ąą‚ĄĚ^{D√óD'}$ are also eigenvectors of $\tilde{\V}[X]$, which represent most spread out directions in the whitened data, or the most spread out directions between unit-norm centroids of whitened data.

- **Nearest centroid classification**: Centroid $k$ has mean $\bar{y}_k=W^‚ä§\bar{x}_k=V^‚ä§S_W^{-1/2}\bar{x}_k$. Then $x_i$ is assigned to class $\arg\min_k\n{y_i-\bar{y}_k}^2=\n{V^‚ä§S_W^{-1/2}(x_i-\bar{x}_k)}^2$.

  - Suppose $P(x_i|C_k)=\Normal(\bar{x}_k,ő£_k)$, then $S_W=\sum_k(N_k-1)\hat{ő£}_k$ and $\hat{ő£}_k=\/{1}{N_k-1}\sum_{i‚ąąC_k}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^‚ä§$ is unbiased estimator for $ő£_k$.
  Euclidean distance $\n{y_i-\bar{y}_k}^2$ in whitened space is equivalent to Malahanobis distance $(x_i-\bar{x}_k)^‚ä§ő£_k^{-1}(x_i-\bar{x}_k)$ in standard space.

**Perceptron discriminant**: binary linear discriminant $y(x_i)=\t{sign}(w^‚ä§Ōē(x_i))$ with activation function $\t{sign}(‚čÖ)$.
We seek $w$ predicting $\BC

w^‚ä§Ōē(x_i)< 0\dg{‚áí}i‚ąąC_2
\EC$ against $t_i‚ąą\{+1_{C_1},-1_{C_2}\}$
with cost $\BC
-w^‚ä§Ōē(x_i)t_i> 0&\t{if misclassified }i‚ąą\m{M}\\
-w^‚ä§Ōē(x_i)t_i=0&\t{if correctly classified}
\EC$.

- **Cost** $E_P(w)=-\sum_{i‚ąą\m{M}}w^‚ä§Ōē(x_i)t_i=\sum_{i=1}^N\max(0,-w^‚ä§Ōē(x_i)t_i)$ is piecewise linear sum of $N$ hockey sticks.
SGD $w^{(ŌĄ+1)}=w^{(ŌĄ)}-ő∑‚ąáE_P(w)$ uses $‚ąáE_P(w)=-Ōē(x_i)t_i$ for random $i‚ąą\m{M}$.
Then $\red{w^{(ŌĄ+1)}=w^{(ŌĄ)}+ő∑Ōē(x_i)t_i}$.

  - Because magnitude is ignored by the activation function and $w^{(0)}=0$, we can assume $ő∑=1$ without loss of generality as long as it stays constant.

  - Naive solution to the problem by simply counting $|\m{M}|$ creates a piecewise constant error, which does not have gradients.
  Perceptron algorithm has piecewise linear error, which allows gradient descent.
  The non-differentiable joints between pieces are not a problem for SGD because $‚ąáE_P(w)=-Ōē(x_i)t_i$ is always defined.

- **Heart of algorithm**: $w:=w+Ōē(x_i)$ for misclassified $t_i=+1$ to align $w$ closer to $Ōē(x_i)$ and $w^‚ä§Ōē(x_i)‚Üí+1$.
Update $w:=w-Ōē(x_i)$ for misclassified $t_i=-1$ to align $w$ farther from $Ōē(x_i)$ and $w^‚ä§Ōē(x_i)‚Üí-1$.

- **Convergence**: If the data points are linearly separable into two classes neatly by a separating hyperplane, then the algorithm is guaranteed to find an exact solution in finite steps.

  - If the data points are not linearly separable then the algorithm will never converge.

  - The algorithm does not generalize to $K>2$.

**Generative approach** models the underlying distributions $p(x_i|C_k)$ and $p(C_k)$, then calculate $p(C_k|x_i)$ the probability that input $x‚ąą‚ĄĚ^D$ belongs to a class $C_1,...,C_K$.
"Generative" refers to the ability to generate synethetic $x_i$ from model.
**Discriminative approach** focuses on the decision boundary using MLE parameters defined through $p(C_k|x_i)$.

- **Logistic and logit**: $p(C_1|x_i)=\/{p(x_i|C_1)p(C_1)}{p(x_i|C_1)p(C_1)+p(x_i|C_2)p(C_2)}=\/{1}{1+\e{-a}}=ŌÉ(a)$ is logistic sigmoid function

$\red{ŌÉ(a)=\/{1}{1+\e{-a}}}=\/{1}{1+\/{p(x_i|C_2)p(C_2)}{p(x_i|C_1)p(C_1)}}=\/{p(x_i|C_1)p(C_1)}{p(x_i|C_1)p(C_1)+p(x_i|C_2)p(C_2)}$.

  $\BA

  \dg{‚áí}&ŌÉ=\/{1}{1+\e{-a}}\dg{‚áí}1=ŌÉ+ŌÉ\e{-a}\dg{‚áí}\e{-a}=\/{1-ŌÉ}{ŌÉ}\dg{‚áí}a=\ln\/{ŌÉ}{1-ŌÉ}\dg{\t{ log-odds link function}}\\

  \dg{‚áí}&ŌÉ'(a)=\/{\e{-a}}{(1+\e{-a})^2}=ŌÉ(a)(1-ŌÉ(a))\dg{\t{ derivative is the cheap-to-compute standard logistic distribution}}\\

$\BA

**Logistic regression**: Given binary $ő¶‚ąą‚ĄĚ^{N√ó(M+1)}$ and $t‚ąą\{1_{C_1},0_{C_2}\}^N$, where $Ōē_i$ is the $i$-th row and $Ōē_{i0}=1$.

$\BA

$\BA

  $\/{t_i-y_i}{y_i(1-y_i)}$ is positive when $t_i=1$ and negative when $t_i=0$.

# ESL: 4 Linear Models for Classification

**OLS classification**: $K$ classes, training data $(X‚ąą‚ĄĚ^{N√ó(p+1)},Y‚ąą‚ĄĚ^{N√óK})$ with one-hot rows in response, and coefficients $\hat{B}=(X^‚ä§X)^{-1}X^‚ä§Y=\small\BM \hat{ő≤}_0\\\hat{ő≤}\EM\normalsize‚ąą‚ĄĚ^{(p+1)√óK}$. Output $\hat{f}(x_i)^‚ä§=(1,x_i^‚ä§)\hat{B}$ assigning class $\hat{G}(x_i)=\arg\max_k\hat{f}_k(x_i)$.

- Classification attempts to reproduce one-hot $y_i$ target.
$\v{1}^‚ä§\hat{f}(x_i)=1$, which is similar to a probability distribution.
It is not a true distribution because individual falues of $\hat{f}(x_i)$ can fall outside of $[0,1]$.

- Masking problem overshadows the middle class $\hat{f}_\t{middle}(x_i)$ by edge classes $\hat{f}_\t{edge1}(x_i)$ and $\hat{f}_\t{edge2}(x_i)$ because the regression line is linear.
The problem can be overcome with quadratic regression.
When $K‚Č•3$ classes are lined up, polynomial regression with terms up to degree $K-1$ might be needed to resolve them.

**Linear discriminant analysis (LDA)** is a generative model based on Gaussian likelihood $p(x_i|G=k)=\Normal(x_i|őľ_k,ő£)$ for each class.
LDA and QDA work well not because real-world data are Gaussian but because of bias-variance trade-off.
They are stable estimators with low variance.
$\BA
\dg{‚áí}&P(G=k|x_i)=\/{p(x_i|G=k)ŌÄ_k}{\sum_{j=1}^Kp(x_i|G=j)ŌÄ_j}\dg{‚áí}\red{p(x_i|G=k)=\Normal(x_i|őľ_k,ő£)}\dg{ - \t{assume Gaussian likelihood and shared covariance }ő£_k=ő£}\\
\dg{‚áí}&\ln\/{P(G=k|x_i)}{P(G=l|x_i)}=\ln\/{\yello{ŌÄ_k}}{ŌÄ_l}-\/{1}{2}[\yello{(x_i-őľ_k)^‚ä§ő£^{-1}(x_i-őľ_k)}-(x_i-őľ_l)^‚ä§ő£^{-1}(x_i-őľ_l)]\dg{ - x_i^‚ä§ő£^{-1}x_i\t{ cancels out in posterior log-odds}}\\

&\dg{‚áí\t{LDA binary discriminant }}f(x_i)=\grass{ő≤^‚ä§}x_i+\brown{ő≤_0}>0‚áí\hat{G}(x_i)=k\dg{\t{ else }}\hat{G}(x_i)=l\\

\dg{‚áí}&\hat{ŌÄ}_k=\/{N_k}{N},\ \hat{őľ}_k=\/{1}{N_k}\sum_{g_i=k}x_i,\ \blue{\hat{ő£}=\/{1}{N-K}\sum_k\sum_{g_i=k}(x_i-\hat{őľ}_k)(x_i-\hat{őľ}_k)^‚ä§}\dg{ - \t{$g_i$ is actual class of $x_i$}}\\

- **Whitening/Sphering**: in practice LDA uses whitening/sphering on data to transform the Mahalanobis distance $\yello{(x_i-őľ_k)^‚ä§ő£^{-1}(x_i-őľ_k)}$ to using Euclidean norm distance $\n{\tilde{x}_i-\tilde{őľ}_k}^2$ for classification.
$\BA
\dg{‚áí}&\hat{ő£}=UDU^‚ä§\dg{‚áí}\tilde{x}_i=D^{-1/2}U^‚ä§x_i,\ \tilde{őľ}_k=D^{-1/2}U^‚ä§őľ_k\dg{ - \t{sphering: $\V[x_i|g_i=k]=\hat{ő£}‚áí\V[\tilde{x}_i|g_i=k]=I$}}\\

\EA$

- **Reduced rank LDA (Fisher)**: $K$ points can form at most $K-1$ linearly independent vectors.
K centroids can span a subspace of at most dimension $K-1$.
When $p‚ČęK$ we can ignore the additional orthogonal dimensions as they would contribute no discriminitive value.

- In binary classification, LDA coefficients $\grass{ő≤=ő£^{-1}(őľ_k-őľ_l)}$ is proportional to Fisher projection axis $w‚ąĚS_W^{-1}(őľ_k-őľ_l)$ and OLS coefficients.

The following proof uses target $y_i=\{+1,-1\}$ for algebraic simplicity.
$\BA

&\dg{‚áí}ő£ő≤=\/{2N_kN_l}{N}(őľ_k-őľ_l)\dg{‚áí}\grass{ő≤‚ąĚő£^{-1}(őľ_k-őľ_l)}
\EA$

  - Recall when target is $y_i=\{+\/{N}{N_k},-\/{N}{N_l}\}$, OLS coefficients are identical to Fisher coefficients.

- **Quadratic discriminant analysis (QDA)**: If we drop the assumption that $ő£_k=ő£$ for all classes, then the quadratic $x_i^‚ä§ő£_k^{-1}x_i$ term remains and decision boundaries are quadratic curves (parabola, ellipsoid, hyperbola).
$\BA

**Logistic regression**: We model posteriors of $K$ classes using anchored softmax with $ő≤‚ąą‚ĄĚ^{p√ó(K-1)}$
$\BA
\dg{‚áí}&k=1,...,K-1\dg{‚áí}\ln\/{P(g_i=k|x_i)}{P(g_i=K|x_i)}=ő≤_{k0}+ő≤_k^‚ä§x_i\dg{‚áí}P(g_i=k|x_i)=P(g_i=K|x_i)\e{ő≤_{k0}+ő≤_k^‚ä§x_i}\\

- Binary classification is trained from MLE $l(őł)=\sum_i\ln p_{g_i}(x_i;őł)$ where notation $p_{g_i}(x_i;őł):=p(g_i|x_i,őł)$.
$\BA
\dg{‚áí}&ő≤:=(ő≤_{10},ő≤_1)‚ąą‚ĄĚ^{p+1},\ x_i:=(1,x_i)‚ąą‚ĄĚ^{p+1}\dg{ - \t{assume intercept is included}}\\

  - $\hat{ő≤}$ is MLE: it is asymptotically consistent $\E[\hat{ő≤}]\arr{p}ő≤$ and efficient $\V[\hat{ő≤}]\arr{p}\/{1}{I_n(ő≤)}=\/{1}{-l''(ő≤)}=(X^‚ä§RX)^{-1}$.

# PRML: 5 Neural Networks

**Feed-forward networks** use hidden layers acting as basis functions instead of fixed basis functions $y(x)=f(w^‚ä§Ōē(x))$.
$\BA
\dg{‚áí}&x‚ąą‚ĄĚ_i^{D+1}‚Üí\block{z_j=h(a_j^{(1)}=w_j^{(1)‚ä§}x)}‚Üíz‚ąą‚ĄĚ_j^{M+1}‚Üí\block{y_k=f(a_k^{(2)}=w_k^{(2)‚ä§}z)}‚Üíy‚ąą‚ĄĚ_k^K\\
&x=\BM 1\\x_1\\\vdots\\x_D\EM‚ąą‚ĄĚ^{D+1}\dg{‚áí}z(x)=\BM1\\h(w_1^{(1)‚ä§}x)\\\vdots\\h(w_M^{(1)‚ä§}x)\EM‚ąą‚ĄĚ^{M+1}
\dg{‚áí}y(x)=\BM f(w_1^{(2)‚ä§}z)\\\vdots\\f(w_K^{(2)‚ä§}z)\EM‚ąą‚ĄĚ^K\dg{ - \t{forward propagation}}\\
&\dg{‚áí\t{Hidden unit $j=1..M$ has input $x$, activation $a_j^{(1)}=a_j^{(1)}(x)=w_j^{(1)‚ä§}x$, and output $z_j=z_j(x)=h(a_j^{(1)})$}}\\
&\dg{‚áí\t{Output unit $k=1..K$ has input $z=z(x)$, activation $a_k^{(2)}=a_k^{(2)}(z)=w_k^{(2)‚ä§}z$, and output $y_k=y_k(x)=f(a_k^{(2)})$}}\\
\EA$

- **Weight index convention**: $w_j^{(1)}$ refers to the weight in hidden layer $z_j=z_j(x)=w_j^{(1)‚ä§}x=\sum_{k=1}^Dw_{jk}^{(1)}x_k$. Here $w_{jk}^{(1)}$ refers to destination $z_j$ from source feature $x_k$.

  - Canonical link activation functions are used. $f(a_k)=a_k$ for linear regression, $f(a_k)=\/{1}{1+\e{-a_k}}$ for binary classification, and $f(a_k)=\/{\e{a_k}}{\sum_j\e{a_j}}$ for multiclass classification.

- **Universal approximation theorem**: We attempt to figure out an unknown underlying rule of the universe that generated outputs from inputs.
A simple input ‚ü∂ hidden ‚ü∂ output two-layer network with enough hidden units can uniformly (no blowup/jump) approximate any continuous function on a limited finite input domain to an arbitrary level of accuracy.

- **Weight-space symmetry**: the choice for weights to achieve the same input ‚ü∂ output mapping is not unique.

  - Sign-flip: activation functions are odd. If we flip input and output sign of hidden unit, then weight also flips sign.

  - Ordering of hidden units don't matter. Indexing of weights change with indexing of hidden units.

**Error backpropagation** refers only to the procedure to calculate $‚ąáE(w)$, before finding $w$ for that layer.
There are $M+K$ units.
**Unit error** $\red{őī_j=\/{‚ąāE}{‚ąāa_j}}$ represents derivative of error wrt activation at every unit.
$\BA
\dg{‚áí}&x‚ąą‚ĄĚ_i^{D+1}‚Üí\block{z_j=h(a_j^{(1)}=w_j^{(1)‚ä§}x)}‚Üíz‚ąą‚ĄĚ_j^{M+1}‚Üí\block{y_k=a_k^{(2)}=w_k^{(2)‚ä§}z}‚Üíy‚ąą‚ĄĚ_k^K\dg{ - \t{regression output layer}}\\
\dg{‚áí}&E(w)=\sum_{n=1}^NE_n(w)\dg{ - \t{total error. De-index for all below $x‚ąą‚ĄĚ^{D+1}$, $y‚ąą‚ĄĚ^K$, $t‚ąą‚ĄĚ^K$, and $E:=E_n$}}\\
\dg{‚áí}&a_k=a_k(z)=w_k^‚ä§z\dg{‚áí}\/{‚ąāE}{‚ąāw_{kj}}=\yello{\/{‚ąāE}{‚ąāa_k}}\/{‚ąāa_k}{‚ąāw_{kj}}\dg{‚áí}\yello{őī_k:=\/{‚ąāE}{‚ąāa_k}}‚ąą‚ĄĚ\dg{‚áí}\red{\/{‚ąāE}{‚ąāw_{\green{kj}}}=\yello{őī_\green{k}}z_{\green{j}}}\\
&\dg{‚áí\t{let $z$ be input, then all layers and units satisfy $‚ąá_{w_{kj}}E=őī_jz_j$. Once we have all the $őī_j$-s we have $‚ąá_wE(w)$}}\\
\dg{‚áí}&\dg{\t{Output layer: }}E=\/{1}{2}\n{y-t}^2=\/{1}{2}\sum_{k=1}^K(y_k-t_k)^2,\ y_k=a_k=w_k^‚ä§z=\sum_{j=0}^Mw_{kj}z_j\\
&\dg{‚áí}‚ąá_{w_k}E=(y_k-t_k)z\dg{‚áí}\/{‚ąāE}{‚ąāw_{kj}}=(\yello{y_k-t_k})z_j\dg{‚áí}\/{‚ąāE}{‚ąāa_k}=\/{‚ąāE}{‚ąāy_k}=y_k-t_k\dg{‚áí}\red{őī_k^{(2)}=y_k-t_k}\dg{ - \t{canonical activation}}\\
\dg{‚áí}&\dg{\t{Hidden layer: }}a_k^{(2)}=w_k^{(2)‚ä§}\BM1\\\\h(a_1^{(1)})\\\vdots\\h(a_M^{(1)})\EM\dg{‚áí}\/{‚ąāa_k^{(2)}}{‚ąāa_j^{(1)}}=w_{kj}^{(2)}h'(a_j^{(1)})\dg{‚áí}\/{‚ąāE}{‚ąāa_j^{(1)}}=\sum_{k=1}^K\/{‚ąāE}{‚ąāa_k^{(2)}}\/{‚ąāa_k^{(2)}}{‚ąāa_j^{(1)}}=h'(a_j^{(1)})\sum_kőī_k^{(2)}w_{kj}^{(2)}\\
&\dg{‚áí}\red{őī_j^{(1)}=h'(a_j^{(1)})\sum_kőī_k^{(2)}w_{kj}^{(2)}}\dg{ - \t{no $t_j$ access for earlier layers. Back-propagate sum of downstream errors}}\\
\dg{‚áí}&\dg{\t{Error function is determined by output layer activation function MLE}}\\
&\dg{‚áí\t{let: }}x‚ąą‚ĄĚ_i^{D+1}‚Üí\block{z_j=h(a_j^{(1)}=w_j^{(1)‚ä§}x)}‚Üíz‚ąą‚ĄĚ_j^{M+1}‚Üí\block{y_k=ŌÉ(a_k^{(2)}=w_k^{(2)‚ä§}z)}‚Üíy‚ąą‚ĄĚ_k^K\dg{ - \t{logistic output layer}}\\
&\dg{‚áí\t{then: }}E=-[t_k\ln y_k+(1-t_k)\ln(1-y_k)]\dg{‚áí}őī_k^{(2)}=\/{‚ąāE}{‚ąāa_k}=-\/{‚ąāE}{‚ąāy_k}\/{‚ąāy_k}{‚ąāa_k}=-(\/{t_k}{y_k}-\/{1-t_k}{1-y_k})y_k(1-y_k)=y_k-t_k\\
\EA$

- **Example**: $\BA
x‚ąą‚ĄĚ_i^{D+1}‚Üí\block{z_j=\tanh(a_j^{(1)}=w_j^{(1)‚ä§}x)}‚Üíz‚ąą‚ĄĚ_j^{M+1}‚Üí\block{y_k=a_k^{(2)}=w_k^{(2)‚ä§}z}‚Üíy‚ąą‚ĄĚ_k^K
\EA$
$\BA
\dg{‚áí}h(a)=\tanh(a)=2ŌÉ(2a)-1=\/{e^a-e^{-a}}{e^a+e^{-a}}\dg{‚áí}h'(a)=1-h(a)^2\quad\dg{‚áí}E=\/{1}{2}\sum_k(y_k-t_k)^2\\
\dg{‚áí\t{forward propagation: }}a_j^{(1)}=w_j^{(1)‚ä§}x\dg{‚áí}z_j=\tanh(a_j^{(1)})\dg{‚áí}a_k^{(2)}=w_k^{(2)‚ä§}z\dg{‚áí}y_k=a_k^{(2)}\\
\dg{‚áí\t{error backpropagation: }}őī_k^{(2)}=y_k-t_k\dg{‚áí}őī_j^{(1)}=(1-z_j^2)\sum_kőī_k^{(2)}w_{kj}^{(2)}\\
\dg{‚áí\t{gradients: }}‚ąá_{w_j^{(1)}}E=őī_j^{(1)}x‚ąą‚ĄĚ^{M+1}\dg{‚áí}‚ąá_{w_k^{(2)}}E=őī_k^{(2)}z‚ąą‚ĄĚ^{K+1}
\EA$

- **Efficiency**: A 2-layer feed-forward network has $\red{W:=(D+1)√óM+(M+1)√óK}$ weights, $M+K$ units, and $(D+1)+(M+1)+K$ nodes.
There are $M+K$ error deltas to calculate. Most of the compute is on forward pass calculating $y$ and $z$ and backward pass calculating the error deltas from $y$ and $z$. Each pass is $O(W)$.

- **Numerical differences**: adjust each $w_j$ by a tiny bit and see the difference in $E(w)$. This method is used for verifying implementation. Very computationally expensive, uses $O(W^2)$.

- **Jacobian backpropagation**: $x‚ąą‚ĄĚ_i^{D+1}\dg{‚áí}\red{J_{ki}:=\/{‚ąāy_k}{‚ąāx_i}}$ represents derivative of a system output wrt an input.
$\BA
\dg{‚áí}J_{ki}=\/{‚ąāy_k}{‚ąāx_i}=\sum_{j=1}^M\yello{\/{‚ąāy_k}{‚ąāa_j^{(1)}}}\green{\/{‚ąāa_j^{(1)}}{‚ąāx_i}}\dg{ - \t{factor $\green{\/{‚ąāa_j^{(1)}}{‚ąāx_i}}$ is known. We need to recursively expand $\yello{\/{‚ąāy_k}{‚ąāa_j^{(1)}}}=\sum_{l=1}^K\purple{\/{‚ąāy_k}{‚ąāa_l^{(2)}}}\brown{\/{‚ąāa_l^{(2)}}{‚ąāa_j^{(1)}}}$}}\\
\quad=\sum_{j=1}^M\left[\sum_{l=1}^K\purple{\/{‚ąāy_k}{‚ąāa_l^{(2)}}}\brown{\/{‚ąāa_l^{(2)}}{‚ąāa_j^{(1)}}}\right]\green{\/{‚ąāa_j^{(1)}}{‚ąāx_i}}=\sum_{j=1}^M\left[\sum_{l=1}^K\purple{\/{‚ąāy_k}{‚ąāa_l^{(2)}}}\brown{h'(a_j)w_{lj}^{(2)}}\right]\green{\/{‚ąāa_j^{(1)}}{‚ąāx_i}}\dg{‚áí}\purple{\/{‚ąāy_k}{‚ąāa_l^{(2)}}}=\BC
I_{kl}ŌÉ'(a_l^{(2)})&\t{sigmoid}\\
I_{kl}y_k-y_ky_l&\t{softmax}\EC\\
\EA$

**Hessian**: Parameter matrices are flattened so that $w_{kj}^{(2)}$ and $w_{ji}^{(1)}$ are in one $‚ĄĚ^W$ vector.
$\/{‚ąāE}{‚ąāw_{kj}}$ and $\/{‚ąāE}{‚ąāw_{ji}}$ also are in one $‚ĄĚ^W$ vector.
Then $H‚ąą‚ĄĚ^{W√óW}$ is a matrix instead of a 3D tensor.

- **Local quadratic approximation**: network $E(w)$ is non-convex surface with many local minima. We can model $E(w)$ using 2nd order Taylors where $H$ dictates the shape of the surface near optimal point.
$\BA
\dg{‚áí}&E(w)‚ČąE(\hat{w})+(w-\hat{w})^‚ä§‚ąáE(\hat{w})+\/{1}{2}(w-\hat{w})^‚ä§H(w-\hat{w}),\ H=‚ąá^2E(\hat{w})\dg{ - \t{Taylor 2nd order in neighborhood of $\hat{w}$}}\\
&\dg{‚áí}E(w)=E(w^*)+\/{1}{2}(w-w^*)^‚ä§H(w-w^*)\dg{ - \t{near $\hat{w}=w^*$ optimal point $‚ąáE(w^*)=0$ and shape is dictated by $H$}}\\
&\dg{‚áí}Hu_i=őĽ_iu_i\dg{‚áí}w-w^*=\sum_iőĪ_iu_i=U^‚ä§őĪ\dg{ - \t{$őĪ_i$ represents coordinate of $w-w^*$ along $u_i$}}\\
\dg{‚áí}&E(w)=E(w^*)+\/{1}{2}\sum_iőĪ_iu_i^‚ä§őĪ_iőĽ_iu_i\dg{‚áí}E(w)=E(w^*)+\/{1}{2}\sum_iőĪ_i^2őĽ_i\dg{ - \t{positive-definite ($őĽ_i>0$) ‚áí minimum $w^*$}}
\EA$

  - Positive-definite $H$ ensures ellipsoidal level-set $(w-w^*)^‚ä§H(w-w^*)=c$ dimensional radii $őĽ_i>0$. If $őĽ_i=0$ then that dimension collapses into a half-pipe and there is no unique minimum.

- **Exact Hessian** can be found via backpropagation at $O(W^2)$
$\BA
\dg{‚áí}&x‚ąą‚ĄĚ_i^{D+1}‚Üí\block{z_j=h(a_j^{(1)}=w_j^{(1)‚ä§}x)}‚Üíz‚ąą‚ĄĚ_j^{M+1}‚Üí\block{y_k=a_k^{(2)}=w_k^{(2)‚ä§}z}‚Üíy‚ąą‚ĄĚ_k^K\\
\dg{‚áí}&\dg{\t{$w_{kj}^{(2)}$ and $w_{k'j'}^{(2)}$: }}E=\/{1}{2}\n{y-t}^2\dg{‚áí}\/{‚ąāE}{‚ąāw_{kj}}=\/{‚ąāE}{‚ąāa_k}\/{‚ąāa_k}{‚ąāw_{kj}}=őī_kz_j\\
&\dg{‚áí}\green{M_{kk'}:=\/{‚ąā^2E}{‚ąāa_k‚ąāa_{k'}}=\/{‚ąāőī_k}{‚ąāa_{k'}}}\dg{‚áí}\/{‚ąā^2E}{‚ąāw_{kj}‚ąāw_{k'j'}}=\/{‚ąāőī_k}{‚ąāw_{k'j'}}z_j=\green{\/{‚ąāőī_k}{‚ąāa_{k'}}}\/{‚ąāa_{k'}}{‚ąāw_{k'j'}}z_j\dg{‚áí}\/{‚ąā^2E}{‚ąāw_{kj}^{(2)}‚ąāw_{k'j'}^{(2)}}=\green{M_{kk'}}z_jz_{j'}\\
\dg{‚áí}&\dg{\t{$w_{ji}^{(1)}$ and $w_{j'i'}^{(1)}$: }}\/{‚ąāE}{‚ąāw_{ji}}=\/{‚ąāE}{‚ąāa_j}\/{‚ąāa_j}{‚ąāw_{ji}}=\sum_k\/{‚ąāE}{‚ąāa_k}\/{‚ąāa_k}{‚ąāa_j}\/{‚ąāa_j}{‚ąāw_{ji}}=\sum_k\yello{őī_kh'(a_j)}w_{kj}x_i\dg{ - \t{need to account for $\/{‚ąā\yello{őī_k}}{w_{j'i'}}$, and $\/{‚ąā\yello{h'(a_j)}}{‚ąāw_{j'i'}}$ if $\yello{j=j'}$}}\\
&\dg{‚áí}\/{‚ąā^2E}{‚ąāw_{ji}^{(1)}w_{j'i'}^{(1)}}=x_ix_{i'}h''(a_{j'})\yello{I_{jj'}}\sum_kw_{kj'}^{(2)}őī_k+x_ix_{i'}h'(a_{j'})h'(a_j)\sum_k\sum_{k'}w_{k'j'}^{(2)}w_{kj}^{(2)}\green{M_{kk'}}\dg{ - \t{product rule}}\\
\dg{‚áí}&\dg{\t{$w_{ji}^{(1)}$ and $w_{kj'}^{(2)}$: }}\/{‚ąāE}{‚ąāw_{ji}}=\sum_k\yello{őī_k}h'(a_j)\yello{w_{kj}}x_i\dg{ - \t{need to account for $\/{‚ąā\yello{őī_k}}{‚ąāw_{kj'}}$, and $\/{‚ąā\yello{w_{kj}}}{‚ąāw_{kj'}}$ if $\yello{j=j'}$}}\\
&\dg{‚áí}\/{‚ąā^2E}{‚ąāw_{ji}^{(1)}w_{kj'}^{(2)}}=x_ih'(a_j)őī_k\yello{I_{jj'}}+x_iz_jh'(a_j)\sum_{k'}w_{k'j}^{(2)}\green{M_{kk'}}\dg{ - \t{product rule}}
\EA$

- **Pearlmutter trick** evaluates $v^‚ä§H$ in $O(W)$ without the $O(W^2)$ operation to find $H$ itself.
Operator $\red{\m{R}\{\blue{g}\}:=v^‚ä§‚ąá_w\blue{g}}$ evaluates how much $g$ changes when the network's weights $w$ change by amount $v$.
$\BA
\dg{‚áí}&\BC
1. &\m{R}\{x_i\}=0,\ \m{R}\{t_k\}=0\dg{ - \t{data is treated as a constant wrt change in network's weights}}\\
2. &\m{R}\{w_{ji}\}=v_{ji}\dg{ - v\t{ is the velocity vector corresponding to weights $w$}}\\
3. &\m{R}\{AB\}=\m{R}\{A\}B+A\m{R}\{B\}\dg{ - \t{product rule}}\\
4. &\m{R}\{h(a_j)\}=h'(a_j)\m{R}\{a_j\}\dg{ - \t{chain rule}}
\EC\\
\dg{‚áí}&x‚ąą‚ĄĚ_i^{D+1}‚Üí\block{z_j=h(a_j^{(1)}=w_j^{(1)‚ä§}x)}‚Üíz‚ąą‚ĄĚ_j^{M+1}‚Üí\block{y_k=a_k^{(2)}=w_k^{(2)‚ä§}z}‚Üíy‚ąą‚ĄĚ_k^K\\
&\dg{‚áí}\grass{\m{R}\{z_j\}}=h'(a_j)\m{R}\{a_j\}=h'(a_j)\m{R}\{w_j^‚ä§x\}=h'(a_j)v_j^‚ä§x\\
&\dg{‚áí}\brown{\m{R}\{y_k\}}=\m{R}\{w_k^‚ä§z\}=\m{R}\{w_k\}^‚ä§z+w_k^‚ä§\m{R}\{z\}=v_k^‚ä§z+w_k^‚ä§\m{R}\{z\}=\sum_jv_{kj}z_j+\sum_jw_{kj}h'(a_j)v_j^‚ä§x\\
&\dg{‚áí}őī_k^{(2)}=y_k-t_k\dg{‚áí}\green{\m{R}\{őī_k\}}=\brown{\m{R}\{y_k\}}\\
&\dg{‚áí}őī_j^{(1)}=h'(a_j)\sum_kw_{kj}őī_k^{(2)}\dg{‚áí}\yello{\m{R}\{őī_j\}}=h''(a_j)\m{R}\{a_j\}\sum_kw_{kj}őī_k+h'(a_j)\sum_kv_{kj}őī_k+h'(a_j)\sum_kw_{kj}\brown{\m{R}\{y_k\}}\\
\dg{‚áí}&\dg{\t{backprop: now $v^‚ä§H=\m{R}\{\/{‚ąāE}{‚ąāw_{xx}}\}‚ąą‚ĄĚ^W$ is a function of $v$. Iterating $v=I_l$ for $l=1,...,W$ returns exact Hessian $H$}}\\
&\dg{‚áí}\/{‚ąāE}{‚ąāw_{kj}^{(2)}}=őī_k^{(2)}z_j\dg{‚áí}\m{R}\{\/{‚ąāE}{‚ąāw_{kj}}\}=\green{\m{R}\{őī_k\}}z_j+őī_k\grass{\m{R}\{z_j\}}\\
&\dg{‚áí}\/{‚ąāE}{‚ąāw_{ji}^{(1)}}=őī_j^{(1)}x_i\dg{‚áí}\m{R}\{\/{‚ąāE}{‚ąāw_{ji}}\}=x_i\yello{\m{R}\{őī_j\}}
\EA$


**Hessian inverse approximations**: many algorithms require $H^{-1}$, which is $O(W^3)$.

- **Diagonal approximation**:
Let $w_{ji}$ be a weight in some layer connecting input $z_i$ to activation $a_j$.
True Hessian is strongly non-diagonal, because $w_j$ in one layer could strongly influence downstream $w_k$.
But by assuming different units do not interact, the Hessian is a diagonal matrix and $\/{‚ąā^2E}{‚ąāw_{ji}^2}$ can be solved in $O(W)$ using backpropagation.
Then inverse is also solved in $O(W)$.
$\BA
\dg{‚áí}&\/{‚ąāE}{‚ąāw_{ji}}=\/{‚ąāE}{‚ąāa_j}\/{‚ąāa_j}{‚ąāw_{ji}}=\/{‚ąāE}{‚ąāa_j}z_i\dg{‚áí}\red{\/{‚ąā^2E}{‚ąāw_{ji}^2}=\yello{\/{‚ąā^2E}{‚ąāa_j^2}}z_i^2}\dg{‚áí}\yello{\/{‚ąā^2E}{‚ąāa_j^2}}=h'(a_j)^2\sum_{k=1}^K\sum_{k'=1}^Kw_{kj}w_{k'j}\/{‚ąā^2E}{‚ąāa_k‚ąāa_{k'}}+h''(a_j)\sum_{k=1}^Kw_{kj}\/{‚ąāE}{‚ąāa_k}\\
&\dg{‚áí\t{diagonal $\/{‚ąā^2E}{‚ąāw_{ji}^2}$ requires finding $\yello{\/{‚ąā^2E}{‚ąāa_j^2}}$, which involves cross-interactions between downstream units $k$ and $k'$}}\\
&\dg{‚áí\t{Now assume zero off-diagonal - i.e., assume different units do not interact, then $\/{‚ąā^2E}{‚ąāa_k‚ąāa_{k'}}‚Üí\yello{\/{‚ąā^2E}{‚ąāa_k^2}}$}}\\
\dg{‚áí}&\red{\yello{\/{‚ąā^2E}{‚ąāa_j^2}}=h'(a_j)^2\sum_{k=1}^Kw_{kj}^2\yello{\/{‚ąā^2E}{‚ąāa_k^2}}+h''(a_j)\sum_{k=1}^Kw_{kj}\/{‚ąāE}{‚ąāa_k}}\dg{ - \t{Recursive downstream $\yello{\/{‚ąā^2E}{‚ąāa_k^2}}$ allows Hessian backpropagation}}\\
\EA$

- **Outer product approximation**: iteratively build $H$ and $H^{-1}$ by assuming $\E[y_n]=t_n$.
$\BA
\dg{‚áí}&E=\/{1}{2}\n{y-t}^2\dg{‚áí}‚ąá_wE=\sum_{n=1}^N(y_n-t_n)‚ąá_wy_n\dg{‚áí}‚ąá_w^2E=\sum_{n=1}^N‚ąá_wy_n‚ąá_wy_n^‚ä§+\sum_{n=1}^N(y_n-t_n)‚ąá_w^2y_n\\
&\dg{‚áí\t{assume $\E[y_n-t_n]=0$: }}‚ąá_w^2E‚Čą\sum_{n=1}^N‚ąá_wy_n‚ąá_wy_n^‚ä§\dg{‚áí\t{regression $y_n=a_n$: }}\red{H‚Čą\sum_{n=1}^Nb_nb_n^‚ä§,\ b_n=‚ąá_wa_n}\\
\dg{‚áí}&E=-\sum_{n=1}^N[t_n\ln y_n+(1-t_n)\ln(1-y_n)]\dg{‚áí}‚ąá_wE=\sum_{n=1}^N\/{y_n-t_n}{y_n(1-y_n)}‚ąá_wy_n=\sum_{n=1}^N(y_n-t_n)‚ąá_wa_n\dg{ - ‚ąá_wy_n=y_n(1-y_n)‚ąá_wa_n}\\
&\dg{‚áí}‚ąá_w^2E=\sum_{n=1}^N[‚ąá_wy_n‚ąá_wa_n+(y_n-t_n)‚ąá_w^2a_n]\dg{‚áí\t{assume $\E[y_n-t_n]=0$: }}\red{H‚Čą\sum_{n=1}^Ny_n(1-y_n)b_nb_n^‚ä§,\ b_n=‚ąá_wa_n}\\
\dg{‚áí}&\dg{\t{Woodbury inversion lemma: $(ő£_{22}-ő£_{21}ő£_{11}^{-1}ő£_{12})^{-1}=ő£_{22}^{-1}+ő£_{22}^{-1}ő£_{21}őõ_{11}ő£_{12}ő£_{22}^{-1},\ őõ_{11}=(ő£_{11}-ő£_{12}ő£_{22}^{-1}ő£_{21})^{-1}$}}\\
&\dg{‚áí\t{Iterative: }}\blue{H_{L+1}=H_L+b_{L+1}b_{L+1}^‚ä§}\dg{ - \t{$L$ points processed already}}\dg{‚áí}\BM 1&b_{L+1}^‚ä§\\b_{L+1}&H_L\EM\dg{‚áí}\blue{H_{L+1}^{-1}=H_L^{-1}+\/{H_L^{-1}b_{L+1}b_{L+1}^‚ä§H_L^{-1}}{1+b_{L+1}^‚ä§H_L^{-1}b_{L+1}}}\\
&\dg{‚áí}H_0=ŌĶI\dg{ - \t{for a small $ŌĶ$ we can iteratively build $(H+H_0)^{-1}=(H+ŌĶI)^{-1}$}}
\EA$

  - Efficiency: finding output activation derivative $b_n=‚ąá_wa_n$ is $O(W)$ by backprop.
  Constructing $b_nb_n^‚ä§$ is $O(W^2)$.
  Inverse numerator and denominator is each $O(W^2)$.
  Constructing $H$ and $H^{-1}$ is each $O(NW^2)$.

- **Numerical differences**: numerical differences model costs $O(W^3)$. A gradient backprop model costs $O(W^2)$.
Exact Hessian is almost always used instead for verifying implementation, but numerical differences are foolproof.
$\BA
\dg{‚áí}\/{‚ąā^2E}{‚ąāw_{ji}‚ąāw_{lk}}=\/{1}{4ŌĶ^2}[E(w_{ji}+ŌĶ,w_{lk}+ŌĶ)+E(w_{ji}-ŌĶ,w_{lk}-ŌĶ)-E(w_{ji}+ŌĶ,w_{lk}-ŌĶ)-E(w_{ji}-ŌĶ,w_{lk}+ŌĶ)]\dg{ - \t{each entry }O(W)}\\
\dg{‚áí}\/{‚ąā^2E}{‚ąāw_{ji}‚ąāw_{lk}}=\/{1}{2ŌĶ}[\/{‚ąāE}{‚ąāw_{ji}}(w_{lk}+ŌĶ)-\/{‚ąāE}{‚ąāw_{ji}}(w_{lk}-ŌĶ)]+O(ŌĶ^2)\dg{- \t{each column of $H$ is filled by 2 backprop runs $\/{‚ąāE}{‚ąāw}(w_{lk}¬ĪŌĶ)$}}
\EA$

**Mixture density networks**: in forward problems features $x‚ąą‚ĄĚ^D$ represent a configuration of factors that lead to one outcome represented by unimodal target $p(t|x)$.
In inverse problems $x‚ąą‚ĄĚ^D$ represents an outcome, and many configurations that may have lead to this outcome are represented by multimodal target $p(t|x)$.
$\BA
\dg{‚áí}&p(t|x)=\sum_k\yello{ŌÄ_k(x)}\Normal(t|\yello{őľ_k(x)},\yello{ŌÉ_k^2(x)})\dg{ - \t{3 groups of output activation functions}}\\
&\dg{‚áí}ŌÄ_k(x)=\/{\e{a_k^ŌÄ(x)}}{\sum_{l=1}^K\e{a_l^ŌÄ(x)}}\dg{ - \t{softmax activation function output for $K$ priors}}\\
&\dg{‚áí}őľ_k(x)=a_k^őľ(x)\dg{ - \t{linear regression activation function output for $K$ means}}\\
&\dg{‚áí}ŌÉ_k(x)=\e{a_k^ŌÉ(x)}\dg{ - \t{exponential activation function output for standard deviations}}\\
\dg{‚áí}&\dg{\t{Exponential activation: }}\blue{y_k=\e{a_k}}\dg{ - \t{}}\\
&\dg{‚áí}p(t|y)=\/{y^te^{-y}}{t!}\dg{‚áí}E(y)=-\ln p(t|y)=-(t\ln y-y-\ln(t!))=-t\ln y+y
\EA$
