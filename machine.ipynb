# PRML: 1 Introduction

**Gaussian**: $\Normal(x_i|μ,Σ)=\/{1}{(2π)^{D/2}\det(Σ)^{1/2}}\e{-\/{1}{2}(x_i-μ)^⊤Σ^{-1}(x_i-μ)}$.

- $\Normal(x_i|μ,σ^2)=\/{1}{(2πσ^2)^{1/2}}\e{-\/{1}{2σ^2}(x_i-μ)^2}$. Dataset likelihood $p(x|μ,σ^2)=\prod_{i=1}^N\Normal(x_i|μ,σ^2)$

- Log-likelihood $\ln p(x|μ,σ^2)=-\/{N}{2}\ln(2π)-\/{N}{2}\ln(σ^2)-\/{1}{2σ^2}\sum_i(x_i-μ)^2$

  - Maximum likelihood estimators $μ_\t{ML}=\/{1}{N}\sum_ix_i$ and $σ_\t{ML}^2=\/{1}{N}\sum_i(x_i-μ_\t{ML})^2$. MLE variance $σ_\t{ML}^2$ is biased with correction factor $\/{N}{N-1}$.

**Baye's Theorem**: posterior ∝ likelihood × prior.
$p(w|X)=\/{p(X|w)p(w)}{p(X)},\ p(X)=∫p(X|w)p(w)\ dw$. Historically limited by difficulty of marginalizing $p(X)$, but has changed with MCMC sampling, variational Bayes, and expectation propagation.

- **Curve fitting** $t∈ℝ^N,x∈ℝ^{N×D}$ with polynomial basis coefficients $w∈ℝ^{M+1}$, predictor function $\E[t_i]=y(x_i,w)$ and precision $β=\/{1}{σ^2}$, we have ML estimator
$β_\t{ML}^{-1}=\/{1}{N}\sum_i(t_i-y(x_i,w_\t{ML}))^2$

  - Predictive distribution
  $p(t_i|x_i,w_\t{ML},β_\t{ML})=\Normal(t_i|y(x_i,w_\t{ML}),β_\t{ML}^{-1})$.
  Let prior $p(w|α)=\Normal(w|0,α^{-1}I)$.
  Then posterior
  $p(w|x,t,α,β)∝p(t|x,w,β)p(w|α)$

  - Log-posterior
  $\ln p(w|t)=\sum_i\ln\Normal(t_i|y(x_i,w_\t{ML}),β_\t{ML}^{-1})+\ln\Normal(w|0,α^{-1}I)$
  $=-\/{β}{2}\sum_i(t_i-y(x_i,w))^2-\/{α}{2}w^⊤w$
  regularized least-squares to get $w_\t{MAP}$.

- **Full Bayesian curve fitting**: Marginalizing $w$ in posterior prediction lies in the heart of Bayesian methods for pattern recognition.
Prior $p(w|α)=\Normal(w|0,α^{-1}I)$ and likelihood $p(t_i|x_i,w,β)=\Normal(t_i|ϕ(x_i)^⊤w,β^{-1})$.

  - Dataset posterior:
  $p(w|t)∝p(t|w,β)p(w,α)$
  $=\prod_i\Normal(t_i|ϕ(x_i)^⊤w,β^{-1})\Normal(w|0,α^{-1}I)$.
  
  - Log-posterior: complete squares
  $\ln p(w|t)=-\/{β}{2}\sum_i(t_i-ϕ(x_i)^⊤w)^2-\/{α}{2}w^⊤w+c_1$
  $=-\/{β}{2}\sum_i(w^⊤ϕ(x_i)ϕ(x_i)^⊤w-2t_iϕ(x_i)^⊤w)-\/{α}{2}w^⊤w+c_2$
  $=-\/{1}{2}w^⊤(\purple{αI+β\sum_iϕ(x_i)ϕ(x_i)^⊤})w+\purple{β\sum_it_iϕ(x_i)}^⊤w+c_2$
  $=-\/{1}{2}(w-m)^⊤S^{-1}(w-m)+c_2$
  $\dg{⇒}p(w|t)=\Normal(w|m,S)$
  $\dg{⇒}\green{S^{-1}=αI+β\sum_iϕ(x_i)ϕ(x_i)^⊤},\ \green{m=βS\sum_it_iϕ(x_i)}$.

  - Posterior prediction:
  $p(t_i|x_i,t)=∫p(t_i|x_i,w)p(w|t)\ dw$
  $=∫\Normal(t_i|ϕ(x_i)^⊤w,β^{-1})\Normal(w|m,S)\ dw$.
  By Adam and Eve
  $\E[t_i]=\E[\E[t_i|w]]$
  $=ϕ(x_i)^⊤m$ and
  $\V[t_i]=\E[\V[t_i|w]]+\V[\E[t_i|w]]$
  $=β^{-1}+ϕ(x_i)^⊤Sϕ(x_i)$.
  Therefore
  $\blue{p(t_i|x_i,t)=\Normal(βϕ(x_i)^⊤S\sum_it_iϕ(x_i),β^{-1}+ϕ(x_i)^⊤Sϕ(x_i))}$

**Decision theory**: Let $C_1,C_2$ be presence/absence of cancer and $x_i$ be imaging. Then posterior $p(C_k|x_i)=\/{p(x_i|C_k)P(C_k)}{P(x_i)}$.
Input space is divided into decision regions $\m{R}_k$, that dictate $x_i∈\m{R}_k$ are assigned $C_k$.

- **Joint → Posterior**: Equivalent $p(x_i,C_1)>p(x_i,C_2)$
$\dg{⇒}p(C_1|x_i)p(x_i)>p(C_2|x_i)p(x_i)$
$\dg{⇒}\red{p(C_1|x_i)>p(C_2|x_i)}$

  - Inference: using training data to learn $p(C_k|x_i)$. Decision: use posteriors to make optimal classifications.

- Minimize probability of error $p(x_i∈\m{R}_1,C_2)+p(x_i∈\m{R}_2,C_1)$
$=∫_{\m{R}_1}p(x_i,C_2)\ dx_i+∫_{\m{R}_2}p(x_i,C_1)\ dx_i$.
Number of terms $K(K-1)$.
Equivalently maximize probability of correct $\sum_kp(x_i∈\m{R}_k,C_k)$
$=\sum_k∫_{\m{R}_k}p(x_i,C_k)\ dx_i$

  - Type 1: false positive. Type 2: false negative.

- **Expected loss** $\blue{\E[L]=\sum_{k,j}∫_{\m{R}_j}L_{kj}p(x_i,C_k)\ dx_i}$ where $L_{kj}$ is loss from actual $C_k$ being misclassified to $\m{R}_j$.

  - Drop the integral: classify $x_i$ to $\m{R}_j$ satisfying $\arg\min_j\red{\sum_kL_{kj}p(C_k|x_i)}$ to minimize average loss

  - **Reject option**: if the highest $p(C_k|x_i)< θ$ for some threshold $θ$, then refuse to make a prediction. $θ< \/{1}{K}$ ensures no rejection, while $θ=1$ rejects everything.

- **Generative models**: determine likelihood $p(x_i|C_k)$ and prior $p(C_k)$, then find joint $p(x_i,C_k)$ and posterior $p(C_k|x_i)$. Having joint distribution allows generation of sample data points.

  - **Discriminant models** use a deterministic mapping.

- **Curve fitting**: Let $y(x_i)$ be model predictor with loss $L(t_i,y(x_i))$.
Then $\blue{\E[L]=\iint L(t_i,y(x_i))p(x_i,t_i)\ dx_i\ dt_i}$ is expected loss.
$ℓ_2$ loss $\red{\E[L]=\iint (y(x_i)-t_i)^2p(x_i,t_i)\ dx_i\ dt_i}$

  - Minimize
  $\E[L]=∫[∫(y(x_i)-t_i)^2p(t_i|x_i)\ dt_i]p(x_i)\ dx_i$:
  minimize the inner integral and apply Leibniz rule
  $\BC
  ∂_{y(x_i)}∫(y(x_i)-t_i)^2p(t_i,x_i)\ dt_i
  =∫∂_{y(x_i)}(y(x_i)-t_i)^2p(t_i,x_i)\ dt_i
  =2∫(y(x_i)-t)p(x_i,t_i)\ dt_i=0\\
  \dg{⇒}∫y(x_i)p(x_i,t_i)\ dt_i-∫tp(x_i,t_i)\ dt_i=0
  \dg{⇒}y(x_i)p(x_i)=∫tp(x_i,t_i)\ dt_i
  \dg{⇒}y^*(x_i)=∫t\/{p(x_i,t_i)}{p(x_i)}\ dt_i
  =∫tp(t_i|x_i)\ dt_i
  \EC$
  $\green{\dg{⇒}y(x_i)^*=\E[t_i|x_i]}$ is the optimal model that minimizes expected loss.

  - $\BC
  \E[L]=\iint (y(x_i)-t_i)^2p(x_i,t_i)\ dt_i\ dx_i
  =∫(y(x_i)-\E[t_i|x_i])^2\ob{∫p(x_i,t_i)\ dt_i}{p(x_i)}\ dx_i\\
  +∫\ob{∫(t_i-\E[t_i|x_i])^2p(t_i|x_i)\ dt_i}{\V[t_i|x_i]}p(x_i)\ dx_i
  +2∫(y(x_i)-\E[t_i|x_i])p(x_i)\ob{∫(\E[t_i|x_i]-t_i)p(t_i|x_i)\ dt_i}{\E[\E[t_i|x_i]|x_i]-\E[t_i|x_i]=0}\ dx_i\\
  \dg{⇒}\green{\E[L]=\E_{x_i}[(y(x_i)-\E_{t_i}[t_i|x_i])^2]+\E_{x_i}[\V_{t_i}[t_i|x_i]]}
  \EC$
  $\dg{⇒}\red{\E[L]=∫(y(x_i)-\E[t_i|x_i])^2p(x_i)\ dx_i+\iint(t_i-\E[t_i|x_i])^2p(t_i|x_i)p(x_i)\ dx_i\ dt_i}$.
  First term is model error, which is zero when $y(x_i)=\E[t_i|x_i]$.
  Second term is residual noise error, which does not depend on model $y(x_i)$ and is irreducible.

**Information theory**: information $\blue{h(x)=-\log_2 p(x)}$ is equivalent to surprise and uncertainty, which holds an inverse relationship with probability.
If $x,y$ are independent, then $h(x,y)=h(x)+h(y)$.

- **Discrete entropy**: $\blue{\H[x]=-\sum_xp(x)\log_2 p(x)}=\E[h(x)]$ is average information. Its unit is bits.

  - **8-state proof**: Suppose $x$ has 8 possible states (A-H).
  If uniform $p(x)=\/{1}{8}$ then $\H[x]=-\sum_x\/{1}{8}(-3)=3$ bits binary coding 001,010,011,100,101,110,111.
  If non-uniform distribution $[\/{1}{2},\/{1}{4},\/{1}{8},\/{1}{16},\/{1}{64},\/{1}{64},\/{1}{64},\/{1}{64}]$ then $\H[x]=\/{1}{2}(1)+\/{1}{4}(2)+\/{1}{8}(3)+\/{1}{16}(4)+\/{4}{64}(6)=2$ bits using variable coding 0,10,110,1110,111110.
  Entropy is the absolute lower bound on average number of bits to transmit the state of a variable.

  - **Thermodynamics**: In Maxwell-Boltzmann statistics, gas molecules are distinguishable.
  Number of ways to distribute $N$ particles into bins of size $n_1+...+n_k=N$ is called multiplicity $W=\/{N!}{\prod_in_i!}$.
  Using Stirling approximation ($\ln n!=n\ln n-n$)
  we have
  $\ln W=\ln(\/{N!}{\prod_in_i!})$
  $=\ln N!-\sum_i\ln n_i!$
  $≈N\ln N-N-\sum_in_i\ln n_i+\sum_in_i$
  $=N\ln N-\sum_in_i\ln n_i$
  $=\sum_in_i\ln N-\sum_in_i\ln n_i$
  $=-\sum_in_i\ln(\/{n_i}{N})$.
  Then we divide by $N$ to find entropy per particle
  $\lim\limits_{N→∞}\/{1}{N}\ln W=-\sum_ip_i\ln(p_i)$
  $\dg{⇒}\red{H=-\sum_ip_i\ln(p_i)}$

  - **Maximum entropy** has uniform distribution.
  $\BC\t{minimize}_p&\sum_ip_i\ln(p_i)\\
  \t{subject to}&p≽0,\v{1}^⊤p=1
  \EC$
  $\dg{⇒}L=\sum_ip_i\ln(p_i)+ν(\v{1}^⊤p-1)$
  $\dg{⇒}g(ν)=\inf_p(\sum_ip_i\ln(p_i)+ν\sum_ip_i-ν)$
  $=-ν+\inf_p\sum_i(νp_i+p_i\ln(p_i))$.
  Minimize
  $∂_{p_i}=ν+\ln(p_i)+1=0$
  $\dg{⇒}p_i^*=e^{-ν-1}\ ∀i$
  is a constant (uniform).

- **Differential entropy**: Mean value theorem says the area under PDF for the $i$-th bin has mean value $p(x_i)$ such that the probability is $p(x_i)Δ$ satisfying $\sum_ip(x_i)Δ=1$.
We quantize
$H_Δ=-\sum_ip(x_i)Δ[\ln p(x_i)+\ln Δ]$
$=-\sum_ip(x_i)Δ\ln p(x_i)-\ln Δ$
$\dg{⇒}\lim\limits_{Δ→0}H_Δ=-∫p(x)\ln p(x)\ dx+∞$.
We discard the infinity term and use the first term as differential entropy $\red{\H[x]=-∫p(x)\ln p(x)\ dx},\ x∈ℝ^D$.

  - **Maximum entropy** for PDF $(μ,σ^2)$:
  $\BC\t{maximize}_{p(x)}&-∫p(x)\ln p(x)\ dx\\
  \t{subject to}&∫p(x)\ dx=1,\ ∫xp(x)\ dx=μ,\ ∫(x-μ)^2p(x)\ dx=σ^2
  \EC$
  $L(p)=-∫p\ln(p)\ dx+ν_1(∫p\ dx-1)+ν_2(∫xp\ dx-μ)+ν_3(∫(x-μ)^2p\ dx-σ^2)$
  $=-∫(p\ln(p)+ν_1p+ν_2xp+ν_3(x-μ)^2p)\ dx+c_1$
  $\dg{⇒}∂_pL(p)=0$
  $\dg{⇒}∂_p(-p\ln(p)+ν_1p+ν_2xp+ν_3(x-μ)^2p)=0$
  $\dg{⇒}-\ln(p)-1+ν_1+ν_2x+ν_3(x-μ)^2=0$
  $\dg{⇒}p(x)=\e{-1+ν_1+ν_2x+ν_3(x-μ)^2}$.
  By integrating $p(x)$ against the three constraints, it turns out
  $\blue{p(x)=\/{1}{\sqrt{2π}σ}e^{-(x-μ)^2/2σ^2}}$
  maximizes differential entropy with
  $\blue{\H[\Normal(x|μ,σ^2)]=\/{1}{2}[1+\ln(2πσ^2)]}$.
  Differential entropy is negative when
  $σ^2< \/{1}{2πe}$.

  - **Conditional entropy**: how much uncertainty is left?
  $\H[y|x=x_0]=-∫p(y|x=x_0)\ln p(y|x=x_0)\ dy$
  $\dg{⇒}\H[y|x]=-\iint p(y|x=x_0)\ln p(y|x=x_0)\ dy\ p(x_0)\ dx_0$
  $\dg{⇒}\red{\H[y|x]=-\iint p(x,y)\ln p(y|x)\ dy\ dx}$

  - **Conditional entropy chain rule**: $\red{\H[x,y]=\H[y|x]+\H[x]}$.
  Proof:
  $\H[x,y]=-\iint p(x,y)\ln p(x,y)\ dy\ dx$
  $=-\iint p(x,y)[\ln p(y|x)+\ln p(x)]\ dy\ dx$
  $=-\iint p(x,y)\ln p(y|x)\ dy\ dx-∫ p(x)\ln p(x)\ dx$.

**Kullback-Leibler Divergence**: Suppose $p(x)$ is the true probability of 'A-Z', but $q(x)$ is what the model assumes. KL divergence is the measure of inefficiency that arise from mismatch between the model $q(x)$ and reality $p(x)$.
$\red{\KL(p\|q)=∫p(x)\ln\/{p(x)}{q(x)}\ dx}$
$=-∫p(x)\ln q(x)\ dx-(-∫p(x)\ln p(x)\ dx)$ matches reality $p$ to model $q$.

- $\blue{\KL(p\|q)\neq\KL(q\|p)}$:

  - **Forward KL**: $\KL(p\|q)=∫p(x)\ln\/{p(x)}{q(x)}\ dx$ is expectation wrt true distribution $p(x)$.
  In regions where $q(x)≈0$ and $p(x)>0$, KL blows up. Loss minimization forces $q(x)$ to stretch out and avoid zeros in all regions where true distribution has probability mass.
  This **zero-avoiding** behavior overestimates variance.
  If $p(x)$ is a 2-peak GMM, $q(x)$ tries to position its peak at the mean of the two modes (**mean seeking**).

  - **Reverse KL**: $\KL(q\|p)=∫q(x)\ln\/{q(x)}{p(x)}\ dx$ wants to avoid placing probability mass where true distribution $p(x)≈0$, therefore it is **zero-forcing** for most of its distribution thereby creating a narrow distribution.
  If $p(x)$ is a 2-peak GMM, $q(x)$ tries to position its peak at one mode and zero everywhere else (**mode seeking**).

- $\blue{\KL(p\|q)≥0}$.
Proof: Let $f(y)=-\ln\/{q(x)}{p(x)}$.
Then by Jensen's inequality
$\KL(p\|q)=\E_p[f(y)]≥f(\E_p[y])$
$=-\ln ∫p(x)\/{q(x)}{p(x)}\ dx$
$=-\ln ∫q(x)\ dx$
$=0$.

- **Density estimation**: $\KL(p\|q)=-∫p(x)\ln\/{q(x|θ)}{p(x)}\ dx$
$=-∫p(x)\ln q(x|θ)\ dx+∫p(x)\ln p(x)\ dx$.
We want to tweak $θ$ to make our model $q(x|θ)$ as close as possible to unknown true distribution $p(x)$.
Using empirical approximation: $∫p(x)f(x)\ dx≈\/{1}{N}\sum_if(x_i)$
$\dg{⇒}\red{\KL(p\|q)≈\/{1}{N}\sum_i[-\ln q(x_i|θ)+\ln p(x_i)]}$.
Therefore optimization
$\arg\min_θ\KL(p\|q)$
$≈\arg\min_θ\sum_i-\ln q(x_i|θ)$
$≈\arg\max_θ\sum_i\ln q(x_i|θ)$ is exactly MLE and log-likelihood!

- **Mutual information**:
$\red{\I[x,y]:=\KL(p(x,y)\|p(x)p(y))}$
$=-\iint p(x,y)\ln\/{p(x)p(y)}{p(x,y)}\ dx\ dy$
$\dg{⇒}\red{\I[x,y]=\H[x]-\H[x|y]}$.
It is zero if variables are truly independent.
Remove from $\H[x]$ the initial information about $x$
by $\H[x|y]$ the remaining information about $x$ after learning $y$ gives the mutual information.

  - Proof: $\I[x,y]=-\iint p(x,y)\ln\/{p(x)p(y)}{p(x,y)}\ dx\ dy$
  $=-\iint p(x,y)[\ln p(x)+\ln p(y)-\ln p(x,y)]\ dx\ dy$
  $=-∫p(x)\ln p(x)\ dx-∫p(y)\ln p(y)\ dy+\iint p(x,y)\ln p(x,y)\ dx\ dy$
  $=\H[x]+\H[y]-\H[x,y]$

# Book1: 5 Decision Theory; 6 Information Theory

**Fundamentals**: Let $a∈\m{A}$ be possible actions, $h∈\m{H}$ be true underlying states, then $l(h,a)$ is the loss function incurred by taking action $a$ when the true underlying state is $h$.

- $\red{ρ(a|x):=\E_{p(h|x)}[l(h,a)]=\sum_{h∈\m{H}}l(h,a)p(h|x)}$ is the expected loss for taking action $a$ given observation $x$ and posterior $p(h|x)$.
The best course of action is **Bayes estimator**
$\red{π^*(x)=\arg\min_{a∈\m{A}}\E_{p(h|x)}[l(h,a)]}$.

  - Let $U(h,a)=-l(h,a)$ be utility function, then
  $π^*(x)=\arg\max_{a∈\m{A}}\E_{p(h|x)}[U(h,a)]$.

**Classification**: in zero-one classification problem where input $x_i$ predicts label $\hat{y}$ against true label $y^*$, the loss function is
$l(y^*,\hat{y})=I(y^*\neq\hat{y})$
and expected loss is
$ρ(\hat{y}|x)=\E_{p(y|x)}[I(y^*\neq\hat{y})]$
$=1-p(y^*=\hat{y}|x)$.
Bayes estimator is the mode of posterior or MAP estimator
$π(x)=\arg\min_{y∈\m{Y}}\E_{p(y|x)}[I(y^*\neq\hat{y})]=\arg\max_{y∈\m{Y}}p(y|x)$

- Let $l_{ij}$ be loss incurred when predicting $j$ while true class is $i$, and $p_0+p_1=1$ be probability that true class is 0,1.
Then we should predict $\hat{y}=0$ if
$l_{00}p_0+l_{10}p_1< l_{01}p_0+l_{11}p_1$
$\dg{⇒}p_1< \/{l_{01}}{l_{01}+l_{10}}$

- Reject option

**ROC curves**: Let policy $τ$ be decision threshold for prediction $\red{\hat{y}_τ(x)=I(p(y=1|x)≥1-τ)}$.
From this policy we count $FP_τ=\sum_iI(\hat{y}_τ(x_i)=1,y_i=0)$, $TP_τ$, $FN_τ$, and $FP_τ$
and derive ratios
$TPR_τ=p(\hat{y}=1|y=1)=\/{TP_τ}{TP_τ+FN_τ}$, $TNR_τ$, $FPR_τ$ (type 1), and $FNR_τ$ (type 2).
The **receiver operating characteristic** curve is a plot of TPR vs FPR.
Ideal curve hugs the top-left corner.

- **Precision-recall curves**: When a search engine returns 1000 results, the top ones are TP related to query, but the number of TN is unknown.
When TN is not well defined, FPR is replaced with **precision** $P(τ)=\/{TP_τ}{TP_τ+FP_τ}=\/{\sum_iy_i\hat{y}_i}{\sum_i\hat{y}_i}$ and TPR is replaced with **recall** $R(τ)=\/{TP_τ}{TP_τ+FN_τ}=\/{\sum_iy_i\hat{y}_i}{\sum_iy_i}$.
Ideal curve hugs the top-right corner.

  - $F_β$-score

- Class imbalance of TN and TP (millions of irrelevant search results vs few useful ones) reduce the practical usefulness of ROC curves. PR curves are highly sensitive to it.

Huber $l_δ(h,a)=\BC
r^2/2&\t{if }|r|≤δ\\
δ|r|-δ^2&\t{if }|r|>δ
\EC$.

**Frequentist decision theory**: unknown but fixed true underlying state is $θ$ instead of $h∈\m{H}$. Let $\hat{θ}=δ(x)$ be the estimator for $θ$. Then MSE risk is $R(θ,δ)=\E_{p(x|θ)}[l(θ,δ(x))]$ where $l(θ,δ(x))=(θ-δ(x))^2$.
MSE can be decomposed $\t{MSE}(\hat{θ}|θ)=\V[\hat{θ}]+\t{bias}(\hat{θ})^2$.

**Entropy**: If $p$ has high entropy and a data source generates $x_i∼p$, then $x_i$ is highly unpredictable. Observing actual data provides a lot of new information. Therefore high entropy = high information content.

- **Discrete**: $\red{\H(X)=-\sum_xp(X=x)\log_2 p(X=x)=-E_X[\log_2 p(X)]}$.
Calculating true entropy of a distribution with a massive number of states (e.g., dictionary) requires immense amounts of data.

  - $\BC
  \t{min entropy}&\H(X)≥0&\t{dirac-delta }p(X=x)=I(X=c)\\
  \t{max entropy}&\H(X)≤\log(K)&\t{uniform }p(X=k)=\/{1}{K}
  \EC$.
  Proof (max): Let $u(x)=\/{1}{K}$ be uniform PDF.
  Then $\KL(p\|u)=\sum_xp(x)\ln\/{p(x)}{u(x)}$
  $=\sum_xp(x)\ln p(x)-\sum_xp(x)\ln u(x)$
  $=-\H(x)-\ln(\/{1}{K})≥0$.

  - $\H(p)$ means entropy of the $p$ probability vector $\red{\H(p)=\sum_kp_k\ln p_k}$

- **Cross entropy**: $\red{\H(p,q)=-\sum_kp_k\ln q_k}$

- **Joint entropy**: $\red{\H(X,Y)=-\sum_{x,y}p(x,y)\ln_2 p(x,y)}$ is simultaneous uncertainty across multiple variables.

  - $\BC
  \t{min joint entropy}&\H(X,Y)≥\max\{\H(X),\H(Y)\}≥0&\t{joint uncertainty ≥ most uncertain element}\\
  \t{max joint entropy}&\H(X,Y)≤\H(X)+\H(Y)&\t{equal when $X$ and $Y$ are independent}
  \EC$

  - For states $n∈\{1,...,8\}$ and variables $X=I(n\%2=0)$ and $Y=I(n\t{ prime})$, then $\E[X]=\E[Y]=\/{1}{2}$ and $\H(X)+\H(Y)=2$ bits.
  $\H(X,Y)=-(\/{1}{8}\log_2 \/{1}{8}+\/{3}{8}\log_2 \/{3}{8}+\/{3}{8}\log_2 \/{3}{8}+\/{1}{8}\log_2 \/{1}{8})=1.81$ bits.
  Variables $X,Y$ are not independent: knowing one gives information about the other. Joint entropy is lower than the sum.

- **Conditional entropy**: $\red{\H(Y|X)=\E_{p(x)}[\H(p(Y|X))]=-\sum_xp(x)\sum_yp(y|x)\ln p(y|x)}$
$=-\sum_{x,y}p(x,y)\ln \/{p(x,y)}{p(x)}$
$=-\sum_{x,y}p(x,y)\ln p(x,y)+\sum_{x,y}p(x,y)\ln p(x)$
$\dg{⇒}\red{\H(Y|X)=\H(X,Y)-\H(X)}$

  - **Chain rule** for entropy: $\blue{H(X_1,...,X_n)=\sum_iH(i|X_1,...,X_{i-1})}$

  - $\blue{\H(Y|X)≤\H(Y)}$

- **Perplexity** of LLM with $K$ words is $\t{PP}(p)=2^{\H(p)}$ is the branching factor in predicting the next word in a sentence.
$\t{PP}(p_\m{D},p)=2^{\H(p_\m{D},p)}$ evaluates how well model $p$ predicts distribution of words in dataset $p_\m{D}$.

- **Differential entropy**: $\red{\H(X)=-∫p(x)\ln p(x)\ dx=\E_p[\ln p(x)]}$

  - $\H(\Normal(μ,Σ))=\/{D}{2}\ln(2π)+\/{1}{2}\ln\det(Σ)+\/{1}{2}\E[(x-μ)^⊤Σ^{-1}(x-μ)]$
  where
  $\E[(x-μ)^⊤Σ^{-1}(x-μ)]$
  $=\E[\tr((x-μ)^⊤Σ^{-1}(x-μ))]$
  $=\tr(Σ^{-1}\E[(x-μ)(x-μ)^⊤])$
  $=\tr(Σ^{-1}Σ)$
  $=D$.
  Then
  $\H(\Normal(μ,Σ))=\/{D}{2}(\ln(2π)+1)+\/{1}{2}\ln\det(Σ)$
  $\dg{⇒}\red{\H(\Normal(μ,Σ))=\/{1}{2}\ln((2πe)^D\det(Σ))}$

  - We often calculate differential entropy by binning the distribution into a histogram from $\min(x)$ to $\max(x)$.

**Kullback-Leibler divergence**: $\red{\KL(p\|q)=\sum_kp_k\ln\/{p_k}{q_k}=-\H(p)+\H(p,q)}$ and $\KL(p\|q)=∫p(x)\ln\/{p(x)}{q(x)}\ dx$.
Distance requires Abelian equality and triangle inequality.
Divergence only requires non-negativity and zero with identical operands.

- **Gaussians**: $\KL(\Normal(x|μ_1,Σ_1)\|\Normal(x|μ_2,Σ_2))$
$=\/{1}{2}\E_{μ_1,Σ_1}[\ln\/{\det(Σ_2)}{\det(Σ_1)}-\tr((x-μ_1)^⊤Σ_1^{-1}(x-μ_1))$
$+(x-μ_2)^⊤Σ_2^{-1}(x-μ_2)]$
$=\/{1}{2}\ln\/{\det(Σ_2)}{\det(Σ_1)}-\/{D}{2}+\/{1}{2}\E_{μ_1,Σ_1}[(x-μ_2)^⊤Σ_2^{-1}(x-μ_2)]$
$=\/{1}{2}\ln\/{\det(Σ_2)}{\det(Σ_1)}-\/{D}{2}+\/{1}{2}[\tr(Σ_2^{-1}Σ_1)+(μ_2-μ_1)^⊤Σ_2^{-1}(μ_2-μ_1)]$
$\dg{⇒}\KL(\Normal(x|μ_1,σ_1^2)\|\Normal(x|μ_2,σ_2^2))=\/{1}{2}\ln\/{σ_2^2}{σ_1^2}-\/{1}{2}+\/{σ_1^2+(μ_1-μ_2)^2}{2σ_2^2}$

- $\KL(p\|q)≥0$ Proof: Jensen's inequality $\KL(p\|q)=-\sum_kp_k\ln\/{q_k}{p_k}≥\ln(\sum_kp_k\/{q_k}{p_k})=0$

- **Minimum $\KL(p_\m{D}\|q)$ = maximum likelihood**: Let $p_\m{D}(x)=\/{1}{N}\sum_iI(x-x_i)$ gives a histogram counter representing empirical distribution.
By delta-dirac convolution,
$q^*=\arg\min_q\KL(p_\m{D}\|q)$
$=-\arg\min_q∫p_\m{D}(x)\ln q(x)\ dx$
$=-\arg\min_q∫[\/{1}{N}\sum_iI(x-x_i)]\ln q(x)\ dx$
$=-\arg\min_q\/{1}{N}\sum_i\ln q(x_i)$
$\dg{⇒}θ^*=\arg\max_θ\sum_i\ln q(x_i|θ)$.

  - **Cross entropy**: $\KL(p\|q)=\sum_yp(y)\ln\/{p(y)}{q(y)}=-\H(p)+\H(p,q)$
  $\dg{⇒}\red{\arg\min_q\KL(p\|q)=\arg\min_q\H(p,q)}$ heavily penalizes wrong predictions for rare events. Brier score is an alternative: $\/{1}{N}\sum_y(q(y)-p(y))^2$

  - **Log loss**: If one-hot $p(y)=I(y=c)$, then
  $\H(I(y=c),q)=-\sum_yI(y=c)\ln q(y)$
  $=-\ln q(c)$.

- **Forward KL** (zero-avoiding): $\KL(p\|q)=∫p(x)\ln\/{p(x)}{q(x)}\ dx$ forces $q(x)$ to stretch out and cover anywhere $p(x)>0$.

- **Reverse KL** (zero-forcing): $\KL(q\|p)=∫p(x)\ln\/{q(x)}{p(x)}\ dx$ seeks to replicate the mode of $p(x)$.


**Mutual information**: $\red{\I(X,Y)=\KL(p(x,y)\|p(x)p(y))=\sum_{y,x}p(x,y)\ln\/{p(x,y)}{p(x)p(y)}}≥0$

- $\blue{\BC
\I(X,Y)=\H(X)-\H(X|Y)=\H(Y)-\H(Y|X)\\
\I(X,Y)=\H(X,Y)-H(X|Y)-\H(Y|X)\\
\I(X,Y)=\H(X)+\H(Y)-H(X,Y)
\EC}$
$\dg{⇒}\blue{\BC
\I(X,Y)&\H(X∩Y)\\
\H(X,Y)&\H(X∪Y)\\
\H(X|Y)&\H(X\backslash Y)
\EC}$ by Venn diagram.

  - $\BC
  \t{min}&\I(X,Y)≥0\\
  \t{max}&\I(X,Y)≤\min(\H(X),\H(Y))
  \EC$

  - **Normalized mutual information**: $\t{NMI}(X,Y)=\/{\I(X,Y)}{\min(\H(X),\H(Y))}≤1$. **Maximal information coefficient** is for quantized continuous data.

- **Gaussians**: For $\Normal(\BM x\\y\EM|0,\BMσ^2&ρσ^2\\ρσ^2&σ^2\EM)$, $\H(X,Y)=\/{1}{2}\ln((2πe)^2\det(Σ))$ and $\H(X)=\H(Y)=\/{1}{2}\ln(2πeσ^2)$.
Then $\I(X,Y)=\H(X)+\H(Y)-\H(X,Y)$
$=\ln(2πeσ^2)-\/{1}{2}\ln((2πe)^2σ^4(1-ρ^2))$
$=\/{1}{2}\ln\/{(2πe)^2σ^4}{(2πe)^2σ^4(1-ρ^2)}$
$\dg{⇒}\I(X,Y)=-\/{1}{2}\ln(1-ρ^2)$
$=\BC∞&|ρ|→1\\
0&ρ=0
\EC$

# PRML: 2 Probability Distributions

**Softmax** $\red{\/{\e{a_k}}{\sum_j\e{a_j}}}$: in classification we use $\arg\max[a_1=1, a_2=5, a_3=2]=2$ to find the winning index, and use that to encode a one-hot vector $[0, 1, 0]$.
Each element in $\{0,1\}$ discrete range cannot be used for gradient descent.
The softmax output $[0.017,0.936,0.047]$ is a probability distribution that mimics $[0,1,0]$ if one element is much larger than the others.
Each element in $[0,1]$ continuous range, which can be used in gradient descent.

- **LogSumExp** $\red{\ln(\sum_j\e{a_j})}[1,5,2]=5.05$ is the actual smooth maximum function.
It is the log of the denominator of softmax.
$\/{∂}{∂a_k}\ln(\sum_j\e{a_j})=\/{\e{a_k}}{\sum_j\e{a_j}}$ derivative of log-sum-exp is the softmax function.

- **Log-partition function**: $\BA
p(x_i|η)=\/{h(x_i)\e{η^⊤ϕ(x_i)}}{∫h(u)\e{η^⊤ϕ(u)}\ du}=h(x_i)\e{-A(η)}\e{η^⊤u(x_i)}
\EA$ is exponential family with partition function $A(η)=\ln(∫h(u)\e{η^⊤ϕ(u)}\ du)$ and $∇_ηA(η)=\E[ϕ(x_i)|η]$.

  - The discrete form $\/{\e{a_k}}{\sum_j\e{a_j}}$ is the softmax function, where $A(η)=\ln(\sum_j\e{a_j})$ is the log-partition function.
  In categorical distribution, the one-hot vector $u(x_i)=[0,1,0]$ is the sufficient statistic computed from data $x_i$, while $η$ are the comparative logits.
  Then $∇_ηA(η)=\E[u(x)|η]=[0.017,0.936,0.047]$ is the softmax output.

**Binomial**: Bernoulli $x_i∈\{0,1\}$ then
$p(x_i=1|μ)=μ$,
$\E[x_i]=μ$,
$\V[x_i]=μ(1-μ)$.

- MLE $p(x|μ)=\prod_{i=1}μ^{x_i}(1-μ)^{1-x_i}$
$\dg{⇒}\ln p(x|μ)=\sum_ix_i\lnμ+(1-x_i)\ln(1-μ)$
$\dg{⇒}\blue{μ_\t{ML}=\/{1}{N}\sum_ix_i}$

- **Binomial**: $p(m|N,μ)=\binom{N}{m}μ^m(1-μ)^{N-m}$,
$\E[m]=Nμ$,
$\V[m]=Nμ(1-μ)$.

- **Beta**: prior $\red{p(μ|a,b)=\/{Γ(a+b)}{Γ(a)Γ(b)}μ^{a-1}(1-μ)^{b-1}}$,
$\E[μ]=\/{a}{a+b}$,
$\V[μ]=\/{ab}{(a+b)^2(a+b+1)}$,
where $\BC
a=a_0+m\\
b=b_0+(N-m)
\EC$ are hyperparameters representing prior observations of 1s and 0s.
**Sequential learning**: Start with $p(μ|a,b)∼\t{beta}(a,b)$.
At every turn old posterior becomes new prior $p(μ|a,b)=p(μ|a+1,b)$.

  - Conjugacy: beta prior × binomial likelihood = beta
  $\blue{p(μ|N,m,a,b)=\/{Γ(a+b+N)}{Γ(a+m)Γ(b+N-m)}μ^{m+a-1}(1-μ)^{N-m+b-1}}$

  - Predicting next trial $p(x_n|x)=∫_0^1p(x_n=1|μ)p(μ|x)\ dμ$
  $=∫_0^1μp(μ|x)\ dμ$
  $=\E[μ|x]$
  $=\/{a+m}{a+b+N}$.

- **Variance reduction**: by Adam and Eve $\E[θ]=\E_x[\E[θ|x]]=∫p(x)[∫θp(θ|x)\ dθ]\ dx$ prior expectation is average posterior expectation over all data.
$\E_x[\V[θ|x]]=\V[θ]-\V_x[\E[θ|x]]$ the average posterior variance is lower than the prior variance.


**Multinomial**: $K$ mutually exclusive states, represented by one-hot $x_i∈\{0,1\}^K$ with distribution $p(x_i|μ)=\prod_kμ_k^{x_{ik}}$ with mean simplex
$μ∈ℝ^K,μ≽0,\v{1}^⊤μ=1$.
Then
$\E[x_i]=\sum_{x_i}x_ip(x_i|μ)=I_Kμ=μ$.

- MLE $p(x|μ)=\prod_ip(x_i|μ)$
$=\prod_kμ_k^{\sum_ix_{ik}}$
where $\sum_ix_{ik}$ is sufficient statistic for $μ_k$ by factorization theorem.
Convex problem $\BC\t{maximize}_μ&\ln p(x|μ)=\sum_k(\sum_ix_{ik})\ln(μ_k)\\
\t{subject to}&μ≽0, \v{1}^⊤μ=1\EC$
with simplified Lagrangian $\sum_k(\sum_ix_{ik})\ln(μ_k)+ν(\v{1}^⊤μ-1)$
$\dg{⇒}∂_{μ_k}=\/{\sum_ix_{ik}}{μ_k}+ν=0$
$\dg{⇒}μ_k^*=-\/{\sum_ix_{ik}}{ν}$
$\dg{⇒}\sum_k-\/{\sum_ix_{ik}}{ν}=1$
$\dg{⇒}ν^*=-N$
$\dg{⇒}\blue{μ_{k,\t{ML}}=\/{1}{N}\sum_ix_{ik}}$.

- **Multinomial**: $p(m_1,...,m_K|μ,N)=\binom{N}{m_1...m_K}\prod_kμ_k^{m_k},\ \sum_km_k=N$

- **Dirichlet**: prior $\red{p(μ|α)=\/{Γ(\v{1}^⊤α)}{Γ(α_1)...Γ(α_K)}\prod_kμ_k^{α_k-1}}$
posterior $\blue{p(μ|x,α)=\/{Γ(\v{1}^⊤α+N)}{Γ(α_1+m_1)...Γ(α_K+m_K)}\prod_kμ_k^{α_k+m_k-1}}$

- **Overparameterization (Identifiability)**: Let $x_i∈\{0,1\}^K$ be one-hot, and $μ∈ℝ^K$ satisfies $\v{1}^⊤μ=1,μ≽0$.
$\BA
\dg{⇒}&p(x_i|μ)=\prod_{k=1}^Kμ_k^{x_{ik}}=\e{\sum_{k=1}^Kx_{ik}\ln μ_k}\\
\dg{⇒}&\dg{\t{PROBLEM: if we let }}η_k=\ln μ_k\dg{⇒}p(x_i|η)=\e{\sum_{k=1}^Kη_kx_{ik}}\dg{⇒}p(x_i|η+c\v{1})=\e{\sum_{k=1}^K(η_k+c)x_{ik}}\\
&=\e{\sum_{k=1}^Kη_kx_{ik}+c\v{1}^⊤x_i}=\e{c}p(x_i|η)\dg{ - \t{$\e{c}$ gets normalized}}\dg{⇒}p(x_i|η)=\e{\sum_{k=1}^Kη_kx_{ik}-A(η)}\\
&\dg{⇒}μ_k=p(x_{ik}=1,x_{ij}=0|η)=\e{η_k-A(η)}\dg{⇒}\sum_{k=1}^K\e{η_k-A(η)}=1\dg{⇒}\e{A(η)}=\sum_{j=1}^K\e{η_j}\\
&\dg{⇒}\red{μ_k=\/{\e{η_k}}{\sum_{j=1}^K\e{η_j}}}\dg{ - \t{$η$ not identifiable: $μ_k(η)=μ_k(η+c\v{1})$ - one $μ$ value corresponds to infinite number of $η$ values}}\\
\dg{⇒}&\dg{\t{SOLUTION: anchor }}\red{η_K=0}\dg{⇒}\sum_{j=1}^K\e{η_j}=1+\sum_{j=1}^{K-1}\e{η_j}\dg{⇒}\red{μ_k=\/{\e{η_k}}{1+\sum_{j=1}^{K-1}\e{η_j}}}\dg{ - \t{anchored softmax}}\\
&\dg{⇒}μ_K=\/{1}{1+\sum_{j=1}^{K-1}\e{η_j}}\dg{⇒}μ_k=μ_K\e{η_k}\dg{⇒}\red{η_k=\ln\/{μ_k}{μ_K}}\dg{ - \t{natural parameter $η_k$ is log-odds wrt $μ_K$}}\\
&\dg{⇒\t{identifiability problem $μ_k(η)=μ_k(η+c\v{1})$ is anchored by $η_K=0$}}
\EA$

  - **Anchored softmax** $\blue{η_k=\ln\/{μ_k}{μ_K}\dg{⇒}μ_k=\/{\e{η_k}}{1+\sum_{j=1}^{K-1}\e{η_j}}}$ is rooted in statistics.
  $η$ is identifiable.
  Hessian is positive definite and invertible.
  Newton's method can run without crashing.
  However it breaks symmetry. Parameters depend on which class is arbitrarily chosen as anchor.

  - **Symmetrical softmax** $\blue{η_k=\ln μ_k\dg{⇒}μ_k=\/{\e{η_k}}{\sum_{j=1}^K\e{η_j}}}$ is rooted in modern machine learning.
  $η$ is symmetrical but not identifiable.
  Hessian is singular and non-invertible.
  Regularization on weights with Gaussian prior makes the Hessian invertible again.
  Used in practice today.

**Gaussian**: $\red{\Normal(x_i|μ,Σ)=\/{1}{(2π)^{D/2}}\/{1}{\det(Σ)^{1/2}}\e{-\/{1}{2}(x_i-μ)^⊤Σ^{-1}(x_i-μ)}},\ x_i∈ℝ^D$

- **Covariance**: $x∈ℝ^D\dg{⇒}Σ=\Cov[x]=\E[(x-\E[x])(x-\E[x])^⊤]$.

  - Let $X∈ℝ^{N×D}$ and $x_{:i}$ be $i$-th column of $X$, then $\Cov[x_{:i},x_{:j}]=\E[(x_{:i}-\E[x_{:i}])(x_{:j}-\E[x_{:j}])]=\E[x_{:i}x_{:j}]-\E[x_{:i}]\E[x_{:j}]$.

  - $\E[xx^⊤]=Σ+\E[x]\E[x]^⊤$.
  Affine precomposition $\Cov[Ax+b]=A\Cov[x]A^⊤$.

  - **Bivariate Gaussian**: $D=2$. Then $Σ=\BMσ_1^2&ρσ_1σ_2\\ρσ_1σ_2&σ_2^2\EM$ where $ρ=\/{\Cov[X_1,X_2]}{\sqrt{\V[X_1]\V[X_2]}}$

- **Whitening/Sphering**: $p(x_i)=\Normal(x_i|μ,Σ)$ has **anisotropic** variance (shaped like an ellipsoidal cloud). $x_i$ can be rebased along eigenvectors of $Σ^{-1}$ so that its variance becomes **isotropic** (shaped like a unit-norm sphere) and Euclidean distance can be used instead of Mahalanobis distance.
$\BA
\dg{⇒}&Σ=UDU^⊤\dg{⇒}Σ^{-1}=UD^{-1}U^⊤\dg{⇒}\red{Σ^{-1/2}=UD^{-1/2}U^⊤}\dg{⇒}\BC
Σ^{-1/2}ΣΣ^{-1/2}=[UD^{-1/2}U^⊤][UDU^⊤][UD^{-1/2}U^⊤]=I\\
Σ^{-1/2}Σ^{-1/2}=[UD^{-1/2}U^⊤][UD^{-1/2}U^⊤]=Σ^{-1}
\EC\\
&\dg{⇒}(x_i-μ)^⊤Σ^{-1}(x_i-μ)=[Σ^{-1/2}(x_i-μ)]^⊤[Σ^{-1/2}(x_i-μ)]=\n{Σ^{-1/2}(x_i-μ)}_2^2=\n{UD^{-1/2}U^⊤(x_i-μ)}_2^2\\
&=\n{\red{D^{-1/2}U^⊤}(x_i-μ)}_2^2\dg{ - \t{$U$ is orthogonal matrix therefore can be removed from norm}}\\
\dg{⇒}&(D^{-1/2}U^⊤)Σ(D^{-1/2}U^⊤)=(D^{-1/2}U^⊤)(UDU^⊤)(D^{-1/2}U^⊤)^⊤=I\dg{ - \t{however $(D^{-1/2}U^⊤)(D^{-1/2}U^⊤)\neqΣ^{-1}$}}\\
&\dg{⇒}\tilde{x}_i:=D^{-1/2}U^⊤x_i\dg{⇒}\V[x_i]=Σ\dg{⇒}\V[\tilde{x}_i]=\V[D^{-1/2}U^⊤x_i]=(D^{-1/2}U^⊤)Σ(D^{-1/2}U^⊤)^⊤=I\dg{ - \t{norm-sphere variance}}\\
\dg{⇒}&(x_i-μ)^⊤Σ^{-1}(x_i-μ)=\n{\tilde{x}_i-\tilde{μ}}^2=\n{D^{-1/2}U^⊤(x_i-μ)}_2^2\dg{ - \t{Euclidean distance in the whitened space}}
\EA$

- **Moments**: Let $z_i=x_i-μ\dg{⇒}z_i=Uy_i$.
  
  - $\E[x_i]=∫(z_i+μ)\/{1}{(2π)^{D/2}}\/{1}{\det(Σ)^{1/2}}\e{-\/{1}{2}z_i^⊤Σ^{-1}z_i}\ dz_i$
  $=\E[z_i]+μ$
  $\dg{⇒}\blue{\E[x_i]=μ}$.

  - $\E[x_ix_i^⊤]=\E[(z_i+μ)(z_i+μ)^⊤]$
  $=\E[z_iz_i^⊤]+\E[z_iμ^⊤]+\E[μz_i^⊤]+\E[μμ^⊤]$
  $=\E[z_iz_i^⊤]+μμ^⊤$
  $=U\E[y_iy_i^⊤]U^⊤+μμ^⊤$
  $=UDU^⊤+μμ^⊤$
  $\dg{⇒}\blue{\E[x_ix_i^⊤]=Σ+μμ^⊤}$.

- **Maximum likelihood**: $\red{\ln p(x|μ,Σ)=-\/{ND}{2}\ln(2π)-\/{N}{2}\ln\det(Σ)-\/{1}{2}\sum_i(x_i-μ)^⊤Σ^{-1}(x_i-μ)},\ x∈ℝ^{N×D}$

  - Exponential family $(x_i-μ)^⊤Σ^{-1}(x_i-μ)=x_i^⊤Σ^{-1}x_i-2x_i^⊤Σ^{-1}μ+c$
  $=\tr(Σ^{-1}x_ix_i^⊤)-2x_i^⊤Σ^{-1}μ+c$.
  Sufficient statistics $\blue{(\sum_ix_i,\sum_ix_ix_i^⊤)}$.

  - $\BC
  ∂_μ\ln p(x|μ,Σ)=\sum_i-2(x_i-μ)^⊤Σ^{-1}=0&\dg{⇒}\red{μ_\t{ML}=\/{1}{N}\sum_ix_i}\\
  ∂_Σ\ln p(x|μ,Σ)=-\/{N}{2}Σ^{-1}+\/{1}{2}Σ^{-1}[\sum_i(x_i-μ)(x_i-μ)^⊤]Σ^{-1}=0&\dg{⇒}\red{Σ_\t{ML}=\/{1}{N}\sum_i(x_i-μ_\t{ML})(x_i-μ_\t{ML})^⊤}
  \EC$

- **Sequential update**: Given $x_N$ and $μ_\t{ML}^{(N-1)}$ we have $μ_\t{ML}^{(N)}=\/{1}{N}\sum_ix_i$
$=\/{1}{N}x_N+\/{1}{N}\sum_{i=1}^{N-1}x_i$
$=\/{1}{N}x_N+\/{N-1}{N}μ_\t{ML}^{(N-1)}$
$\dg{⇒}\blue{μ_\t{ML}^{(N)}=μ_\t{ML}^{(N-1)}+\/{1}{N}(x_N-μ_\t{ML}^{(N-1)})}$

  - **Robins-Monro**: ML estimate is root of score: $l'(θ|x_i)=\/{∂}{∂θ}\ln p(x_i|θ)\dg{⇒}\lim\limits_{N→∞}\/{1}{N}\sum_il'(θ|x_i)=\E_{x_i}[l'(θ|x_i)]$ satisfies $\E_{x_i}[l'(θ_\t{ML}|x_i)]=0$ and $\E_{x_i}[l'(θ_\t{ML}|x_i)^2]< ∞$, then
  $\blue{\BC
  θ^{(N)}=θ^{(N-1)}+a_{N-1}l'(θ^{(N-1)}|x_N)\\
  a_{N-1}=\/{1}{NI_1(θ^{(N-1)})}=\/{1}{N\E_{x_i}[l'(θ^{(N-1)}|x_N)^2]}

\dg{⇒}&t_\m{D}∈ℝ^N\dg{ - \t{training data}}\dg{⇒}\red{E(w):=\/{1}{2}\sum_i(y(x_i,w)-t_i)^2=\/{1}{2}\n{t_\m{D}-Φw}^2}\dg{ - \t{OLS error}}\\
\dg{⇒}&p(t_\m{D}|w)=\prod_i\Normal(t_i|y(x_i,w),β^{-1})\dg{⇒}\red{-\ln p(t_\m{D}|w)=\/{β}{2}\sum_i(t_i-y(x_i,w))^2=\/{β}{2}\n{t_\m{D}-Φw}^2=βE(w)}\dg{ - \t{MLE error}}\\
&\dg{⇒}p(t_\m{D}|w,β)=\prod_i\Normal(t_i|w^⊤ϕ_i,β^{-1})
\dg{⇒}\ln p(t_\m{D}|w,β)=\/{N}{2}\ln β+\sum_i\/{-β}{2}(t_i-w^⊤ϕ_i)^2+c
=\/{N}{2}\ln β-\/{β}{2}\n{t_\m{D}-Φw}^2+c\\
&\dg{⇒}∇_w=β(t_\m{D}-Φw)^⊤Φ=0
\dg{⇒}Φ^⊤t_\m{D}=Φ^⊤Φw
\dg{⇒}\red{w_\t{ML}=(Φ^⊤Φ)^{-1}Φ^⊤t_\m{D}}\dg{ - \t{MLE for weights}}\\
&\dg{⇒}∇_β=\/{N}{β}-\n{t_\m{D}-Φw}^2=0
\dg{⇒}\red{\/{1}{β_\t{ML}}=\/{1}{N}\n{t_\m{D}-Φw_\t{ML}}^2}\dg{ - \t{MLE for precision is biased $\/{1}{N}$}}\\
\dg{⇒}&E(w)=\/{β}{2}\sum_i(t_i-w^⊤ϕ_i)^2=\/{β}{2}\sum_i(t_i-w_0-w_{1..M}^⊤ϕ_{i,1..M})^2\dg{⇒}\/{∂E}{∂w_0}=β\sum_i(t_i-w_0-w_{1..M}^⊤ϕ_{i,1..M})=0\\
&\dg{⇒}N\bar{t}-Nw_0-w_{1..M}^⊤(N\bar{ϕ}_{1..M})=0\dg{⇒}\red{w_0=\bar{t}-w_{1..M}^⊤\bar{ϕ}_{1..M}}\dg{ - \t{MLE for bias term $w_0$}}\\
\dg{⇒}&\E[t|x,\m{D}]=y_\m{D}(x)=w_\t{MLE}^⊤ϕ(x)\dg{-\t{prediction}}
\EA$

- **Sequential learning**: Gradient descent with step size $η$ we have $w^{(k+1)}=w^{(k)}-η∇_wE_D$.
SGD uses random $∇_wE_i$ we have $=w^{(k)}+η(t_i-ϕ(x_i)^⊤w^{(k)})ϕ(x_i)$.

- **Regularization**: $E_R(w)=\/{1}{2}\n{t_\m{D}-Φw}^2+\/{λ}{2}\n{w}^2=E(w)+\/{λ}{2}\n{w}^2$
$\dg{⇒}\red{w^*=(λI+Φ^⊤Φ)^{-1}Φ^⊤t_\m{D}}$.

  - **General regularizer**: $E_R(w)=\/{1}{2}\n{t_\m{D}-Φw}_2^2+\/{λ}{2}\n{w}_q^q$
  $=\/{1}{2}\sum_{i=1}^N(t_i-w^⊤ϕ(x_i))^2+\/{λ}{2}\sum_{j=1}^M|w_j|^q$.
  Unregularized RSS error $E(w)=-\/{1}{2}\n{t_\m{D}-Φw}^2=-\/{1}{2}w^⊤(Φ^⊤Φ)w-w^⊤Φ^⊤t_\m{D}+\/{1}{2}t_\m{D}^⊤t_\m{D}$ is a bowl with ellipsoid level sets.
  In regularization we search for the bottom of the bowl while constrained to the origin in a contour defined by the regularizer.

  - **Contours**: $|w_1|^q+|w_2|^q$ is plotted on a $w_1$-$w_2$ graph.
  $\blue{q=0.5}$ creates a non-convex star shape with sharp narrow spikes pinched toward axes.
  It penalizes having multiple non-zero weights.
  $\blue{q=1}$ **lasso** regularizer forms a (convex) diamond shape centered at origin. Moving toward the bottom of the bowl likely ends up on axis at one of the corners resulting in a sparse solution.
  $\blue{q=2}$ **ridge** regularizer forms a circle centered at origin. Moving toward the bottom of the bowl is almost never exactly on axis.
  $\blue{q=4→∞}$ regularizer looks more and more like a square. Moving toward the bottom of the bowl likely ends up in one of the corners resulting in equal weights in magnitude.
  
- **Multiple outputs**: $y(x,W)=W^⊤ϕ(x)∈ℝ^K$
$\dg{⇒}W_\t{ML}=(Φ^⊤Φ)^{-1}Φ^⊤T,\ w_k=(Φ^⊤Φ)^{-1}Φ^⊤t_k$

**Bias-Variance**: Suppose optimal $y^*(x)=\E[t|x]$, model $y_\m{D}(x)$ trained on dataset $\m{D}$.
Then expected loss is
$\BA
\dg{⇒}&\E[L]=\iint[t-y(x)]^2p(x)\ dx\ dt
=∫(y(x)-y^*(x))^2p(x)\ dx+\iint(t-y^*(x))^2p(t|x)\ dt\ p(x)\ dx\\
&=\E[(y(x)-y^*(x))^2]+\purple{\E[\V_{t}[t|x]]}\\
\dg{⇒}&(y_\m{D}(x)-y^*(x))^2=(y_\m{D}(x)-\E[y_\m{D}(x)])^2+(\E[y_\m{D}(x)]-y^*(x))^2+2(y_\m{D}(x)-\E[y_\m{D}(x)])(\E[y_\m{D}(x)]-y^*(x))\\
\dg{⇒}&\E[(y_\m{D}(x)-y^*(x))^2]=\E[(\E[y_\m{D}(x)]-y^*(x))^2]+\E[(y_\m{D}(x)-\E[y_\m{D}(x)])^2]
=\green{(\E[y_\m{D}(x)]-y^*(x))^2}+\yello{\V_\m{D}[y_\m{D}(x)]}\\
\dg{⇒}&\E[L]=\green{(\E[y_\m{D}(x)]-y^*(x))^2}+\yello{\V_\m{D}[y_\m{D}(x)]}+\purple{\E[\V_{t}[t|x]]}\\
\EA$

- $\green{(\E[y_\m{D}(x)]-y^*(x))^2}$ is bias-squared.
$\yello{\V_\m{D}[y_\m{D}(x)]}$ is variance of estimate.
$\purple{\E[\V_{t}[t|x]]}$ is irreducible expected measurement noise.
Fundamental trade-off between variance and bias.

**Generative linear regression**:
$\BA
\dg{⇒}&Φ∈ℝ^{N×(M+1)},t_\m{D}∈ℝ^N\dg{⇒}\BC
\t{prior}&p(w)=\Normal(w|m_0,S_0)\\
\t{likelihood}&p(t_\m{D}|w)=\prod_i\Normal(t_i|w^⊤ϕ_i,β^{-1})
\EC\\
\dg{⇒}&\dg{\t{unlike joint distribution, posterior distribution only needs to complete the square in $w$}}\\
\dg{⇒}&\ln p(w|t_\m{D})=\ln p(w)+\ln p(t_\m{D}|w)+c_1=-\/{1}{2}(w-m_0)^⊤S_0^{-1}(w-m_0)-\/{β}{2}\n{t_\m{D}-Φw}^2+c_2\\
&=-\/{1}{2}[w^⊤S_0^{-1}w-2w^⊤S_0^{-1}m_0+βw^⊤Φ^⊤Φw-2βw^⊤Φ^⊤t_\m{D}]+c_3=-\/{1}{2}[w^⊤(\green{S_0^{-1}+βΦ^⊤Φ})w-2w^⊤(\green{S_0^{-1}m_0+βΦ^⊤t_\m{D}})]+c_3\\
\dg{⇒}&\red{p(w|t_\m{D})=\Normal(w|m_N,S_N)\dg{⇒}\BC
S_N=(S_0^{-1}+βΦ^⊤Φ)^{-1}\dg{ - \t{posterior covariance}}\\
m_N=S_N(S_0^{-1}m_0+βΦ^⊤t_\m{D})\dg{ - \t{$w_\t{MAP}$ - the peak of the posterior distribution}}\EC}\\
\dg{⇒}&\dg{\t{isotropic prior: }}p(w|α)=\Normal(w|0,α^{-1}I)\dg{⇒}\BC
p(w|t_\m{D})=\Normal(w|m_N,S_N),\ S_N=(αI+βΦ^⊤Φ)^{-1},\ m_N=βS_NΦ^⊤t_\m{D}\\
\green{\ln p(w|t_\m{D})}=\ln p(w)+\ln p(t_\m{D}|w)+c_1=\green{-\/{α}{2}\n{w}^2-\/{β}{2}\n{t_\m{D}-Φw}^2}+c_2
\EC\\
&\dg{⇒}\blue{-\ln p(w|t_\m{D})=βE(w)+α\n{w}^2}\dg{-\t{MAP error is MLE error + isotropic Gaussian prior error each weighed by accuracy}}\\
\EA$

- **Generalized Gaussian prior**: regularized least squares matching prior $p(w_j|α)∝\e{-\/{α}{2}|w_j|^q}$
$\BC
\dg{⇒}&∫_{-∞}^∞p(w_j|α)\ dw_j=∫_{-∞}^∞\e{-\/{α}{2}|w_j|^q}\ dw_j=2∫_0^∞\e{-\/{α}{2}w_j^q}\ dw_j\\
\dg{⇒}&u=\/{α}{2}w_j^q,\ dw_j=\/{1}{q}(\/{α}{2})^{-1/q}u^{1/q-1}\ du
\dg{⇒}∫_{-∞}^∞p(w_j|α)\ dw_j=\/{2}{q}(\/{α}{2})^{-1/q}∫_0^∞u^{1/q-1}e^{-u}\ du=\/{2}{q}(\/{α}{2})^{-1/q}Γ(\/{1}{q})\\
\dg{⇒}&p(w_j|α)=\/{q}{2}(\/{α}{2})^{1/q}\/{1}{Γ(1/q)}\e{-\/{α}{2}|w_j|^q}
\dg{⇒}\red{p(w|α)=\left[\/{q}{2}(\/{α}{2})^{1/q}\/{1}{Γ(1/q)}\right]^M\e{-\/{α}{2}\n{w}_q^q}}
\EC$

  - Finding $w_\t{MAP}$ using generalized Gaussian prior with $\sum_j|w_j|^q$ in its exponent is regularized least-squares with penalty $\sum_j|w_j|^q$.

  - Setting $q=2$ gives standard Gaussian.

- **Generative prediction**: To predict $ϕ=ϕ(x)∈ℝ^{M+1},t∈ℝ$ pair we have by Adam and Eve
$\BA
\dg{⇒}&p(t|t_\m{D})=∫p(t|w)p(w|t_\m{D})\ dw=∫\Normal(t|w^⊤ϕ,β^{-1})\Normal(w|m_N,S_N)\ dw\\
&\dg{⇒}\E[t|t_\m{D}]=\E_w[\E[t|w]|t_\m{D}]=\E_w[w^⊤ϕ|t_\m{D}]=\E[w|t_\m{D}]^⊤ϕ=\yello{m_N^⊤ϕ}\\
&\dg{⇒}\V[t|t_\m{D}]=\E_w[\V[t|w]|t_\m{D}]+\V_w[\E[t|w]|t_\m{D}]=\E_w[β^{-1}|t_\m{D}]+\V_w[w^⊤ϕ|t_\m{D}]=\green{β^{-1}+ϕS_Nϕ^⊤}\\
\dg{⇒}&p(t|t_\m{D})=\Normal(t|\yello{m_N^⊤ϕ},\green{β^{-1}+ϕS_Nϕ^⊤})\dg{⇒}\E[t|x,\m{D}]=w_\t{MAP}^⊤ϕ(x)\dg{-\t{generative because it also has a whole distribution}}\\
\EA$

- **3.8**: Having observed $N$ data points, for $(x_{N+1},t_{N+1})$ we have $\BC
p(w|t_\m{D})=\Normal(w|m_N,S_N)\\
p(t_{N+1}|w)=\Normal(t_{N+1}|w^⊤ϕ(x_{N+1}),β^{-1})
\EC$
  $\BA
  \dg{⇒}&p(w|t_\m{D},t_{N+1})∝p(t_{N+1}|w)p(w|t_\m{D})\\
  \dg{⇒}&E=-\/{1}{2}(w-m_N)^⊤S_N^{-1}(w-m_N)-\/{β}{2}(t_{N+1}-w^⊤ϕ(x_{N+1}))^⊤(t_{N+1}-w^⊤ϕ(x_{N+1}))\\
  &=-\/{1}{2}[w^⊤S_N^{-1}w-2w^⊤S_N^{-1}m_N+w^⊤βϕ(x_{N+1})^⊤ϕ(x_{N+1})w-2βw^⊤ϕ(x_{N+1})t_{N+1}]+c\dg{ - \t{complete the squares in $w$}}\\
  &=-\/{1}{2}[w^⊤(S_N^{-1}+βϕ(x_{N+1})^⊤ϕ(x_{N+1}))w-2w^⊤(S_N^{-1}m_N+βϕ(x_{N+1})t_{N+1})]+c\\
  &=-\/{1}{2}(w-m_{N+1})^⊤S_{N+1}^{-1}(w-m_{N+1})\\
  \dg{⇒}&S_{N+1}^{-1}=S_N^{-1}+βϕ(x_{N+1})^⊤ϕ(x_{N+1}),\ m_{N+1}=S_{N+1}(s_N^{-1}m_N+βϕ(x_{N+1})t_{N+1})
  \EA$

  - **3.9**: Use the conditional formulas. Complete the squares in two parameters
  $\BA
  \dg{⇒}&p(t_{N+1},w|t_\m{D})=p(w|t_\m{D})p(t_{N+1}|w)\\
  \dg{⇒}&E=-\/{1}{2}(w-m_N)^⊤S_N^{-1}(w-m_N)-\/{β}{2}(t_{N+1}-w^⊤ϕ(x_{N+1}))^⊤(t_{N+1}-w^⊤ϕ(x_{N+1}))\\
  &=-\/{1}{2}[w^⊤S_N^{-1}w-2w^⊤S_N^{-1}m_N+w^⊤βϕ(x_{N+1})^⊤ϕ(x_{N+1})w-2βw^⊤ϕ(x_{N+1})t_{N+1}+βt_{N+1}^2]+c\\
  &=-\/{1}{2}[\BM w\\t_{N+1}\EM^⊤\BM S_N^{-1}+βϕ(x_{N+1})^⊤ϕ(x_{N+1})&-βϕ(x_{N+1})\\-βϕ(x_{N+1})^⊤&β\EM\BM w\\t_{N+1}\EM-2\BM w\\t_{N+1}\EM^⊤\BM S_N^{-1}m_N\\0\EM]+c\\
  \dg{⇒}&Σ_{w|t_{N+1}}^{-1}=Λ_{11}=S_N^{-1}+βϕ(x_{N+1})^⊤ϕ(x_{N+1}),\ μ_{w|t_{N+1}}=Λ_{11}^{-1}(η_1-Λ_{12}t_{N+1})=Σ_{w|t_{N+1}}(S_N^{-1}m_N+βϕ(x_{N+1})t_{N+1})
  \EA$

**Equivalent kernel**: Suppose isotropic prior $p(w|α)=\Normal(w|0,α^{-1}I)$ then $\E[t|x,t_\m{D}]$ is a linear combination of training $t_i$ weighted by dot product similarity between $x$ and $x_i$ (**linear smoother**).
$\BA
\dg{⇒}&\E[t|t_\m{D}]=y(x,m_N)=m_N^⊤ϕ=βϕ^⊤S_NΦ^⊤t_\m{D}=βϕ^⊤S_N\sum_{i=1}^Nϕ_it_i=\sum_{i=1}^Nβϕ^⊤S_Nϕ_it_i=\sum_{i=1}^Nk(x,x_i)t_i\dg{ - \t{weighted $t_\m{D}$}}\\
&\dg{⇒}S_N=\V[w|t_\m{D}]=(αI+βΦ^⊤Φ)^{-1}=UDU^⊤\dg{⇒}\red{k(x,x_i):=βϕ^⊤S_Nϕ_i}=β(U^⊤ϕ)^⊤D(U^⊤ϕ_i)=\sum_{j=1}^M\/{β}{α+\yello{λ_j}}(u_j^⊤ϕ)^⊤(u_j^⊤ϕ_i)\\
&\dg{⇒\t{eigenvectors $U$ of $\V[w|t_\m{D}]=S_N$ and $\V[Φ]=Φ^ΤΦ$ are identical, but have nearly inverse eigenvalues}}\\
&\dg{⇒\t{$ϕ^⊤S_Nϕ_i$ de-emphasizes dot product $(u_j^⊤ϕ)^⊤(u_j^⊤ϕ_i)$ along principal axes where feature covariance $\yello{λ_i}$ are high}}\\
&\dg{⇒}\tilde{ϕ}=β^{1/2}S_N^{1/2}ϕ\dg{⇒}\V[\tilde{Φ}]≈I\dg{\t{ (isotropic)}}\dg{⇒}k(x,x_i)=\tilde{ϕ}^⊤\tilde{ϕ}_i\dg{ - \t{dot product in whitened space}}\\
\dg{⇒}&\V[y(x,w),y(x_i,w)|t_\m{D}]=\V[ϕ^⊤w,w^⊤ϕ_i|t_\m{D}]=ϕ^⊤\V[w|t_\m{D}]ϕ_i=ϕ^⊤S_Nϕ_i=β^{-1}k(x,x_i)\\
&\dg{⇒\t{covariance in model predictions is determined by $k(x,x_i)$}}\\
\dg{⇒}&t_\m{D}=\v{1}\dg{⇒}\E[t|t_\m{D}]=\sum_ik(x,x_i)t_i=1\dg{⇒}\red{\sum_{i=1}^Nk(x,x_i)=1}\dg{ - \t{sum of the weights must equal to 1}}
\EA$

- **Non-parametric kernel method**: Going down the path of using parametric (weights) models with basis functions $y(x)=w_\t{ML}^⊤ϕ$ or $y(x)=w_\t{MAP}^⊤ϕ$ or $y(x)=w_\t{ridge}^⊤ϕ$ to its logical conclusion always leads to $y(x)=\sum_ik(x,x_i)t_i$.

**Bayesian evidence framework**: full Bayesian treatment with integration over all hyperparameters.
$\BA
\dg{⇒}&\BC
p(t_\m{D}|w,β)=\prod_i\Normal(t_i|ϕ_i^⊤w,β^{-1})\\
p(w|α)=\Normal(0,α^{-1}I)\\
p(w|t_\m{D})=\Normal(w|m_N,S_N)
\EC\dg{⇒}\BC
S_N=(αI+βΦ^⊤Φ)^{-1}\\
m_N=βS_NΦ^⊤t=β(αI+βΦ^⊤Φ)^{-1}Φ^⊤t
\EC\\
\dg{⇒}&p(t_\m{D}|x,t_\m{D})=\iiint p(t_\m{D}|x,w,β)p(w|t_\m{D},α,β)\purple{p(α,β|t)}\ dw\ dα\ dβ≈p(t_\m{D}|x,t_\m{D},\hat{α},\hat{β})=∫p(t_\m{D}|w,\hat{β})p(w|t_\m{D},\hat{α},\hat{β})\ dw\\
&\dg{⇒}\hat{α},\hat{β}=\arg\max_{α,β}\purple{p(α,β|t_\m{D})}\dg{ - \t{MAP estimators}}\\
&\dg{⇒}\purple{p(α,β|t_\m{D})}∝\brown{p(t_\m{D}|α,β)}p(α,β)\dg{⇒}\hat{α},\hat{β}=\arg\max_{α,β}\brown{p(t_\m{D}|α,β)}\dg{ - \t{assuming prior $p(α,β)$ is relatively flat}}\\
\dg{⇒}&\dg{\t{Completing squares: let $E(w)$ be the MAP error of posterior (instead of MLE error of likelihood)}}\\
\dg{⇒}&\brown{p(t_\m{D}|α,β)}=∫p(t_\m{D},w|α,β)\ dw=∫p(t_\m{D}|w,β)p(w|α)\ dw=∫(\/{β}{2π})^{N/2}(\/{α}{2π})^{M/2}\e{-\green{E(w)}}\ dw\\
&\dg{⇒}\blue{2E(w)=β\n{t_\m{D}-Φw}^2+α\n{w}^2}=β(w^⊤Φ^⊤Φw-2w^⊤Φ^⊤t_\m{D}+t_\m{D}^⊤t_\m{D})+αw^⊤w=w^⊤S_N^{-1}w-2w^⊤S_N^{-1}m_N+βt_\m{D}^⊤t_\m{D}\\
&\quad=w^⊤S_N^{-1}w-2w^⊤S_N^{-1}m_N+βt_\m{D}^⊤t_\m{D}+m_N^⊤S_N^{-1}m_N-m_N^⊤S_N^{-1}m_N=\yello{(w-m_N)^⊤S_N^{-1}(w-m_N)}+\green{βt_\m{D}^⊤t_\m{D}-m_N^⊤S_N^{-1}m_N}\\
&\dg{⇒}\green{βt_\m{D}^⊤t_\m{D}-m_N^⊤S_N^{-1}m_N}=βt_\m{D}^⊤t_\m{D}-2m_N^⊤(S_N^{-1}m_N)+m_N^⊤S_N^{-1}m_N=βt_\m{D}^⊤t_\m{D}-2m_N^⊤(βΦ^⊤t_\m{D})+m_N^⊤(αI+βΦ^⊤Φ)m_N\\
&\quad=β(t_\m{D}^⊤t_\m{D}-2m_N^⊤Φ^⊤t_\m{D}+m_N^⊤Φ^⊤Φm_N)+αm_N^⊤m_N=β\n{t_\m{D}-Φm_N}^2+α\n{m_N}^2=\green{2E(m_N)}\\
&\dg{⇒}\brown{p(t_\m{D}|α,β)}=∫p(t_\m{D},w|α,β)\ dw=(\/{β}{2π})^{N/2}(\/{α}{2π})^{M/2}\e{-\green{E(m_N)}}∫\e{-\/{1}{2}\yello{(w-m_N)^⊤S_N^{-1}(w-m_N)}}\ dw\\
&\quad=(\/{β}{2π})^{N/2}(\/{α}{2π})^{M/2}(2π)^{M/2}\det(S_N)^{1/2}\e{-\green{E(m_N)}}\\
\dg{⇒}&\dg{\t{Adam-Eve convolution: another approach to get the marginal likelihood}}\\
\dg{⇒}&\brown{p(t_\m{D}|α,β)}=∫p(t_\m{D},w|α,β)\ dw=∫p(t_\m{D}|w,β)p(w|α)\ dw=∫\Normal(t_\m{D}|Φw,β^{-1}I)\Normal(w|0,α^{-1}I)\ dw=\Normal(t_\m{D}|\E[t_\m{D}],\yello{\V[t_\m{D}]})\\
&\dg{⇒}\E[t_\m{D}]=\E[\E[t_\m{D}|w]]=\E[Φw]=0\\
&\dg{⇒}\V[t_\m{D}]=\E[\V[t_\m{D}|w]]+\V[\E[t_\m{D}|w]]=\E[β^{-1}I]+\V[Φw]\dg{⇒}\yello{\V[t_\m{D}]=β^{-1}I+α^{-1}ΦΦ^⊤}\\
&\dg{⇒}\dg{\t{Find $\yello{\V[t_\m{D}]}^{-1}$ by mapping to Woodbury identity }}\yello{\V[t_\m{D}]}=A_{22}+A_{21}A_{11}^{-1}A_{12}\dg{⇒}A=\BMαI&Φ^⊤\\Φ&β^{-1}I \EM\\
&\dg{⇒}(A_{22}+A_{21}A_{11}^{-1}A_{12})^{-1}=A_{22}^{-1}-A_{22}^{-1}A_{21}(A_{11}+A_{12}A_{22}^{-1}A_{21})^{-1}A_{12}A_{22}^{-1}\dg{⇒}\yello{\V[t_\m{D}]}^{-1}=βI-β^2Φ(αI+βΦ^⊤Φ)Φ^⊤\\
&\dg{⇒}t_\m{D}^⊤\yello{\V[t_\m{D}]}^{-1}t_\m{D}=t_\m{D}^⊤[βI-β^2Φ(αI+βΦ^⊤Φ)Φ^⊤]t_\m{D}=βt_\m{D}^⊤t_\m{D}-β^2t_\m{D}^⊤ΦS_N^{-1}Φ^⊤t_\m{D}=\green{βt_\m{D}^⊤t_\m{D}-m_N^⊤S_N^{-1}m_N}\\
&\quad=βt_\m{D}^⊤t_\m{D}-2m_N^⊤(S_N^{-1}m_N)+m_N^⊤S_N^{-1}m_N=βt_\m{D}^⊤t_\m{D}-2m_N^⊤(βΦ^⊤t_\m{D})+m_N^⊤(αI+βΦ^⊤Φ)m_N\\
&\quad=β(t_\m{D}^⊤t_\m{D}-2m_N^⊤Φ^⊤t_\m{D}+m_N^⊤Φ^⊤Φm_N)+αm_N^⊤m_N=β\n{t_\m{D}-Φm_N}^2+α\n{m_N}^2=\green{2E(m_N)}\\
&\dg{⇒}\det(\yello{\V[t_\m{D}]})=\det(β^{-1}I+α^{-1}ΦΦ^⊤)=\det(β^{-1}I_N)\det(α^{-1}I_M)\det(αI_N+βΦΦ^⊤)=β^{-N}α^{-M}\det(αI_M+βΦ^⊤Φ)\\
&\quad=β^{-N}α^{-M}\det(S_N^{-1})\\
&\dg{⇒}\brown{p(t_\m{D}|α,β)}=\Normal(t_\m{D}|0,\yello{\V[t_\m{D}]})=\/{1}{(2π)^{N/2}}\det(\yello{\V[t_\m{D}]})^{-1/2}\e{-\green{E(m_N)}}\\
&\quad=\/{1}{(2π)^{N/2}}α^{M/2}β^{N/2}\det(S_N)^{1/2}\e{-\green{E(m_N)}}\\
\dg{⇒}&\red{\ln \brown{p(t_\m{D}|α,β)}=\/{M}{2}\ln α+\/{N}{2}\ln β-E(m_N)+\/{1}{2}\ln\det(S_N)-\/{N}{2}\ln(2π)}\dg{-\t{marginal likelihood/evidence}}\\
\EA$

- **Maximizing evidence**: marginal likelihood (evidence) exposes measurement error $β^{-1}$ and isotropic prior variance $α^{-1}$ along with $m_N$ and $S_N$ to solve iteratively.
$\BA
\dg{⇒}&\red{γ:=\sum_{i=1}^M\/{λ_i}{α+λ_i}}\\
\dg{⇒}&\ln\det(S_N)=-\ln\det(S_N^{-1})=-\ln\prod_{i=1}^M(α+λ_i)=-\sum_{i=1}^M\ln(α+λ_i)\dg{⇒}\yello{\/{∂}{∂α}\ln\det(S_N)=-\sum_{i=1}^M\/{1}{α+λ_i}}\\
&\dg{⇒}(βΦ^⊤Φ)u_i=λ_iu_i\dg{⇒}\/{∂λ_i}{∂β}u_i=\/{∂(βΦ^⊤Φ)}{∂β}u_i=(Φ^⊤Φ)u_i=\/{λ_i}{β}u_i\dg{⇒}\/{∂}{∂β}λ_i=\/{λ_i}{β}\dg{⇒}\yello{\/{∂}{∂β}\ln\det(S_N)=-\sum_{i=1}^M\/{1}{α+λ_i}\/{λ_i}{β}=-\/{γ}{β}}\\
\dg{⇒}&E(m_N)=\/{1}{2}β\n{t_\m{D}-Φm_N}^2+\/{1}{2}α\n{m_N}^2\dg{⇒}\/{∂}{∂α}E(m_N)=\/{∂E}{∂α}+\/{∂E}{∂w}|_{w=m_N}\/{∂m_N}{∂α}=\/{∂E}{∂α}=\/{1}{2}\n{m_N}^2\\
&\dg{⇒}\/{∂}{∂β}E(m_N)=\/{∂E}{∂β}+\/{∂E}{∂w}|_{w=m_N}\/{∂m_N}{∂β}=\/{1}{2}\n{t_\m{D}-Φm_N}^2\dg{-\t{in both cases }\/{∂E}{∂w}|_{w=w_\t{MAP}}=0\t{ for posterior error curve}}\\
\dg{⇒}&\/{∂}{∂α}\ln \brown{p(t_\m{D}|α,β)}=\/{M}{2α}-\/{1}{2}\n{m_N}^2-\/{1}{2}\sum_{i=1}^M\/{1}{α+λ_i}=0\dg{⇒}α\n{m_N}^2=M-\sum_{i=1}^M\/{α}{α+λ_i}=\sum_{i=1}^M\/{λ_i}{α+λ_i}\dg{⇒}\blue{α^+=\/{γ}{\n{m_N}^2}}\\
\dg{⇒}&\/{∂}{∂β}\ln \brown{p(t_\m{D}|α,β)}=\/{N}{2β}-\/{1}{2}\n{t_\m{D}-Φm_N}^2-\/{γ}{2β}=0\dg{⇒}\blue{\/{1}{β^+}=\/{1}{N-γ}\n{t_\m{D}-Φm_N}^2}
\EA$

- **Bayesian regularization**: Non-Bayesian regularized $(Φ^⊤Φ+λI)^{-1}Φ^⊤t_\m{D}$ finds adhoc $λ=\/{α}{β}$ by cross-validation, but Bayesian regularized $w_\t{MAP}$ finds $α,β$ from training data by maximizing model evidence.
$\BA
\dg{⇒}&∇_w\ln p(w|t_\m{D})=∇_w[-\/{α}{2}\n{w}^2-\/{β}{2}\n{t_\m{D}-Φw}^2]=-αw^⊤-β(t_\m{D}-Φw)^⊤(-Φ)=-αw^⊤+βt_\m{D}^⊤Φ-βw^⊤Φ^⊤Φ=0\\
\dg{⇒}&w^⊤(αI+βΦ^⊤Φ)=βt_\m{D}^⊤Φ\dg{⇒}w^*=β(αI+βΦ^⊤Φ)^{-1}Φ^⊤t_\m{D}\\
\dg{⇒}&\red{w_\t{MAP}=\arg\max_wp(w|t_\m{D})=m_N=β(αI+βΦ^⊤Φ)^{-1}Φ^⊤t_\m{D}}
\EA$

- **Well-determined parameters**: PCA of empirical covariance $\purple{βΦ^⊤Φ}=UΛU^⊤$ gives anisotropic shrinkage along each principal axis
$\red{u_i^⊤w_\t{MAP}=\/{λ_i}{α+λ_i}u_i^⊤w_\t{ML}}$.
Well determined parameters have high $λ_i$ to pull $u_i^⊤w_\t{MAP}→u_i^⊤w_\t{ML}$ vs $0$.
$\BC
\dg{⇒}&\yello{w_\t{ML}=(Φ^⊤Φ)^{-1}Φ^⊤t_\m{D}}\quad\dg{⇒}\green{w_\t{MAP}=β(αI+βΦ^⊤Φ)^{-1}Φ^⊤t_\m{D}}=(αI+βΦ^⊤Φ)^{-1}(\purple{βΦ^⊤Φ})\yello{w_\t{ML}}\\
\dg{⇒}&u_i^⊤(αI+βΦ^⊤Φ)^{-1}(βΦ^⊤Φ)=[(αI+βΦ^⊤Φ)^{-1}(\purple{βΦ^⊤Φ})]^⊤u_i=(\purple{βΦ^⊤Φ})\/{1}{α+λ_i}u_i=\/{λ_i}{α+λ_i}u_i\\
\dg{⇒}&u_i^⊤\green{w_\t{MAP}}=u_i^⊤(αI+βΦ^⊤Φ)^{-1}(\purple{βΦ^⊤Φ})w_\t{ML}=\/{λ_i}{α+λ_i}u_i^⊤\yello{w_\t{ML}}\dg{⇒}u_i^⊤\green{w_\t{MAP}}=\red{\/{λ_i}{α+λ_i}}u_i^⊤\yello{w_\t{ML}}
\EC$

  - **Effective number of parameters** $\red{γ=\sum_{i=1}^M\/{λ_i}{α+λ_i}}$ is a count of well-determined parameters, or degrees of freedom used to determine parameters.
  The data has enough strong eigenvalues to determine $γ$ of $M$ parameters.

  - $N≫M\dg{⇒}Φ^⊤Φ=\sum_iϕ(x_i)ϕ(x_i)^⊤$ has high eigenvalues: $γ=M,\ α=\/{M}{\n{m_N}^2},\ β=\/{N}{\n{t_\m{D}-Φm_N}^2)}$

**Bayesian model selection**: Suppose we have models $\{M_1,...,M_L\}$ and training data $\m{D}$. Then posterior distribution over models is
$\purple{p(M_i|\m{D})}∝p(M_i)\yello{p(\m{D}|M_i)}$ and prediction distribution is
$p(t_k|x_k,\m{D})=\sum_ip(t_k|x_k,M_i,\m{D})\purple{p(M_i|\m{D})}$.
Model selection chooses the model with highest posterior $\purple{p(M_i|\m{D})}$, which is equivalent to choosing highest evidence $\yello{p(\m{D}|M_i)}$.

- Model evidence is
$\yello{p(\m{D}|M_i)}=∫p(\m{D}|w,M_i)p(w|M_i)\ dw$.
To approximate the model evidence integral, let
$Δw_\t{posterior}< Δw_\t{prior}$ be widths of the modal peaks.

  - Let $w∈ℝ$ then
  $p(\m{D})Δw_\t{prior}≈p(\m{D}|w_\t{MAP})Δw_\t{posterior}$
  $\dg{⇒}p(\m{D})=∫p(\m{D}|w)p(w)\ dw≈p(\m{D}|w_\t{MAP})\/{Δw_\t{posterior}}{Δw_\t{prior}}$
  $\dg{⇒}\ln p(\m{D})=\ln p(\m{D}|w_\t{MAP})+\ln\/{Δw_\t{posterior}}{Δw_\t{prior}}$.
  Although data fit improves with parameters $\ln p(\m{D})< \ln p(\m{D}|w_\t{MAP})$,
  the second term $\ln\/{Δw_\t{posterior}}{Δw_\t{prior}}< 0$ is a penalty on complexity.

  - Let $w∈ℝ^M$ then $\ln p(\m{D})=\ln p(\m{D}|w_\t{MAP})+M\ln\/{Δw_\t{posterior}}{Δw_\t{prior}}$
  and the complexity penalty is greater.

- **Bayes factor** $\/{\yello{p(\m{D}|M_1)}}{\yello{p(\m{D}|M_2)}}$ is the ratio of evidence across different models. Assuming $M_1$ is the true distribution and $M_2$ is an incorrect model, then we can average the Bayes factor across all possible datasets $\m{D}$ to get the KL divergence $\KL(\yello{p(\m{D}|M_1)}\|\yello{p(\m{D}|M_2)})=∫\ln\/{\yello{p(\m{D}|M_1)}}{\yello{p(\m{D}|M_2)}}\yello{p(\m{D}|M_1)}\ d\m{D}$

# ESL: 3 Linear Models for Regression

**Setup**: $y_i=\E[y_i|x_i]+ϵ=β_0+x_i^⊤β_{1..p}+ϵ$ from training data $x∈ℝ^{N×p},y∈ℝ^N$ and $X=\BM\v{1}&x\EM∈ℝ^{N×(p+1)}$.
OLS estimate $\BA
\hat{β}=\arg\min_β\t{RSS}(β)
=\arg\min_β\n{y-β_0\v{1}-xβ_{1..p}}^2
=\arg\min_β\n{y-Xβ}^2
\dg{⇒}\red{\hat{β}=(X^⊤X)^{-1}X^⊤y}∈ℝ^{p+1}
\EA$.

- **Degrees of freedom**: OLS estimates $\hat{β}∈ℝ^{p+1}$ with $p+1$ degrees of freedom.
Statistical definition of degrees of freedom is how sensitive the fitted values $\hat{y}_i$ are wrt changes in observed values $y_i$.
Here $\hat{y}=Hy\dg{⇒}\hat{y}_i=H_iy$ where $H=X(X^⊤X)^{-1}X^⊤∈ℝ^{N×N}$ is the hat matrix.

  - $\BA
  \red{\t{df}(\hat{y})=\sum_{i=1}^N\/{\V[\hat{y}_i,y_i]}{\V[y_i]}}
  =\sum_i\/{\V[\sum_jH_{ij}y_j,y_i]}{\V[y_i]}
  =\sum_i\/{\V[H_{ii}y_i,y_i]}{\V[y_i]}
  =\sum_iH_{ii}
  \dg{⇒}\red{\t{df}(\hat{y})=\tr(H)}
  \EA$.

- Assume $\V[y_i]=σ^2$ and uncorrelated.
Then $\red{\hat{β}∼\Normal(β,(X^⊤X)^{-1}σ^2)}$ and $\V[\hat{β}_j]=v_j\hat{σ}^2$ where $v_j=[(X^⊤X)^{-1}]_{jj}$.
$\BA
\dg{⇒}&\V[\hat{β}]=\V[(X^⊤X)^{-1}X^⊤y]=[(X^⊤X)^{-1}X^⊤]\V[y][(X^⊤X)^{-1}X^⊤]^⊤=(X^⊤X)^{-1}X^⊤σ^2IX(X^⊤X)^{-1}=(X^⊤X)^{-1}σ^2
\EA$

  - $\hat{σ}^2=\/{1}{N-p-1}\sum_i(y_i-\hat{y}_i)^2$
  $\dg{⇒}\blue{\/{(N-p-1)\hat{σ}^2}{σ^2}∼χ_{N-p-1}^2}$.

  - Quadratic form: $Z^2=\/{(x-μ)^2}{σ^2}∼χ_1^2$
  $\dg{⇒}\blue{(\hat{β}-β)^⊤Σ^{-1}(\hat{β}-β)∼χ_{p+1}^2}$

  - $H_0:β_j=0$ Z-test $\blue{z_j=\/{\hat{β}}{\hat{σ}\sqrt{v_j}}}$.

- **Gauss-Markov theorem**: Linear estimator $\hat{θ}=c^⊤y$ is a linear combination of observed response $y$.  LS estimator $\hat{θ}=a^⊤\hat{β}=a^⊤(x^⊤x)^{-1}x^⊤y$ has the smallest variance among linear unbiased estimators (BLUE).

  - Quality of estimator should be measured by estimator MSE, which measures how close estimates are to true parameter values
  $\t{MSE}(\hat{θ})=\E[(\hat{θ}-θ)^2]$
  $=\V[\hat{θ}]+(\E[\hat{θ}]-θ)^2$
  even if biased $\E[\hat{θ}]-θ\neq0$.

**Gram-Schmidt multiple regression**: $p=1$ no intercept regressing $y$ on $x$ to get $\hat{β}=\/{x^⊤y}{x^⊤x}$ with residual $r=y-x\hat{β}$.

- Consider $\BM 1&x_1\\\vdots&\vdots\\1&x_N\EM$. Regressing $x$ on $z_0=\v{1}$ to get $\bar{x}=\/{\v{1}^⊤x}{\v{1}^⊤\v{1}}$ with residual $z=x-\v{1}\bar{x}$ centering $x$.
Now $x=z+\v{1}\bar{x}$ is the sum of two orthogonal components satisfying $(x-\bar{x}\v{1})^⊤\v{1}=0$.
Then $\hat{β}_1=\/{z^⊤y}{z^⊤z}$

- Consider $\BM 1&x_{11}&x_{21}\\\vdots&\vdots&\vdots\\1&x_{1N}&x_{2N}\EM$.
$\BC
\t{Regress $x_1$ on $\yello{z_0}=\green{v_0}=\v{1}$}&\hat{γ}_{01}=\/{\yello{z_0}^⊤x_1}{\yello{z_0}^⊤\yello{z_0}}=\bar{x}_1&\yello{z_1}=x_1-\hat{γ}_{01}\yello{z_0}\\
\t{Regress $x_2$ on $\yello{z_0},\yello{z_1}$}&\hat{γ}_{02}=\/{\yello{z_0}^⊤x_2}{\yello{z_0}^⊤\yello{z_0}}=\bar{x}_2&\green{v_1}=x_2-\hat{γ}_{02}\green{v_0}\\
&\hat{γ}_{12}=\/{\yello{z_1}^⊤x_2}{\yello{z_1}^⊤\yello{z_1}}&\yello{z_2}=x_2-\hat{γ}_{02}\yello{z_0}-\hat{γ}_{12}\yello{z_1}&\dg{⇒}\hat{β}_2=\/{\yello{z_2}^⊤y}{\yello{z_2}^⊤\yello{z_2}}\\
\t{Regress $x_1$ on $\green{v_1}$}&\hat{γ}_{11}=\/{\green{v_1}^⊤x_1}{\green{v_1}^⊤\green{v_1}}&\green{v_2}=x_1-\hat{γ}_{01}\green{v_0}-\hat{γ}_{11}\green{v_1}&\dg{⇒}\hat{β}_1=\/{\green{v_2}^⊤y}{\green{v_2}^⊤\green{v_2}}
\EC$

  - Rearrange to have $x_1$ as the last column to find $\hat{β}_1$.
  Similarly have $\v{1}$ as the last column to find $\hat{β}_0$.
  
  - **Deflation**: Projecting $y$ onto $z_p$ gives $\hat{β}_p$ because $z_p$ represents the only unique independent information in $x_p$ after removing influence of $x_0,...,x_{p-1}$ by deflating away $z_0,...,z_{p-1}$ from $x_p$.

- Consider $x∈ℝ^p$. $\BC
1.&z_0=x_0=\v{1}\\
2.&\t{for }j=1,2,...,p\\
&\t{Regress }x_j\t{ on }z_0,...,z_{j-1}\t{ to produce }\hat{γ}_{lj}=\/{z_l^⊤x_j}{z_l^⊤z_l}, z_j=x_j-\sum_{k=0}^{j-1}\hat{γ}_{kj}z_k\\
3.&\red{\hat{β}_p=\/{z_p^⊤y}{z_p^⊤z_p}}\t{ where }\V[\hat{β}_p]=\/{σ^2}{z_p^⊤z_p}
\EC$

- Gram-Schmidt orthogonalization of $X=\BM\v{1}&x\EM∈ℝ^{N×(p+1)}$ finds $Z=\BM z_0&...&z_p\EM∈ℝ^{N×(p+1)}$ as orthogonal basis for the column space of $X=ZΓ$.
Normalizing $Z$ to have unit length columns using diagonal $D$ yields QR decomposition $X=(ZD^{-1})(DΓ)=QR$ where $Q∈ℝ^{N×(p+1)}$ and $Q^⊤Q=I$.
Then $\red{\hat{β}=R^{-1}Q^⊤y}$ and $\red{\hat{y}=QQ^⊤y}$.

**Shrinkage**: shrinkage methods reduce variance by shrinking the coefficients toward zero. Because regression is not unit invariant, the model will pull intercept toward zero if it is included in ridge penalty thus resulting in different slopes for Celsius vs Kelvin. Standard practice is to center $x$ and $y$ so that $\bar{x}=\bar{y}=0$ and $\hat{β}_0=0$. Let $X∈ℝ^{N×p}$ and $\hat{β}∈ℝ^p$.

- **Ridge**: $\red{\hat{β}^\t{ridge}=\arg\min_β(\n{y-Xβ}^2+λ\n{β}^2)}$
$=\blue{(X^⊤X+λI)^{-1}X^⊤y}$. Let $X=UΣV^⊤$ with thin $U∈ℝ^{N×D}$. Then $X^⊤X=VΣ^2V^⊤$
$\dg{⇒}\BC
X\hat{β}^\t{ls}=X(X^⊤X)^{-1}X^⊤y=UΣV^⊤(VΣ^2V^⊤)^{-1}VΣ^⊤U^⊤y=UU^⊤y\dg{⇒}\red{\hat{y}^\t{ls}=\sum_{j=1}^pu_ju_j^⊤y}\\
X\hat{β}^\t{ridge}=X(X^⊤X+λI)^{-1}X^⊤y=UΣ(Σ^2+λI)^{-1}Σ^⊤U^⊤y\dg{⇒}\red{\hat{y}^\t{ridge}=\sum_{j=1}^pu_j\/{σ_j^2}{σ_j^2+λ}u_j^⊤y}
\EC$

  - **Anisotropic**: OLS regression is isotropic and uniform along all axes. $UU^⊤$ is projection matrix onto the column space of $X$. Along each axis $u_ju_j^⊤y$ is projection of $y$ onto unit vector $u_j$. Ridge regression is anisotropic and axis-dependent. Along each axis $\/{σ_j^2}{σ_j^2+λ}u_ju_j^⊤y$ applies shrinkage according to singular values.

  - **PCA**: Let $X^⊤X=VΣ^2V^⊤$, then code $Z_M=XV_M=U_MΣ_M∈ℝ^{N×M}$. The first principal component is the first column $z_1=Xv_1∈ℝ^N$ with sample variance $\V[z_1]=\/{1}{N}z_1^⊤z_1$
  $=\/{1}{N}v_1^⊤X^⊤Xv_1$
  $=\/{σ_1^2}{N}v_1^⊤v_1$
  $\dg{⇒}\blue{\V[z_1]=\/{σ_1^2}{N}}$.
  Then $Σ^2$ in $\/{σ_j^2}{σ_j^2+λ}$ are just variances of the data along principal axes.
  Higher variance axes are easy to estimate slopes - ridge leaves this alone. Low variance slopes can flip wildly to effect of noise - ridge aggressively attunates the coordinates along this axis toward zero (robust).

  - **Degrees of freedom**: Anisotropic shrinkage pulls each parameter toward zero and reduces freedom.
  $\BA
  \t{df}(λ)=\tr[H_λ]
  =\tr[X(X^⊤X+λI)^{-1}X^⊤]
  =\tr[UΣ(Σ^2+λI)^{-1}Σ^⊤U^⊤]
  =\tr[Σ^2(Σ^2+λI)^{-1}]
  \dg{⇒}\blue{\t{df}(λ)=\sum_j\/{σ_j^2}{σ_j^2+λ}}
  \EA$
  is the effective degrees of freedom of the ridge regression fit.

- **Lasso**: $\red{\hat{β}^\t{lasso}=\arg\min_β(\n{y-Xβ}_2^2+λ\n{β}_1)}$ uses penalty $\sum_j|β_j|$ instead of $\sum_jβ_j^2$. It has no closed-form solution, and requires convex optimization to solve. In 2004 LAR was introduced to compute lasso. Then in 2007 coordinate descent became the gold standard algorithm for lasso and elastic net.

- **Elastic net**: $λ\sum_{j=1}^p(αβ_j^2+(1-α)|β_j|)$ has sharp corners while $\sum_j|β_j|^q$ with q=1.x has more rounded corners.

- **3.6**: Under prior $β∼\Normal(0,τI)$ and likelihood $y∼\Normal(Xβ,σ^2I)$, the mean/mode of the posterior distribution is $\hat{β}^\t{ridge}$.
$\BA
\dg{⇒}&p(β|y,X)∝p(y|X,β)p(β)=\Normal(y|Xβ,σ^2I)\Normal(β|0,τI)
\dg{⇒}E=-\/{1}{2σ^2}\n{y-Xβ}^2-\/{1}{2τ}\n{β}^2\\
&=-\/{1}{2}[β^⊤(\/{1}{σ^2}X^⊤X+\/{1}{τ}I)β-2β^⊤X^⊤y]+c_1
\dg{⇒}p(β|y,X)=\Normal(β|m_N,S_N),\BC
S_N^{-1}=\/{1}{σ^2}X^⊤X+\/{1}{τ}I\\
m_N=S_NX^⊤y=(\/{1}{σ^2}X^⊤X+\/{1}{τ}I)^{-1}X^⊤y
\EC
\EA$
$\BA
\dg{⇒}&p(β,y|X)=p(y|X,β)p(β)=\Normal(y|Xβ,σ^2I)\Normal(β|0,τI)
\dg{⇒}E=-\/{1}{2σ^2}\n{y-Xβ}^2-\/{1}{2τ}\n{β}^2\\
&=-\/{1}{2}[β^⊤(\/{1}{σ^2}X^⊤X+\/{1}{τ}I)β-2β^⊤X^⊤y+\/{1}{σ^2}y^2]=-\/{1}{2}\BM β\\y\EM^⊤\BM\/{1}{σ^2}X^⊤X+\/{1}{τ}I&-X^⊤\\-X&\/{1}{σ^2}I\EM\BM β\\y\EM
\dg{⇒}\BC
Σ_{β|y}=(\/{1}{σ^2}X^⊤X+\/{1}{τ}I)^{-1}\\
μ_{β|y}=(\/{1}{σ^2}X^⊤X+\/{1}{τ}I)^{-1}X^⊤y
\EC
\EA$

**Least angle regression (LAR)**: Standardize $X∈ℝ^{N×p}$ (zero mean, unit norm). Center $y$ (zero mean). Set $β=0$. Calculate residual $r_0=y-\bar{y}$. Find column $x_j$ most correlated with $r_0$ and add to active predictor set $\m{A}_1=\{x_j\}$.

- **Step-$k$** has prediction $\blue{\hat{y}_k=X_{\m{A}_k}β_{\m{A}_k}}$, residual $\red{r_k=y-X_{\m{A}_k}β_{\m{A}_k}}$, and direction $\red{δ_k=(X_{\m{A}_k}^⊤X_{\m{A}_k})^{-1}X_{\m{A}_k}^⊤r_k}$.

  - $\BA
  \Corr(x_j,r_k)=\/{\V[x_j,r_k]}{\sqrt{\V[x_j]\V[r_k]}}=\/{\/{1}{N}x_j^⊤r_k}{\sqrt{\/{1}{N}\V[r_k]}}=\/{1}{\sqrt{N\V[r_k]}}x_j^⊤r_k
  \dg{⇒}\blue{c_k=X_{\m{A}_k}^⊤r_k}
  \EA$
  represents $\Corr(X_{\m{A}_k},r_k)$.

  - $δ_k∈ℝ^{|\m{A}_k|}$ is the direction of coefficient update.
  As $x_j$ enters $A_k$, coefficient $β_j$ has the same sign as its correlation with residual $x_j^⊤r_k$.
  $\BA
  \dg{⇒}&δ_k=(X_{\m{A}_k}^⊤X_{\m{A}_k})^{-1}\brown{X_{\m{A}_k}^⊤r_k}\dg{⇒}δ_{kj}=\/{\brown{x_j^⊤r_k}}{x_j^⊤x_j}\dg{⇒}β_j(α)=αδ_{kj}=α\/{\brown{x_j^⊤r_k}}{x_j^⊤x_j}
  \EA$

- **Updates**: During step-$k$ gradually increase step size $α=0→1$ and grow active coefficients $\red{β_{\m{A}_k}(α)=β_{\m{A}_k}+αδ_k}$ on top of the model states at the start of step-$k$.
$\BA
\dg{⇒}&r(α)=y-X_{\m{A}_k}β_{\m{A}_k}(α)=y-X_{\m{A}_k}(β_{\m{A}_k}+αδ_k)=r_k-αX_{\m{A}_k}δ_k
\dg{⇒}\brown{r(α)=r_k-αX_{\m{A}_k}δ_k}\\
\dg{⇒}&c(α)=X_{\m{A}_k}^⊤r(α)=X_{\m{A}_k}^⊤(r_k-αX_{\m{A}_k}δ_k)=c_k-αX_{\m{A}_k}^⊤X_{\m{A}_k}δ_k=c_k-αX_{\m{A}_k}^⊤X_{\m{A}_k}(X_{\m{A}_k}^⊤X_{\m{A}_k})^{-1}X_{\m{A}_k}^⊤r_k\dg{⇒}\brown{c(α)=(1-α)c_k}\\
\dg{⇒}&\hat{y}(α)=X_{\m{A}_k}β_{\m{A}_k}(α)=X_{\m{A}_k}(β_{\m{A}_k}+αδ_k)=\hat{y}_k+αX_{\m{A}_k}δ_k\dg{⇒}\brown{\hat{y}(α)=\hat{y}_k+αu_k}
\EA$

  - As coefficients move away from zero, the residual shrinks by $αX_{\m{A}_k}δ_k$ and the correlation of all active predictors with the residual are tied to $c(α)$ and decrease at the same rate $(1-α)$.

  - Equiangular direction $\blue{u_k=X_{\m{A}_k}δ_k}∈ℝ^N$ is the direction of prediction update that ensures the correlation of all active predictors with the residual decrease at the same rate. Geometrically it sits exactly in the middle of all active predictors, and creates the least possible angle that is equal among all of them. For example if k=3, then $u_k$ is the vector right down the center of the cone formed by columns $x_1,x_2,x_3$.

  - $α∈[0,1]$. When $α=1$ we have the OLS coefficients on the active set $X_{\m{A}_k}$
  $\BA
  β_{\m{A}_k}+δ_k=β_{\m{A}_k}+(X_{\m{A}_k}^⊤X_{\m{A}_k})^{-1}X_{\m{A}_k}^⊤r_k=β_{\m{A}_k}+(X_{\m{A}_k}^⊤X_{\m{A}_k})^{-1}X_{\m{A}_k}^⊤(y-X_{\m{A}_k}β_{\m{A}_k})=(X_{\m{A}_k}^⊤X_{\m{A}_k})^{-1}X_{\m{A}_k}^⊤y\dg{⇒}\green{β_{\m{A}_k}(1)=\hat{β}_{\m{A}_k}^\t{ls}}
  \EA$

- **Next step**: As $α$ increases and $c(α)$ decreases, another column $x_j'$ satisfying $|x_j'^⊤r(α)|=c(α)$ is added to $\m{A}_{k+1}$ to start next step. As such LAR always takes $p$ steps to get to the full OLS coefficients.

  - Along the way LAR gives a sparse model with 1, 2, ..., p predictors with non-zero coefficients smoothly growing toward their OLS values.

- **LAR-Lasso** adds an additional rule: when $β_j$ shrinks to exact zero, $x_j$ is removed from $\m{A}_{k+1}$ to start the next step.
LAR-Lasso takes more than $p$ steps to get the full OLS coefficients, and along the way traces the entire continuous path of lasso-regularized coefficients with $λ=∞→0$. Any $λ>\max_j|x_j^⊤y|$ just gives $β=0$.

  - Lasso ojective $\min_β(\/{1}{2}\n{y-Xβ}_2^2+λ\n{β}_1)$ using subgradient $∇\n{β}_1$ has KKT stationarity conditions
  $\BA
  ∇_β=(y-Xβ)^⊤(-X)+λ∇\n{β}_1^⊤=0\dg{⇒}X^⊤r=λ∇\n{β}_1
  \dg{⇒}\BC
  β_j\neq0&∂_j\n{β}_1=\t{sign}(β_j)&\brown{x_j^⊤r=λ⋅\t{sign}(β_j)}\\
  β_j=0&∂_j\n{β}_1∈[-1,1]&\brown{|x_j^⊤r|≤λ}
  \EC\\
  \EA$.
  LAR-lasso active predictors ($β_j\neq0$) have tied correlation $\brown{x_j^⊤r(α)=c(α)⋅\t{sign}_j}$ while inactive predictors ($β_j=0$) satisfy $\brown{|x_j'^⊤r(α)|≤c(α)}$.
  LAR-lasso satisfies KKT stationarity as long as $\t{sign}_j=\t{sign}(β_j)$, which only fails during $β_j$ zero-crossing.
  Kicking out $x_j$ from $\m{A}_k$ prevents the failure of alignment with KKT stationarity.

  - LAR-lasso has same order of computation as a least squares fit with QR decomposition of $X$.

- **Degrees of freedom**: At step-$k$ the degrees of freedom is exactly $k$.

**Principal components regression**: Let $X=UΣV^⊤∈ℝ^{N×p}$ with thin $U∈ℝ^{N×p}$, then principal components $Z_M=XV_M$ where $M≤p$ already has orthogonal columns and doesn't need deflation.
Coefficients along principal axes $\red{\hat{θ}_j=\/{z_j^⊤y}{z_j^⊤z_j}}$ can be translated to standard basis
$\blue{\hat{β}_M^\t{pcr}=V_M\hat{θ}}=\sum_{j=1}^M\hat{θ}_jv_j$.
Predictions are $\red{\hat{y}_M^\t{pcr}=\bar{y}\v{1}+Z_M\hat{θ}}=\bar{y}\v{1}+\sum_{j=1}^M\hat{θ}_jz_j$.

- Ridge regression scales principal components of $X$ by scaling factor $\/{σ_j^2}{σ_j^2+λ}$ while principal components regression truncates lower $p-M$ components of $X$ using scaling factor $\{1,0\}$.


# PRML: 4 Linear Models for Classification

**Discriminant function** takes an input $x∈ℝ^D$ and assigns to a class $C_1,...,C_K$.
Binary problem has target $t∈\{1_{C_1},0_{C_2}\}$ while $K>2$ use one-hot $t∈\{0,1\}^K$.
Linear regression has prediction function $y(x)=w^⊤x+w_0$.
Classification has **activation function** $y(x)=f(w^⊤x+w_0)$ mapping linear output to a probability (sigmoid).


- **Binary linear discriminant** has a decision hyperplane $y(x)=w^⊤x+w_0=0$ as class boundary. A hyperplane is defined to be
$\red{\{x|w^⊤x=-w_0\}=\{x|x∈x_0+w^\perp\}}$ where
$x_0=-\/{w_0}{\n{w}}\/{w}{\n{w}}$ is the projection of origin. Distance from hyperplane to any point $x_1$ is the distance from $x_0$ to projection of $x_1$ onto $w$, which is
$\n{\/{w^⊤x_1}{\n{w}}\/{w}{\n{w}}-x_0}=\/{y(x_1)}{\n{w}}$.

  - Naively extending to $K>2$ multiclass by combining multiple binary classifiers.
  **One-vs-rest** trains $K$ classifiers independently.
  **One-vs-one** trains $K(K-1)/2$ classifiers independently.
  There will be ambiguous regions in the input space with points claimed by multiple classifiers and points claimed by no classifier.

- **$K$-class linear discriminant** trains $K$ discriminant functions simultaneously
$\red{y_k(x)=w_k^⊤x+w_{k0}}$
with decision region
$\red{R_k=\{x|y_k(x)≥y_j(x)\ ∀j\}}$ and decision boundary
$y_k(x)=y_j(x)$
$\dg{⇒}(w_k-w_j)^⊤x+(w_{k0}-w_{j0})=0$.

  - The decision region is convex.
  Proof: $R_k=⋂_j\{x|(w_j-w_k)^⊤x≤w_{k0}-w_{j0}\}$ is an intersection of halfspaces.

**OLS linear discriminant**: $\BA
W=\BM w_{10}&...&w_{K0}\\w_1&...&w_K\EM∈ℝ^{(D+1)×K},X∈ℝ^{N×(D+1)}\dg{⇒}XW=\BM y_1(x_1)&...&y_K(x_1)\\\vdots&\ddots&\vdots\\y_1(x_N)&...&y_K(x_N)\EM∈ℝ^{N×K}
\EA$ is trained against $T∈\{0,1\}^{N×K}$ by minimizing RSS error $\red{\/{1}{2}\n{XW-T}_F^2}=\/{1}{2}\tr[(XW-T)^⊤(XW-T)]$.

- $\red{W^*=(X^⊤X)^{-1}X^⊤T}$ where every row $y(x_i)^⊤=(XW)_i∈ℝ^K$ guarantees $\v{1}^⊤y(x_i)=1$ but not $y(x_i)≽0$ and is therefore not a probability distribution.

- OLS coefficients represents the normal of the separating hyperplane. On a 2D scatterplot of $x_1,x_2$ axes, regression predicts $x_2$ based on $x_1$.
Binary classification uses $x_1$ and $x_2$ to predict a third axis where $t=0$ is lower and $t=1$ is higher.
The boundary is the waterline at half-height.

- **Too correct problem** (Figure 4.4):
OLS tilts decision boundary to outlier training points placed far from boundary because it assumes $t_i∼\Normal(y(x_i),σ^2)∈(-∞,∞)$ full range against actual $t_i∈\{0,1\}$ discrete range.
Outlier $x_i$ makes large predictions far from $\{0,1\}$.
For example $y(x_i)=5$ has penalty $\n{5-1}^2$ (too correct), thus tilts the boundary.

- **Masking problem** (Figure 4.5): $K=3$ parallel clusters of data has targets (1,0,0), (0,1,0), and (0,0,1). $y_1(x)$ and $y_3(x)$ easily predicts the two outer classes. Middle column goes from 0 to 1 to 0, but affine $y_2(x)$ cannot bend and therefore fits to a middling constant, say, $y_2(x)=0.33$, and is easily masked by $y_1(x)$ and $y_2(x)$ in $\arg\max_ky_k(x)$.

**Fisher linear discriminant**: $\red{y(x_i)=w^⊤x_i+w_0}$ projects points onto axis $w$ against a threshold placed at $-w_0$.

- **Scatter matrices**: Covariance matrix is a sum of within-class and between-class scatters. Let $X∈ℝ^{N×D}$
$\BA
\dg{⇒}&\V[X]=(X-\v{1}\bar{x}^⊤)^⊤(X-\v{1}\bar{x}^⊤)=\sum_i(x_i-\bar{x})(x_i-\bar{x})^⊤=\sum_k\sum_{i∈C_k}(x_i-\bar{x})(x_i-\bar{x})^⊤\\
&=\sum_k\sum_{i∈C_k}(x_i-\bar{x}_k+\bar{x}_k-\bar{x})(x_i-\bar{x}_k+\bar{x}_k-\bar{x})^⊤\quad\dg{⇓\sum_{i∈C_k}(x_i-\bar{x}_k)=0}\\
&=\sum_k\sum_{i∈C_k}[(x_i-\bar{x}_k)(x_i-\bar{x}_k)^⊤+(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^⊤+(\bar{x}_k-\bar{x})(x_i-\bar{x}_k)^⊤+(x_i-\bar{x}_k)(\bar{x}_k-\bar{x})^⊤]\\
&=\sum_k\sum_{i∈C_k}[(x_i-\bar{x}_k)(x_i-\bar{x}_k)^⊤+(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^⊤]=\green{\sum_k\sum_{i∈C_k}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^⊤}+\yello{\sum_kN_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^⊤}\\
\dg{⇒}&\V[X]=\green{S_W}+\yello{S_B}\dg{⇒}\BC
\green{S_W=\sum_k\sum_{i∈C_k}(x_i-\bar{x})(x_i-\bar{x})^⊤}=\sum_k(N_k-1)\hat{Σ}_k\dg{ - \t{within-class scatter}}\\
\yello{S_B=\sum_kN_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^⊤}\dg{ - \t{between-class scatter}}
\EC
\EA$

- **Binary scatter**: Let $k=\{1,2\}$ and $\bar{x}_k=\/{1}{N_k}\sum_{i∈C_k}x_i$. Let $y_i=w^⊤x_i$ be data points projected onto axis $w$.
$\BA
\dg{⇒}&\dg{\t{between-class }}(\bar{y}_1-\bar{y}_2)^2=[w^⊤(\bar{x}_1-\bar{x}_2)]^2=w^⊤[\yello{(\bar{x}_1-\bar{x}_2)(\bar{x}_1-\bar{x}_2)^⊤}]w=w^⊤\yello{S_{12}}w\dg{ - \t{$S_{12}$ and $S_B$ are related as below}}\\
&\dg{⇒}\bar{x}=\/{N_1}{N}\bar{x}_1+\/{N_2}{N}\bar{x}_2
\dg{⇒}\bar{x}_1-\bar{x}=\bar{x}_1-(\/{N_1}{N}\bar{x}_1+\/{N_2}{N}\bar{x}_2)\dg{⇒}\brown{\bar{x}_1-\bar{x}}=\/{N_2}{N}(\bar{x}_1-\bar{x}_2)\dg{⇒}\grass{\bar{x}_2-\bar{x}}=-\/{N_1}{N}(\bar{x}_1-\bar{x}_2)\\
&\dg{⇒}\yello{S_B}=N_1(\brown{\bar{x}_1-\bar{x}})(\brown{\bar{x}_1-\bar{x}})^⊤+N_2(\grass{\bar{x}_2-\bar{x}})(\grass{\bar{x}_2-\bar{x}})^⊤=(N_1\/{N_2^2}{N^2}+N_2\/{N_1^2}{N^2})(\bar{x}_1-\bar{x}_2)(\bar{x}_1-\bar{x}_2)^⊤\dg{⇒}\yello{S_B=\/{N_1N_2}{N}S_{12}}\\
\dg{⇒}&\dg{\t{within-class }}s_k^2=\sum_{i∈C_k}(y_i-\bar{y}_k)^2=\sum_{i∈C_k}[w^⊤(x_i-\bar{x}_k)]^2=w^⊤[\green{\sum_{i∈C_k}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^⊤}]w=w^⊤\green{S_{k}}w\\
\dg{⇒}&\V[X]=\green{S_1+S_2}+\yello{\/{N_1N_2}{N}S_{12}},\ \green{S_W=S_1+S_2},\ \yello{S_B=\/{N_1N_2}{N}S_{12}}
\EA$

- **Fisher criterion** is $\red{J(w):=\/{w^⊤\yello{S_B}w}{w^⊤\green{S_W}w}}$ is the ratio of projected scatter along axis $w$.
We pick axis $w^*=\arg\sup_wJ(w)$ that maximizes $\/{(\bar{y}_2-\bar{y}_1)^2}{s_1^2+s_2^2}=\/{w^⊤\yello{S_{12}}w}{w^⊤[\green{S_1+S_2}]w}$.
Note that this does not pick an intercept for the decision boundary.
$\BA
\dg{⇒}&λ_\min(\green{S_W}^{-1}\yello{S_B})≤\red{\/{w^⊤\yello{S_B}w}{w^⊤\green{S_W}w}}≤λ_\max(\green{S_W}^{-1}\yello{S_B})\dg{ - \t{generalized Rayleigh quotient}}\\
\dg{⇒}&J(w)=(w^⊤\yello{S_B}w)(w^⊤\green{S_W}w)^{-1}\dg{⇒}∇_wJ(w)=[∇_w(w^⊤\yello{S_B}w)](w^⊤\green{S_W}w)^{-1}+(w^⊤\yello{S_B}w)[∇_w(w^⊤\green{S_W}w)^{-1}]\\
&=2w^⊤\yello{S_B}(w^⊤\green{S_W}w)^{-1}-(w^⊤\yello{S_B}w)(w^⊤\green{S_W}w)^{-2}2w^⊤\green{S_W}=\/{2w^⊤\yello{S_B}(w^⊤\green{S_W}w)-2w^⊤\green{S_W}(w^⊤\yello{S_B}w)}{(w^⊤\green{S_W}w)^2}=0\\
&\dg{⇒}w^⊤\yello{S_B}(w^⊤\green{S_W}w)-w^⊤\green{S_W}(w^⊤\yello{S_B}w)=0\dg{⇒}(w^⊤\green{S_W}w)\yello{S_B}w=(w^⊤\yello{S_B}w)\green{S_W}w\dg{⇒}\green{S_W}^{-1}\yello{S_B}w=\red{\/{w^⊤\yello{S_B}w}{w^⊤\green{S_W}w}}w\\
&\dg{⇒}\green{S_W}^{-1}\yello{S_B}w=\green{S_W}^{-1}(\bar{x}_2-\bar{x}_1)[(\bar{x}_2-\bar{x}_1)^⊤w]\dg{⇒}\red{w^*∝\green{S_W}^{-1}(\bar{x}_2-\bar{x}_1)}\\
&\dg{⇒\t{$w$ is eigenvector of $\green{S_W}^{-1}\yello{S_B}$ corresponding to $λ=\sup_w\/{w^⊤\yello{S_B}w}{w^⊤\green{S_W}w}$}}
\EA$

- Modify OLS for binary classification with $t_i=\/{N}{N_1}\ ∀x_i∈C_1$ and $t_i=-\/{N}{N_2}\ ∀x_i∈C_2$ where $X∈ℝ^{N×D}$ shows that binary Fisher discriminant is identical to OLS linear discriminant
$\BA
\dg{⇒}&\v{1}^⊤t=0\dg{⇒}w_0=-w^⊤\bar{x}\\
\dg{⇒}&∇_w\n{Xw+w_0\v{1}-t}^2=2(Xw+w_0\v{1}-t)^⊤X=0\dg{⇒}X^⊤(Xw-\v{1}\bar{x}^⊤w-t)=0\dg{⇒}X^⊤(X-\v{1}\bar{x}^⊤)w=X^⊤t\\
&\t{RHS: }X^⊤t=\sum_ix_i^⊤t_i=\/{N}{N_1}\sum_{i∈C_1}x_i-\/{N}{N_2}\sum_{i∈C_2}x_i=N(\bar{x}_1-\bar{x}_2)\\
&\t{LHS: }(\v{1}\bar{x}^⊤)^⊤(X-\v{1}\bar{x}^⊤)=\bar{x}\v{1}^⊤X-\bar{x}\v{1}^⊤\v{1}\bar{x}^⊤=\bar{x}N\bar{x}^⊤-\bar{x}N\bar{x}^⊤=0\dg{⇒}X^⊤(X-\v{1}\bar{x}^⊤)=(X-\v{1}\bar{x}^⊤)^⊤(X-\v{1}\bar{x})\\
\dg{⇒}&(\green{S_W}+\yello{S_B})w=N(\bar{x}_1-\bar{x}_2)\dg{⇒}\red{w^*∝\green{S_W}^{-1}(\bar{x}_2-\bar{x}_1)}
\EA$

**Multiclass Fisher**: For $K>2$, we have $y(x_i)=W^⊤x_i∈ℝ^{D'}$ where $W∈ℝ^{D×D'}$ and $1< D'≤K-1$.
$\BA
\dg{⇒}&\BC
\tilde{S}_W=\sum_k\sum_{i∈C_k}(y_i-\bar{y}_k)(y_i-\bar{y}_k)^⊤=\sum_k\sum_{i∈C_k}W^⊤(x_i-\bar{x}_k)(x_i-\bar{x}_k)^⊤W=W^⊤\yello{S_W}W\\
\tilde{S}_B=\sum_kN_k(\bar{y}_k-\bar{y})(\bar{y}_k-\bar{y})^⊤=\sum_kN_kW^⊤(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^⊤W=W^⊤\green{S_B}W\EC\\
\dg{⇒}&\red{J(W):=\tr[\tilde{S}_W^{-1}\tilde{S}_B]=\tr[(W^⊤\yello{S_W}W)^{-1}(W^⊤\green{S_B}W)]}\dg{ - \t{$\tilde{S}_W$ and $\tilde{S}_B$ are projected scatter}}\\
&\dg{⇒}∇_WJ(W)=2\green{S_B}W(w^⊤\yello{S_W}W)^{-1}-2\yello{S_W}W(W^⊤\yello{S_W}W)^{-1}(W^⊤\green{S_B}W)(W^⊤\yello{S_W}W)^{-1}=0\\
&\dg{⇒}\green{S_B}W-\yello{S_W}W(W^⊤\yello{S_W}W)^{-1}(W^⊤\green{S_B}W)=0\dg{⇒}\red{\yello{S_W}^{-1}\green{S_B}W=WΛ}\\
&\dg{⇒\t{$W$ are eigenvectors of $\yello{S_W}^{-1}\green{S_B}$ wrt eigenvalues $Λ=\sup_W(W^⊤\yello{S_W}W)^{-1}(W^⊤\green{S_B}W)$}}\\
\EA$

- **Dimensionality/Rank**: $K$ points can only make up to $K-1$ linearly independent vectors.
$K$ centroids can only span up to $K-1$ dimensions.
When $D≫K$ the additional orthogonal dimensions provide zero discriminitive value.
$S_B=\sum_kN_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^⊤∈ℝ^{D×D}$ is a sum of $K$ rank-1 matrices.
$\rk(S_B)=K$ but $\bar{x}$ used 1 degree of freedom. Only $K-1$ of those matrices are linearly independent.
$S_W^{-1}S_B$ has at most $K-1$ non-zero eigenvalues.
Therefore $1< D'≤K-1$.

- **Whitening and rank reduction**: Conceptually $\red{y_i=W^⊤x_i=V^⊤S_W^{-1/2}x_i}$.
$\BA
\dg{⇒}&\red{W=S_W^{-1/2}V}\dg{⇒}\brown{S_W^{-1}S_B}W=WΛ\dg{⇒}S_W^{-1}S_B(S_W^{-1/2}V)=(S_W^{-1/2}V)Λ\dg{⇒}\brown{S_W^{-1/2}S_BS_W^{-1/2}}V=VΛ\dg{⇒}\tilde{S}_W^{-1}\tilde{S}_BV=VΛ\\
&\BC
\tilde{S}_W=S_W^{-1/2}S_WS_W^{-1/2}=I\dg{ - \t{$\tilde{S}_W$ is whitened from an ellipsoid to unit-norm sphere}}\\
\tilde{S}_B=S_W^{-1/2}S_BS_W^{-1/2}\dg{ - \t{$V$ is orthogonal matrix containing eigenvectors of whitened $\tilde{S}_B$}}\\
\tilde{\V}[X]v=(I+\tilde{S}_B)v=(1+λ)v\dg{ - \t{whitened $\tilde{\V}[X]$ and $\tilde{S}_B$ share the same eigenvectors in $V$}}
\EC\\
\dg{⇒}&\tilde{x}_i=S_W^{-1/2}x_i\dg{⇒}\tilde{S}_w=\sum_k\sum_{i∈C_k}(\tilde{x}_i-\tilde{\bar{x}}_k)(\tilde{x}_i-\tilde{\bar{x}})^⊤=\sum_k\sum_{i∈C_k}S_W^{-1/2}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^⊤S_W^{-1/2}\\
&=S_W^{-1/2}S_WS_W^{-1/2}=I\dg{ - \t{if $\V[x_i|C_k]=S_W$ then $\tilde{x}_i=S_W^{-1}x_i$ variance $\V[\tilde{x}_i|C_k]=I$ is a unit-norm sphere}}\\
&\dg{⇒\t{ OR }}\V[x_i|C_k]=S_W\dg{⇒}\V[\tilde{x}_i|C_k]=\V[S_W^{-1/2}x_i|C_k]=S_W^{-1/2}\V[x_i|C_k]S_W^{-1/2⊤}=S_W^{-1/2}S_WS_W^{-1/2}=I
\EA$

  - **Whitening / Sphering**: We start with $K$ classes of points in standard space, each grouped in ellipsoid cluster whose shape is defined by their within-class covariance.
  $S_W$ is the summation (therefore average) of $K$ cluster shapes.
  $\tilde{x}_i=S_W^{-1/2}x_i$ brings all points to a new basis where $\tilde{S}_W=I$ is a unit-variance sphere instead of an ellipsoid.
  Each transformed cluster on average looks like a unit-norm ball (**centroid**).

  - **Rank reduction**: PCA code $z_i=V_M^⊤x_i$ is coordinates of $x_i$ along the first $M$ eigenvectors of $\V[X]$ representing $M$ highest variance directions in $X$.
  The $D'$ eigenvectors of $\tilde{S}_B$ contained in $V∈ℝ^{D×D'}$ are also eigenvectors of $\tilde{\V}[X]$, which represent most spread out directions in the whitened data, or the most spread out directions between unit-norm centroids of whitened data.

- **Nearest centroid classification**: Centroid $k$ has mean $\bar{y}_k=W^⊤\bar{x}_k=V^⊤S_W^{-1/2}\bar{x}_k$. Then $x_i$ is assigned to class $\arg\min_k\n{y_i-\bar{y}_k}^2=\n{V^⊤S_W^{-1/2}(x_i-\bar{x}_k)}^2$.

  - Suppose $P(x_i|C_k)=\Normal(\bar{x}_k,Σ_k)$, then $S_W=\sum_k(N_k-1)\hat{Σ}_k$ and $\hat{Σ}_k=\/{1}{N_k-1}\sum_{i∈C_k}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^⊤$ is unbiased estimator for $Σ_k$.
  Euclidean distance $\n{y_i-\bar{y}_k}^2$ in whitened space is equivalent to Malahanobis distance $(x_i-\bar{x}_k)^⊤Σ_k^{-1}(x_i-\bar{x}_k)$ in standard space.

**Perceptron discriminant**: binary linear discriminant $y(x_i)=\t{sign}(w^⊤ϕ(x_i))$ with activation function $\t{sign}(⋅)$.
We seek $w$ predicting $\BC
w^⊤ϕ(x_i)> 0\dg{⇒}i∈C_1\\
w^⊤ϕ(x_i)< 0\dg{⇒}i∈C_2
\EC$ against $t_i∈\{+1_{C_1},-1_{C_2}\}$
with cost $\BC
-w^⊤ϕ(x_i)t_i> 0&\t{if misclassified }i∈\m{M}\\
-w^⊤ϕ(x_i)t_i=0&\t{if correctly classified}
\EC$.

- **Cost** $E_P(w)=-\sum_{i∈\m{M}}w^⊤ϕ(x_i)t_i=\sum_{i=1}^N\max(0,-w^⊤ϕ(x_i)t_i)$ is piecewise linear sum of $N$ hockey sticks.
SGD $w^{(τ+1)}=w^{(τ)}-η∇E_P(w)$ uses $∇E_P(w)=-ϕ(x_i)t_i$ for random $i∈\m{M}$.
Then $\red{w^{(τ+1)}=w^{(τ)}+ηϕ(x_i)t_i}$.

  - Because magnitude is ignored by the activation function and $w^{(0)}=0$, we can assume $η=1$ without loss of generality as long as it stays constant.

  - Naive solution to the problem by simply counting $|\m{M}|$ creates a piecewise constant error, which does not have gradients.
  Perceptron algorithm has piecewise linear error, which allows gradient descent.
  The non-differentiable joints between pieces are not a problem for SGD because $∇E_P(w)=-ϕ(x_i)t_i$ is always defined.

- **Heart of algorithm**: $w:=w+ϕ(x_i)$ for misclassified $t_i=+1$ to align $w$ closer to $ϕ(x_i)$ and $w^⊤ϕ(x_i)→+1$.
Update $w:=w-ϕ(x_i)$ for misclassified $t_i=-1$ to align $w$ farther from $ϕ(x_i)$ and $w^⊤ϕ(x_i)→-1$.

- **Convergence**: If the data points are linearly separable into two classes neatly by a separating hyperplane, then the algorithm is guaranteed to find an exact solution in finite steps.

  - If the data points are not linearly separable then the algorithm will never converge.

  - The algorithm does not generalize to $K>2$.

**Generative approach** models the underlying distributions $p(x_i|C_k)$ and $p(C_k)$, then calculate $p(C_k|x_i)$ the probability that input $x∈ℝ^D$ belongs to a class $C_1,...,C_K$.
"Generative" refers to the ability to generate synethetic $x_i$ from model.
**Discriminative approach** focuses on the decision boundary using MLE parameters defined through $p(C_k|x_i)$.

- **Logistic and logit**: $p(C_1|x_i)=\/{p(x_i|C_1)p(C_1)}{p(x_i|C_1)p(C_1)+p(x_i|C_2)p(C_2)}=\/{1}{1+\e{-a}}=σ(a)$ is logistic sigmoid function
where $\red{a=\ln\/{p(x_i|C_1)p(C_1)}{p(x_i|C_2)p(C_2)}}$ is logit (log-odds) function.
$\red{σ(a)=\/{1}{1+\e{-a}}}=\/{1}{1+\/{p(x_i|C_2)p(C_2)}{p(x_i|C_1)p(C_1)}}=\/{p(x_i|C_1)p(C_1)}{p(x_i|C_1)p(C_1)+p(x_i|C_2)p(C_2)}$.

  - Logistic $p(C_k|x_i)=σ(a)$ and logit $a=\ln\/{σ}{1-σ}$ only works for $K=2$. It is one-vs-rest in $K>2$ and cannot guarantee $\sum_kp(C_k|x_i)=1$ because it creates ambiguities in the input space.
  Let $\red{a_k=\ln p(x|C_k)p(C_k)}$ then
  $\BA
  \dg{⇒}&a=\ln\/{p(x_i|C_1)p(C_1)}{p(x_i|C_2)p(C_2)}=\ln\/{p(C_1|x_i)}{p(C_2|x_i)}\dg{\t{ log-odds in posterior probabilities = log-odds in joint probabilities}}\\
  \dg{⇒}&a=\ln p(x_i|C_1)p(C_1)-\ln p(x_i|C_2)p(C_2)=a_1-a_2\dg{\t{ is the difference in log joint probabilities}}\\
  \dg{⇒}&σ=\/{1}{1+\e{-a}}\dg{⇒}1=σ+σ\e{-a}\dg{⇒}\e{-a}=\/{1-σ}{σ}\dg{⇒}a=\ln\/{σ}{1-σ}\dg{\t{ log-odds link function}}\\
  \dg{⇒}&σ(a)=\/{1}{1+\e{-a}}=\/{\e{a}}{1+\e{a}}\dg{\t{ logistic equivalent form}}\\
  \dg{⇒}&σ'(a)=\/{\e{-a}}{(1+\e{-a})^2}=σ(a)(1-σ(a))\dg{\t{ derivative is the cheap-to-compute standard logistic distribution}}\\
  \dg{⇒}&σ(a)=\/{\e{a_1-a_2}}{1+\e{a_1-a_2}}=\/{\e{a_1}}{\e{a_1}+\e{a_2}}\dg{\t{ logistic is binary softmax of log joint probabilities}}\\
  \dg{⇒}&σ(-a)=1-σ(a)\dg{\t{ logistic is symmetric}}\\
  \EA$

  - $K>2$: $\BA \red{p(C_k|x_i)=\/{p(x_i|C_k)p(C_k)}{\sum_jp(x_i|C_j)p(C_j)}=\/{\e{a_k}}{\sum_j\e{a_j}}}\EA$ is softmax (normalized exponential) function.

- **Shared covariance**: If we assume $Σ$ is shared among all $p(x_i|C_k)$ we arrive at a linear discriminant model.
$\BA
\dg{⇒}&p(x_i|C_k)=\Normal(x_i|μ_k,Σ)=\/{1}{(2π)^{D/2}\det(Σ)^{1/2}}\e{-\/{1}{2}(x_i-μ_k)^⊤Σ^{-1}(x_i-μ_k)}\\
\dg{⇒}&a(x_i)=\ln p(x_i|C_1)-\ln p(x_i|C_2)+\ln\/{p(C_1)}{p(C_2)}
=-\/{1}{2}(x_i-μ_1)^⊤Σ^{-1}(x_i-μ_1)+\/{1}{2}(x_i-μ_2)^⊤Σ^{-1}(x_i-μ_2)+\ln\/{p(C_1)}{p(C_2)}\\
&=\green{(μ_1-μ_2)^⊤Σ^{-1}}x_i\yello{-\/{1}{2}[μ_1^⊤Σ^{-1}μ_1-μ_2^⊤Σ^{-1}μ_2]+\ln\/{p(C_1)}{p(C_2)}}\dg{⇒}\red{a(x_i)=\green{w^⊤}x_i+\yello{w_0}\dg{⇒}p(C_1|x_i)=σ(\green{w^⊤}x_i+\yello{w_0})}
\EA$
$\BA
\dg{⇒}&a_k(x_i)=\purple{c_1}-\/{1}{2}(x_i-μ_k)^⊤Σ^{-1}(x_i-μ_k)+\ln p(C_k)=\purple{c_1-\/{1}{2}x_i^⊤Σ^{-1}x_i}-\/{1}{2}(-2μ_k^⊤Σ^{-1}x_i+μ_k^⊤Σ^{-1}μ_k)+\ln p(C_k)\\
&=\purple{c_1-\/{1}{2}x_i^⊤Σ^{-1}x_i}+\green{μ_k^⊤Σ^{-1}}x_i\yello{-\/{1}{2}μ_k^⊤Σ^{-1}μ_k+\ln p(C_k)}\dg{⇒}\red{a_k(x_i)=\green{w_k^⊤}x_i+\yello{w_{k0}}\dg{⇒}p(C_k|x_i)=\/{\e{\green{w_k^⊤}x_i+\yello{w_{k0}}}}{\sum_j\e{\green{w_j^⊤}x_i+\yello{w_{j0}}}}}
\EA$

  - $\purple{c_1-\/{1}{2}x_i^⊤Σ^{-1}x_i}$ gets canceled out against denominator, and we arrive at K-class linear discriminant decision regions $R_k=\{x|y_k(x)≥y_j(x)\ ∀j\}$ with logistic sigmoid ($K=2$) or softmax ($K>2$) activation functions.

  - $\BA
  p(C_k|x_i)=p(C_j|x_i)\dg{⇒}a_k(x_i)=a_j(x_i)\dg{⇒}(w_k-w_j)^⊤x_i+(w_{k0}-w_{j0})=0
  \EA$
  Linearity of the discriminant derives from linearity of $a_k$.
  Relaxing shared covariance condition gives rise to quadratic $a_k$ and quadratic discriminant.

  - To use $p(C_k|x_i)$ we need to find $μ_k$ and $Σ$ from training data through likelihood function $p(t)$.
  In practice we skip all these steps and go straight to linear discriminant model: logistic regression.

- **MLE**: we find $μ_k$ and $Σ$ from likelihood $\blue{p(t)=\prod_ip(x_i,C_1)^{t_i}p(x_i,C_2)^{1-t_i}}$ where $x_i∈ℝ^D,t_i∈\{1_{C_1},0_{C_2}\}$.
Then $\BA
\dg{⇒}&\BC
p(C_1)=π\dg{⇒}p(x_i,C_1)=π\Normal(x_i|μ_1,Σ)\\
p(C_2)=1-π\dg{⇒}p(x_i,C_2)=(1-π)\Normal(x_i|μ_2,Σ)
\EC
\dg{⇒}p(t|π,μ_1,μ_2,Σ)=\prod_i(π\Normal(x_i|μ_1,Σ))^{t_i}((1-π)\Normal(x_i|μ_2,Σ))^{1-t_i}\\
\dg{⇒}&\ln p(t|π)=\sum_i[t_i\lnπ+(1-t_i)\ln(1-π)]+c_1\dg{⇒}∂_π=\/{\sum_it_i}{π}-\/{N-\sum_it_i}{1-π}=0\dg{⇒}N_1-πN_1=πN-πN_1\dg{⇒}\blue{π_\t{ML}=\/{N_1}{N}}\\
\dg{⇒}&\ln p(t|μ_1)=\sum_i-\/{1}{2}t_i(x_i-μ_1)^⊤Σ^{-1}(x_i-μ_1)+c_2\dg{⇒}∂_{μ_1}=\sum_it_i(x_i-μ_1)^⊤Σ^{-1}=0\dg{⇒}\sum_it_ix_i=\sum_it_iμ_1\\
&\dg{⇒}\blue{μ_{1\t{ML}}=\/{\sum_it_ix_i}{N_1}}\quad\dg{⇒}\sum_i(1-t_i)x_i=\sum_i(1-t_i)μ_2\dg{⇒}\blue{μ_{2\t{ML}}=\/{\sum_i(1-t_i)x_i}{N_2}}\\
\dg{⇒}&\ln p(t|Σ)=\sum_it_i[-\/{1}{2}\ln\det(Σ)-\/{1}{2}(x_i-μ_1)^⊤Σ^{-1}(x_i-μ_1)]+\sum_i(1-t_i)[-\/{1}{2}\ln\det(Σ)-\/{1}{2}(x_i-μ_2)^⊤Σ^{-1}(x_i-μ_2)]\\
&=-\/{N}{2}\ln\det(Σ)-\sum_i\/{t_i}{2}\tr[Σ^{-1}(x_i-μ_1)(x_i-μ_1)^⊤]-\sum_i\/{1-t_i}{2}\tr[Σ^{-1}(x_i-μ_2)(x_i-μ_2)^⊤]\\
&=-\/{N}{2}\ln\det(Σ)-\/{1}{2}\tr[Σ^{-1}(\sum_it_i(x_i-μ_1)(x_i-μ_1)^⊤+\sum_i(1-t_i)(x_i-μ_2)(x_i-μ_2)^⊤)]\\
&=-\/{N}{2}\ln\det(Σ)-\/{1}{2}\tr[Σ^{-1}(N_1Σ_1+N_2Σ_2)]\dg{⇒}∂_Σ=-\/{N}{2}Σ^{-1}+\/{1}{2}(N_1Σ_1+N_2Σ_2)Σ^{-2}=0\dg{⇒}\blue{Σ_\t{ML}=\/{N_1}{N}Σ_1+\/{N_2}{N}Σ_2}\\
&\dg{⇒}Σ_1=\/{1}{N_1}\sum_it_i(x_i-μ_1)(x_i-μ_1)^⊤,\ Σ_2=\/{1}{N_2}\sum_i(1-t_i)(x_i-μ_2)(x_i-μ_2)^⊤\\
\dg{⇒}&\dg{\t{Unbiased: }}\hat{Σ}_1=\/{1}{N_1-1}\sum_it_i(x_i-μ_1)(x_i-μ_1)^⊤,\ \hat{Σ}_2=\/{1}{N_2-1}\sum_i(1-t_i)(x_i-μ_2)(x_i-μ_2)^⊤\dg{⇒}\hat{Σ}=\/{(N_1-1)\hat{Σ}_1+(N_2-1)\hat{Σ}_2}{N-2}\\
\EA$

  - The results extend to $K>2$.
  From training data we calculate $μ_k,Σ$, and build a generative model for $p(x_i|C_k)$ which derives $p(C_k|x_i)$ to classify data.
  MLE parameters are not robust and are vulnerable to outliers.

- **Discrete features**: Let $x_i∈\{0,1\}^D$ and $μ_k∈ℝ^D$, then likelihood $\blue{p(x_i|C_k)=\prod_{j=1}^Dμ_{kj}^{x_{ij}}(1-μ_{kj})^{1-x_{ij}}}$
$\BA
\dg{⇒}&a_k(x_i)=\ln p(x_i|C_k)+\ln p(C_k)=\sum_j[x_{ij}\ln μ_{kj}+(1-x_{ij})\ln (1-μ_{kj})]+\ln p(C_k)\\
&=\sum_j\ln\/{μ_{kj}}{1-μ_{kj}}x_{ij}+\sum_j\ln(1-μ_{kj})+\ln p(C_k)=\green{\left[\ln\/{μ_{kj}}{1-μ_{kj}}\right]_{j=1..D}^⊤}x_i+\yello{\sum_j\ln(1-μ_{kj})+\ln p(C_k)}\dg{⇒}a_k=\green{w_k^⊤}x_i+\yello{w_{k0}}
\EA$

  - Cross entropy: true distribution $p=[x_{ij},1-x_{ij}]$ and predicted distribution $q=[μ_{kj},1-μ_{kj}]$.
  Then
  $\BA
  -\H_j[p,q]=\sum_lp_l\ln q_l=x_{ij}\ln μ_{kj}+(1-x_{ij})\ln(1-μ_{kj})
  \EA$

- **Exponential family**: Let $\blue{p(x_i|w_k,s)=\/{1}{s}h(\/{1}{s}x_i)g(w_k)\e{\/{1}{s}w_k^⊤x_i}}$ where classes share the same spread/width through $s$.
Normal distribution has $w_k=Σ^{-1}μ_k$ and $s=1$.
Then $\blue{\BC
a(x_i)=(w_1-w_2)^⊤x_i+\ln\/{g(w_1)}{g(w_2)}+\ln\/{p(C_1)}{p(C_2)}\\
a_k(x_i)=w_k^⊤x_i+\ln g(w_k)+\ln p(C_k)
\EC}$

**Logistic regression**: Given binary $Φ∈ℝ^{N×(M+1)}$ and $t∈\{1_{C_1},0_{C_2}\}^N$, where $ϕ_i$ is the $i$-th row and $ϕ_{i0}=1$.
Logistic regression performs linear regression on alternative target space to find $w$ in activation function $p(C_1|ϕ_i)=σ(w^⊤ϕ_i)$.
$\BA
\dg{⇒}&p(C_1|ϕ_i)=σ(a_i)=y(ϕ_i)=\red{y_i=\/{1}{1+\e{-a_i}}},\ \red{a_i(y_i)=w^⊤ϕ_i}\dg{⇒}y_i'(a_i)=\/{e^{-a_i}}{(1+e^{-a_i})^2}=y_i(1-y_i)\dg{⇒}a_i'(y_i)=\/{1}{y_i(1-y_i)}
\EA$

- **Effective target**: We cannot regress $w^⊤ϕ_i∈(-∞,∞)$ against binary target $t_i∈\{0,1\}$.
We use logit **link function** (inverse activation function) $a_i(y_i)=σ^{-1}(y_i)=\ln\/{y_i}{1-y_i}∈(-∞,∞)$ instead.
$a_i(t_i)=\ln\/{t_i}{1-t_i}=\{-∞,∞\}$ is undefined, but its Taylor's approximation in the neighborhood of $y_i$ is a valid target.
$\BA
\dg{⇒}&a_i(y)=a_i(y_i)+a_i'(y_i)(y-y_i)=w^⊤ϕ_i+\/{y-y_i}{y_i(1-y_i)}\dg{⇒}\red{z_i=a_i(t_i)=w^⊤ϕ_i+\/{t_i-y_i}{y_i(1-y_i)}}\dg{\t{ 1st-order Taylor expansion}}\\
\dg{⇒}&\V[z_i]=\/{\V[t_i]}{y_i^2(1-y_i)^2}\dg{⇒}\red{\V[z_i]=\/{1}{y_i(1-y_i)}}
\EA$

  - $p(t_i)=\t{Bernoulli}(t_i|y_i)\dg{⇒}\V[t_i]=y_i(1-y_i)$ is zero at $y_i=0,y_i=1$ and $\/{1}{4}$ at $y_i=\/{1}{2}$ when uncertainty is high.
  $\/{t_i-y_i}{y_i(1-y_i)}$ is positive when $t_i=1$ and negative when $t_i=0$.
  It is amplified by $\V[t_i]^{-1}$.

- **Weighted least squares (WLS)**: OLS assumes homoscedasticity $p(t_i)=\Normal(t_i|w^⊤ϕ_i,\yello{σ^2})\dg{⇒}Σ=σ^2I$. WLS assumes $p(t_i)=\Normal(t_i|w^⊤ϕ_i,\yello{σ_i^2})\dg{⇒}\V[t]=Σ=\diag[σ_i^2]$, and $\red{R=\V[t]^{-1}}$ is a precision matrix that weighs each point based on the precision of the target.
$\BA
\dg{⇒}&E(w)=\/{1}{2}(t-Φw)^⊤Σ^{-1}(t-Φw)\dg{⇒}∇_wE(w)=(t-Φw)^⊤Σ^{-1}(-Φ)=0\\
&\dg{⇒}Φ^⊤Σ^{-1}Φw=Φ^⊤t\dg{⇒}w=(Φ^⊤Σ^{-1}Φ)^{-1}Φ^⊤Σ^{-1}t=(Φ^⊤RΦ)^{-1}Φ^⊤Rt\dg{ - \t{sphering $(\tilde{Φ}^⊤\tilde{Φ})^⊤\tilde{Φ}^⊤\tilde{t}$ where $\tilde{Φ}=Σ^{-1/2}Φ$}}
\EA$

- **Logistic WLS**: We regress $a_i(y_i)=w^⊤ϕ_i$ against Taylor approximation of $a_i(t_i)$ in the neighborhood of $y_i$.
However $\V[z_i]=\/{1}{y_i(1-y_i)}$ varies widely across data points so we have to use WLS.
$\BA
\dg{⇒}&\red{R≜\diag[y_i(1-y_i)]}\dg{⇒}\purple{z=Φw+R^{-1}(t-y)}\dg{⇒}\V[z]=\V[Φw+R^{-1}(t-y)]=R^{-1}\V[t]R^{-1}=R^{-1}RR^{-1}=R^{-1}\\
&\dg{⇒}\blue{w=(Φ^⊤RΦ)^{-1}Φ^⊤R\purple{z}}\dg{ - \t{both $R$ and $z$ depend on $w$, so it is not a closed form solution}}\\
\dg{⇒}&E(w)=\/{1}{2}(z-Φw)^⊤R(z-Φw)=\/{1}{2}(R^{-1}(t-y))^⊤R(R^{-1}(t-y))=\/{1}{2}(t-y)^⊤R^{-1}(t-y)\\
&\dg{⇒}∇_wE(w)=(t-y)^⊤R^{-1}∇_wy=(t-y)^⊤R^{-1}RΦ=\green{(t-y)^⊤Φ=0}\\
&\dg{⇒}\dg{\t{ this is where nonlinearity of $y$ prevents closed-form solution}}
\EA$

  - $R=\V[z]^{-1}=\V[t]$.
  Logistic WLS weighs each point by precision of the effective target $z_i$ but also by variance of the real target $\V[t_i]$.
  Points with higher $\V[t_i]=y_i(1-y_i)$ contribute more to the determination of the decision boundary.
  These points are located near the decision boundary where $y_i$ is closer to 0.5.
  Outliers have $y_i$ closer to 0 or 1 and therefore very low $\V[t_i]$.
  This makes logistic regression very robust to outliers.

**Iterative reweighted least squares (IRLS)**: logistic regression has no closed-form solution.
IRLS is an application of Newton's method $\red{w^\t{(new)}=w^\t{(old)}-H^{-1}∇E(w^\t{(old)})}$ involving weighing matrix $R$.

- B&V: $w^+=w+tΔw_\t{nt}$. Newton's step $Δw_\t{nt}=-∇^2E(w)^{-1}∇E(w)$. Step size $t=1$ with no line search.

- **IRLS for OLS**: Newton's method finds endpoint $w$ after 1 iteration.
$\BA
\dg{⇒}&E(w)=\/{1}{2}\n{t-Φw}^2\dg{⇒}∇_wE(w)=[(t-Φw)^⊤(-Φ)]^⊤=Φ^⊤Φw-Φ^⊤t\dg{⇒}H=Φ^⊤Φ\\
&\dg{⇒}w^\t{(new)}=w^\t{(old)}-H^{-1}∇E(w^\t{(old)})=w^\t{(old)}-w^\t{(old)}+(Φ^⊤Φ)^{-1}Φ^⊤t\dg{⇒}w^\t{(new)}=(Φ^⊤Φ)^{-1}Φ^⊤t
  \EA$

- **IRLS for logistic regression**: $R$ depends on $w^\t{(old)}$, which is iteratively updated (reweighted).
$\BA
\dg{⇒}&p(C_1|ϕ_i)=σ(w^⊤ϕ_i)=y(ϕ_i)=\red{y_i=\/{1}{1+\e{-w^⊤ϕ_i}}}\dg{⇒}∇_wy_i=\/{\e{-w^⊤ϕ_i}ϕ_i}{(1+\e{-w^⊤ϕ_i})^2}=y_i(1-y_i)ϕ_i\\
\dg{⇒}&p(t|w)=\prod_iy_i^{t_i}(1-y_i)^{1-t_i}\dg{⇒}E(w)=-\sum_i[t_i\ln y_i+(1-t_i)\ln(1-y_i)]\dg{\t{ binomial cross entropy MLE error}}\\
&\dg{⇒}∇_wE(w)=-\sum_i(\/{t_i}{y_i}-\/{1-t_i}{1-y_i})∇_wy_i=-\sum_i\/{t_i-y_i}{y_i(1-y_i)}y_i(1-y_i)ϕ_i\green{\dg{⇒}∇_wE(w)=\sum_i(y_i-t_i)ϕ_i=Φ^⊤(y-t)}\\
&\dg{⇒}H=∇_w^2E(w)=∇_w(\sum_iy_iϕ_i)=\sum_iy_i(1-y_i)ϕ_iϕ_i^⊤=Φ^⊤\diag[y_i(1-y_i)]Φ=Φ^⊤RΦ\\
\dg{⇒}&w^\t{(new)}=w^\t{(old)}-H^{-1}∇E(w^\t{(old)})
=(Φ^⊤RΦ)^{-1}(Φ^⊤RΦ)w^\t{(old)}-(Φ^⊤RΦ)^{-1}Φ^⊤(y-t)\\
&\dg{⇒}\red{w^\t{(new)}=(Φ^⊤RΦ)^{-1}Φ^⊤R(\purple{Φw^\t{(old)}-R^{-1}(y-t)})}=\blue{(Φ^⊤RΦ)^{-1}Φ^⊤R\purple{z}}
\EA$

- **IRLS for multiclass** $K>2$: Let $T∈\{0,1\}^{N×K}$ be one-hot target where $t_i$ is the $i$-th row, $t_{:j}$ is the $j$-th column, $\sum_kt_{ik}=1$ and $\yello{\sum_kt_{ik}I_{kj}=t_{ij}}$.
Let $W=(w_1,...,w_K)$.
$\BA
\dg{⇒}&p(C_k|ϕ_i)=y_k(ϕ_i)=\red{y_{ik}=\/{\e{a_{ik}}}{\sum_j\e{a_{ij}}}},\ \red{a_{ik}=w_k^⊤ϕ_i}\dg{⇒}y_{ik}'(a_{ij})=-\/{\e{a_{ik}}\e{a_{ij}}}{(\sum_j\e{a_{ij}})^2}+\/{\e{a_{ik}}}{\sum_j\e{a_{ij}}}I_{kj}=y_{ik}(I_{kj}-y_{ij})\\
&\dg{⇒}∇_{w_j}y_{ik}=y_{ik}'(a_{ij})a_{ij}'(w_j)=y_{ik}(I_{kj}-y_{ij})ϕ_i\\
\dg{⇒}&p(T|W)=\prod_{i=1}^N\prod_{k=1}^Kp(C_k|ϕ_i)^{t_{ik}}=\prod_i\prod_ky_{ik}^{t_{ik}}
\dg{⇒}E(W)=-\sum_i\sum_kt_{ik}\ln y_{ik}\dg{\t{ multinomial cross entropy error}}\\
&\dg{⇒}∇_{w_j}E(W)=∇_{y_{ik}}E(W)∇_{w_j}y_{ik}=-\sum_i\sum_k\/{t_{ik}}{y_{ik}}y_{ik}(I_{kj}-y_{ij})ϕ_i=-\sum_i\sum_k(\yello{t_{ik}I_{kj}}-t_{ik}y_{ij})ϕ_i=-\sum_i(\yello{t_{ij}}-y_{ij})ϕ_i\\
&\dg{⇒}\green{∇_{w_j}E(W)=\sum_i(y_{ij}-t_{ij})ϕ_i=Φ^⊤(y_{:j}-t_{:j})}\\
&\dg{⇒}H_{kj}=∇_{w_j}∇_{w_k}E(W)=∇_{w_j}(\sum_iy_{ik}ϕ_i)=\sum_iy_{ik}(I_{kj}-y_{ij})ϕ_iϕ_i^⊤=Φ^⊤\diag[y_{ik}(I_{kj}-y_{ij})]Φ=Φ^⊤R_{kj}Φ\\
\dg{⇒}&w_\t{flat}=\BM w_1\\\vdots\\w_K\EM∈ℝ^{K(M+1)},\
∇_{w_\t{flat}}E(W)=\BM ∇_{w_1}E(W)\\\vdots\\∇_{w_K}E(W)\EM∈ℝ^{K(M+1)},\
H=\BM H_{11}&...&H_{1K}\\\vdots&\ddots&\vdots\\H_{K1}&...&H_{KK}\EM∈ℝ^{K(M+1)×K(M+1)}
\EA$



**Activation function**: in general $p(t_i=1|ϕ_i)=f(a_i)$ for activation function $f$ and $a_i=w^⊤ϕ_i$.
Suppose training labels are generated on hidden threshold $θ∼p(θ)$ with $t_i=\BC1&\blue{a_i≥θ}\\0&\t{otherwise}\EC$, then $\blue{f(a_i)=∫_{-∞}^{a_i}p(θ)\ dθ}$.
Logistic regression $y_i=f(a_i)=σ(a_i)$ uses threshold $p(θ)=σ'(θ)=\/{\e{-θ}}{(1+\e{-θ})^2}=σ(θ)(1-σ(θ))$ the standard logistic distribution.

- **Probit regression**: $p(t_i=1|ϕ_i)=Φ(a_i)$ where $a_i=w^⊤ϕ_i$ uses threshold $p(θ)=\Normal(θ|0,1)$. Probit $\e{-θ^2}$ has much thinner tail than logistic $\e{-θ}$.
Fatter tails are more robust for outliers.
Technically probit function is $Φ^{-1}$.

- **Robust activation**: Let $ϵ$ be probability that data is mislabelled.
Then $p(t|a_i)=(1-ϵ)σ(a_i)+ϵ(1-σ(a_i))$

**Canonical link function**: inverse activation function $f^{-1}$ is link function.
$∇_wE(w)=\sum_i(y_i-t_i)ϕ_i=Φ^⊤(y-t)$ always simplifies to (prediction - target) × feature if $t_i$ belongs to an exponential family, and its canonical link inverse is chosen as activation function.
$\BA
\dg{⇒}&\red{p(t_i|η,s)=\/{1}{s}h(\/{t_i}{s})g(η)\e{\purple{\/{ηt_i}{s}}}}\dg{ - \t{linear term $\purple{\/{ηt_i}{s}}$ in exponential of $p(t_i|η)$}}\\
\dg{⇒}&∇_η[∫\/{1}{s}h(\/{t_i}{s})g(η)\e{\/{1}{s}ηt_i}\ dt_i]=0\dg{⇒}∇_ηg(η)∫\/{1}{s}h(\/{t_i}{s})\e{\/{1}{s}ηt_i}\ dt_i+g(η)∫\/{1}{s}h(\/{t_i}{s})\e{\/{1}{s}ηt_i}\/{1}{s}t_i\ dt_i=0\\
&\dg{⇒}\/{∇_ηg(η)}{g(η)}∫p(t_i|η,s)\ dt_i+∫\/{1}{s}t_ip(t_i|η,s)\ dt_i=0\dg{⇒}\/{∇_ηg(η)}{g(η)}+\E[\/{1}{s}t_i|η]=0\dg{⇒}\blue{y_i=\E[t_i|η]=-s∇_η\ln g(η)}\dg{⇒}\green{η=ψ(y_i)}\\
&\dg{⇒\t{$y_i=f(a_i)$ is a function of $η$, therefore the inverse is true and $η=ψ(f(a_i))$}}\\
\dg{⇒}&∇_w\ln p(t|η,s)=∇_w\sum_i[\ln g(η)+\/{1}{s}ηt_i]=\sum_i[-\/{1}{s}(-s∇_η\ln g(η))+\/{t_i}{s}]ψ'(y_i)f'(a_i)a_i'(w)=\/{1}{s}\sum_i(t_i-y_i)\yello{ψ'(y_i)f'(a_i)}ϕ_i\\
&\dg{⇒}∇_wE(w)=-∇_w\ln p(t|η,s)=\sum_i(y_i-t_i)ϕ_i\dg{⇒}\yello{ψ'(f)f'(y)=1}⇔ψ(f(y))=y\dg{⇒}f(y)=ψ^{-1}(y)\dg{⇒}\red{f^{-1}(y)=ψ(y)}\\
&\dg{⇒\t{this says if $∇_wE(w)=\sum_i(y_i-t_i)ϕ_i$ then $f^{-1}(y)=ψ(y)$ must be true}}\\
&\dg{⇒}\red{ψ(y_i):=a_i⇔y_i=f(a_i)}\dg{ - \t{$ψ=f^{-1}$ is the link function, which is the inverse of the activation function}}
\EA$

- **Gaussian**: Linear activation function $y_i=a_i$ and sum of squares error from MLE.
$\BA
\dg{⇒}&p(t_i|μ,σ^2)=\Normal(t_i|μ,σ^2)=\/{1}{(2πσ^2)^{1/2}}\e{-\/{t_i^2}{2σ^2}+\purple{\/{μt_i}{σ^2}}-\/{μ^2}{2σ^2}}\dg{⇒}\purple{\/{ηt_i}{s}}=\purple{\/{μt_i}{σ^2}}\dg{⇒}\BC η=μ\\s=σ^2\EC\\
&\dg{⇒}\blue{y_i=\E[t_i|η]}=η\dg{⇒}\BC
\green{η=ψ(y_i)}=y_i\\
ψ(y_i)=a_i\EC\dg{⇒}y_i=a_i=w^⊤ϕ(x_i)\dg{ - \t{linear activation function}}\\
&\dg{⇒}E_i(w)=-\ln p(t_i|y_i,σ^2)=\/{1}{2\dg{σ^2}}\n{y_i-t_i}^2\dg{⇒}\yello{\/{∂E_i}{∂a_i}=y_i-t_i}\dg{ - \t{MLE gives RSS as canonical error}}\\
\EA$

- **Bernoulli**: Logistic activation function $y_i=σ(a_i)$ and binary cross-entropy error from MLE.
$\BA
\dg{⇒}&p(t_i|μ)=μ^{t_i}(1-μ)^{1-t_i}=(1-μ)\e{\purple{t_i\ln\/{μ}{1-μ}}}\dg{⇒}\purple{\/{ηt_i}{s}}=\purple{t_i\ln\/{μ}{1-μ}}\dg{⇒}\BC η=\ln\/{μ}{1-μ}\\s=1\EC\\
&\dg{⇒}\blue{y_i=\E[t_i|η]}=μ\dg{⇒}\BC
\green{η=ψ(y_i)}=\ln\/{y_i}{1-y_i}\\
ψ(y_i)=a_i\EC\dg{⇒}y_i=\/{1}{1+\e{-a_i}}=\/{1}{1+\e{-w^⊤ϕ(x_i)}}\dg{ - \t{logistic activation function}}\\
&\dg{⇒}E_i(w)=-\ln p(t_i|y_i)=-[t_i\ln y_i+(1-t_i)\ln (1-y_i)]\dg{ - \t{MLE gives binary cross entropy as canonical error}}\\
&\dg{⇒}\/{∂E_i}{∂a_i}=\/{∂E_i}{∂y_i}\/{∂y_i}{∂a_i}=-[\/{t_i}{y_i}-\/{1-t_i}{1-y_i}]y_i(1-y_i)\dg{⇒}\yello{\/{∂E_i}{∂a_i}=y_i-t_i}\\
\EA$

- **Categorical**: Softmax activation function $y_i=\left[\/{\e{a_k}}{\sum_j\e{a_j}}\right]_{k=1..K}$ and multi-class cross-entropy error from MLE.
$\BA
\dg{⇒}&p(t_i|μ)=\prod_kμ_k^{t_{ik}}=\e{\sum_kt_{ik}\ln μ_k}\dg{⇒}\BC η=[\ln μ_k]_{k=1..K}\\s=1\EC\\
&\dg{⇒}\blue{y_i=\E[t_i|η]}=μ\dg{⇒}\BC
\green{η=ψ(y_i)}=[\ln μ_k]_{k=1..K}\\
ψ(y_i)=a_i\EC\dg{⇒}y_i=\/{\e{a_k}}{\sum_j\e{a_j}}=\/{\e{w_k^⊤ϕ(x_i)}}{\sum_j\e{w_j^⊤ϕ(x_i)}}\dg{ - \t{softmax activation function}}\\
&\dg{⇒}E_i(w)=-\ln p(t_i|y_i)=-\sum_kt_{ik}\ln y_{ik}\dg{ - \t{MLE gives multinomial cross entropy as canonical error}}\\
&\dg{⇒}\/{∂E_i}{∂a_{ik}}=\sum_{j=1}^K\/{∂E_i}{∂y_{ij}}\/{∂y_{ij}}{∂a_{ik}}=-\sum_j\/{t_{ij}}{y_{ij}}y_{ij}(I_{kj}-y_{ik})=-\sum_jt_{ij}(I_{kj}-y_{ik})=y_{ik}-t_{ij}\dg{⇒}\yello{\/{∂E_i}{∂a_{ik}}=y_{ik}-t_{ij}}\\
\EA$

- **Poisson**: exponential activation function $y_i=\e{a_i}$
$\BA
\dg{⇒}&p(t_i|μ)=\/{μ^{t_i}e^{-μ}}{t_i!}=\dg{\e{-t_i!}}\e{-μ}\e{\purple{t_i\ln μ}}\dg{⇒}\purple{\/{ηt_i}{s}}=\purple{t_i\ln μ}\dg{⇒}\BC η=\ln μ\\s=1\EC\\
&\dg{⇒}\blue{y_i=\E[t_i|η]}=μ\dg{⇒}\BC
\green{η=ψ(y_i)}=\ln μ\\
ψ(y_i)=a_i\EC\dg{⇒}y_i=\e{a_i}\dg{ - \t{exponential activation function}}\\
&\dg{⇒}E_i(w)=-\ln p(t_i|y_i)=-[t_i\ln y_i-y_i]\dg{ - \t{MLE gives Poisson error as canonical error}}\\
&\dg{⇒}\/{∂E_i}{∂a_i}=-\/{∂E_i}{∂y_i}\/{∂y_i}{∂a_i}=-[\/{t_i}{y_i}-1]y_i\dg{⇒}\yello{\/{∂E_i}{∂a_i}=y_i-t_i}
\EA$

**Laplace approximation**: find Gaussian $q(z)$ to approximate unimodal true density $p(z)=\/{1}{Z}f(z)$ where $Z=∫f(z)\ dz$ is normalization constant using second-order Taylor approximation of log density $\ln f(z)$ at its peak $z_0$.
$\BA
\dg{⇒}\t{Find mode ($z_0$):}&\/{df}{dz}|_{z=z_0}=0\\
\dg{⇒}\t{Taylor expansion:}&\ln f(z)=\ln f(z_0)-\/{1}{2}\blue{A}(z-z_0)^2,\ \blue{A=-\/{d^2}{dz^2}\ln f(z)|_{z=z_0}}\dg{⇒\t{ 2nd order because }}\/{d}{dz}\ln f(z)|_{z=z_0}=0\\
&\dg{⇒}f(z)=f(z_0)\e{-\/{\blue{A}}{2}(z-z_0)^2}\dg{⇒}\red{q(z)=(\/{\blue{A}}{2π})^{1/2}\e{-\/{\blue{A}}{2}(z-z_0)^2}}=\Normal(z|z_0,\blue{A}^{-1})\\
\dg{⇒}\t{Multivariate $z∈ℝ^M$:}&\ln f(z)=\ln f(z_0)-\/{1}{2}(z-z_0)^⊤\blue{A}(z-z_0),\ \blue{A=∇^2\ln f(z)|_{z=z_0}}∈\v{S}_{++}^{M}\\
&\dg{⇒}q(z)=\/{\det(A)^{1/2}}{(2π)^{M/2}}\e{-\/{1}{2}(z-z_0)^⊤\blue{A}(z-z_0)}=\Normal(z|z_0,\blue{A}^{-1})\\
\EA$

- **Bayesian information criterion (BIC)**: Using $Z=∫f(z)\ dz=f(z_0)\/{(2π)^{M/2}}{\det(A)^{1/2}}$ we can calculate $p(\m{D})$ model evidence.
$\BA
\dg{⇒}&p(\m{D})=∫p(\m{D}|w)p(w)\ dw\dg{⇒}\BC
f(w)=p(\m{D}|w)p(w)\\
Z=p(\m{D})\\
w_0=w_\t{MAP}
\EC
\dg{⇒}p(\m{D})=p(\m{D}|w_\t{MAP})p(w_\t{MAP})\/{(2π)^{M/2}}{\det(A)^{1/2}}\\
&\dg{⇒}\ln p(\m{D})=\ln p(\m{D}|w_\t{MAP})+\yello{\ln p(w_\t{MAP})+\/{M}{2}\ln(2π)-\/{1}{2}\ln\det(A)} \dg{-}\yello{\t{Occam factor}}\dg{\t{ penalizes complexity}}\\
\dg{⇒}&A=-∇^2\ln p(w_\t{MAP}|\m{D})=-∇^2\ln p(\m{D}|w_\t{MAP})p(w_\t{MAP})+c1≈-∇^2\ln p(\m{D}|w_\t{MAP}) \dg{-}\dg{p(w_\t{MAP})\t{ relatively flat}}\\
&\dg{⇒}A≈-∇^2\sum_{i=1}^N\ln p(x_i|w_\t{MAP})≈NH_0\dg{-}\dg{H_0\t{ is average Hessian for each }i}\\
&\dg{⇒}\ln\det(A)=\ln\det(NH_0)=M\ln N+\ln\det(H_0)\\
&\dg{⇒}\yello{\ln p(w_\t{MAP})+\/{M}{2}\ln(2π)-\/{1}{2}\ln\det(A)}=\ln p(w_\t{MAP})+\/{M}{2}\ln(2π)-\/{M}{2}\ln(N)-\/{1}{2}\ln\det(H_0)\arr{N}-\/{M}{2}\ln(N)\\
\dg{⇒}&p(\m{D})≈p(\m{D}|w_\t{MAP})-\/{M}{2}\ln(N)\dg{ - \t{Bayesian information criterion}}
\EA$



**Bayesian logistic regression**:
$\BA
\dg{⇒}&p(w)=\Normal(w|m_0,S_0),\ p(t|w)=\prod_iy_i^{t_i}(1-y_i)^{1-t_i}\dg{⇒}p(w|t)∝p(w)p(t|w)\dg{ - \t{Gaussian prior + cross-entropy likelihood}}\\
\dg{⇒}&E(w)=-\ln p(w|t)=\/{1}{2}(w-m_0)^⊤S_0^{-1}(w-m_0)-\sum_it_i\ln y_i+(1-t_i)\ln(1-y_i)+c\\
&\dg{⇒}∇_wE(w)=S_0^{-1}(w-m_0)+Φ^⊤(y-t)\dg{⇒}∇_w^2E(w)=S_0^{-1}+Φ^⊤RΦ\dg{⇒}w^\t{(new)}=w^\t{(old)}-H^{-1}∇E(w^\t{(old)})\dg{⇒}w_\t{MAP}\dg{ - \t{IRLS}}\\
\dg{⇒}&\BC m_N=w_\t{MAP}\\S_N^{-1}=∇_w^2E(w_\t{MAP})\EC\dg{⇒}q(w|t)=\Normal(w|m_N,S_N)\dg{ - \t{Laplace approximation of the posterior}}\\
&\dg{⇒}\green{p(C_1|ϕ_i,t)}=∫p(C_1|ϕ_i,w)p(w|t)\ dw≈∫σ(w^⊤ϕ_i)q(w|t)\ dw,\ w∈ℝ^M\dg{ - \t{multivariate logistic × Gaussian is intractable}^2}\\
\dg{⇒}&σ(w^⊤ϕ_i)=∫\purple{δ(a-w^⊤ϕ_i)}σ(a)\ da\dg{⇒}∫σ(w^⊤ϕ_i)q(w|t)\ dw=∫[∫\purple{δ(a-w^⊤ϕ_i)}σ(a)\ da]q(w|t)\ dw\dg{ - \t{delta-dirac trick}}\\
&\quad=∫σ(a)[∫\purple{δ(a-w^⊤ϕ_i)}q(w|t)\ dw]\ da\dg{⇒}p(a)=∫\purple{δ(a-w^⊤ϕ_i)}q(w|t)\ dw\dg{ - \t{ project density over $w$ onto $a=w^⊤ϕ_i$}}\\
&\dg{⇒}\BC μ_a=m_N^⊤ϕ_i\\σ_a^2=ϕ_i^⊤S_Nϕ_i\EC\dg{⇒}p(a)=\Normal(a|μ_a,σ_a^2)\dg{⇒}\green{p(C_1|ϕ_i,t)}=∫σ(a)\Normal(a|μ_a,σ_a^2)\ da\dg{ - \t{logistic × Gaussian still intractable}}\\
\dg{⇒}&σ(a)≈Φ(\yello{λ}a)\dg{ - \t{probit trick: both $σ$ and $Φ$ go through (0,0.5) and strech from 0 to 1. Use $λ$ to match slope at origin}}\\
&\BC
σ'(a)=σ(a)(1-σ(a))\dg{⇒}σ'(0)=\/{1}{4}\\
Φ'(\yello{λ}a)=\yello{λ}\Normal(\yello{λ}a|0,1)\dg{⇒}Φ'(0)=\/{\yello{λ}}{\sqrt{2π}}\EC
\dg{⇒}σ'(0)=Φ'(0)\dg{⇒}\yello{λ=\sqrt{\/{8}{π}}}\dg{⇒}\red{σ(a)≈Φ(\sqrt{\/{8}{π}}a)}\\
&\dg{⇒}u:=z-\yello{λ}a∼\Normal(u|-\yello{λ}μ_a,1+\yello{λ}^2σ_a^2)\dg{⇒}Φ(\yello{λ}a)=P(u≤0)=Φ(\/{0-(-\yello{λ}μ_a)}{\sqrt{1+\yello{λ}^2σ_a^2}})\dg{ - \t{get rid of $a$ to take it out of the integral}}\\
&\dg{⇒}\green{p(C_1|ϕ_i,t)}=∫σ(a)\Normal(a|μ_a,σ_a^2)\ da≈∫Φ(\yello{λ}a)\Normal(a|μ_a,σ_a^2)\ da=Φ(\/{\yello{λ}μ_a}{\sqrt{1+\yello{λ}^2σ_a^2}})∫\Normal(a|μ_a,σ_a^2)\ da=Φ(\/{\yello{λ}μ_a}{\sqrt{1+\yello{λ}^2σ_a^2}})≈σ(\/{μ_a}{\sqrt{1+\yello{λ}^2σ_a^2}})\\

\EA$

# ESL: 4 Linear Models for Classification

**OLS classification**: $K$ classes, training data $(X∈ℝ^{N×(p+1)},Y∈ℝ^{N×K})$ with one-hot rows in response, and coefficients $\hat{B}=(X^⊤X)^{-1}X^⊤Y=\small\BM \hat{β}_0\\\hat{β}\EM\normalsize∈ℝ^{(p+1)×K}$. Output $\hat{f}(x_i)^⊤=(1,x_i^⊤)\hat{B}$ assigning class $\hat{G}(x_i)=\arg\max_k\hat{f}_k(x_i)$.

- Classification attempts to reproduce one-hot $y_i$ target.
$\v{1}^⊤\hat{f}(x_i)=1$, which is similar to a probability distribution.
It is not a true distribution because individual falues of $\hat{f}(x_i)$ can fall outside of $[0,1]$.

- Masking problem overshadows the middle class $\hat{f}_\t{middle}(x_i)$ by edge classes $\hat{f}_\t{edge1}(x_i)$ and $\hat{f}_\t{edge2}(x_i)$ because the regression line is linear.
The problem can be overcome with quadratic regression.
When $K≥3$ classes are lined up, polynomial regression with terms up to degree $K-1$ might be needed to resolve them.

**Linear discriminant analysis (LDA)** is a generative model based on Gaussian likelihood $p(x_i|G=k)=\Normal(x_i|μ_k,Σ)$ for each class.
LDA and QDA work well not because real-world data are Gaussian but because of bias-variance trade-off.
They are stable estimators with low variance.
$\BA
\dg{⇒}&P(G=k|x_i)=\/{p(x_i|G=k)π_k}{\sum_{j=1}^Kp(x_i|G=j)π_j}\dg{⇒}\red{p(x_i|G=k)=\Normal(x_i|μ_k,Σ)}\dg{ - \t{assume Gaussian likelihood and shared covariance }Σ_k=Σ}\\
\dg{⇒}&\ln\/{P(G=k|x_i)}{P(G=l|x_i)}=\ln\/{\yello{π_k}}{π_l}-\/{1}{2}[\yello{(x_i-μ_k)^⊤Σ^{-1}(x_i-μ_k)}-(x_i-μ_l)^⊤Σ^{-1}(x_i-μ_l)]\dg{ - x_i^⊤Σ^{-1}x_i\t{ cancels out in posterior log-odds}}\\
&=\ln\/{\yello{π_k}}{π_l}-\/{1}{2}[\yello{-2x_iΣ^{-1}μ_k+μ_k^⊤Σ^{-1}μ_k}+2x_iΣ^{-1}μ_l-μ_l^⊤Σ^{-1}μ_l]=\grass{(μ_k-μ_l)^⊤Σ^{-1}}x_i\brown{-\/{1}{2}(μ_k+μ_l)^⊤Σ^{-1}(μ_k-μ_l)+\ln\/{π_k}{π_l}}\\
&\dg{⇒\t{LDA binary discriminant }}f(x_i)=\grass{β^⊤}x_i+\brown{β_0}>0⇒\hat{G}(x_i)=k\dg{\t{ else }}\hat{G}(x_i)=l\\
&\dg{⇒\t{LDA $K>2$ discriminant }}δ_k(x_i)=\yello{x_i^⊤Σ^{-1}μ_k-\/{1}{2}μ_k^⊤Σ^{-1}μ_k+\ln π_k}\dg{⇒}\hat{G}(x_i)=\arg\max_kδ_k(x_i)\\
\dg{⇒}&\hat{π}_k=\/{N_k}{N},\ \hat{μ}_k=\/{1}{N_k}\sum_{g_i=k}x_i,\ \blue{\hat{Σ}=\/{1}{N-K}\sum_k\sum_{g_i=k}(x_i-\hat{μ}_k)(x_i-\hat{μ}_k)^⊤}\dg{ - \t{$g_i$ is actual class of $x_i$}}\\
\EA$

- **Whitening/Sphering**: in practice LDA uses whitening/sphering on data to transform the Mahalanobis distance $\yello{(x_i-μ_k)^⊤Σ^{-1}(x_i-μ_k)}$ to using Euclidean norm distance $\n{\tilde{x}_i-\tilde{μ}_k}^2$ for classification.
$\BA
\dg{⇒}&\hat{Σ}=UDU^⊤\dg{⇒}\tilde{x}_i=D^{-1/2}U^⊤x_i,\ \tilde{μ}_k=D^{-1/2}U^⊤μ_k\dg{ - \t{sphering: $\V[x_i|g_i=k]=\hat{Σ}⇒\V[\tilde{x}_i|g_i=k]=I$}}\\
\dg{⇒}&δ_k(x_i)=\yello{-\/{1}{2}(x_i-μ_k)^⊤Σ^{-1}(x_i-μ_k)+\ln(π_k)}=\yello{-\/{1}{2}\n{\tilde{x}_i-\tilde{μ}_k}^2+\ln(π_k)}\\
&\dg{⇒}\hat{G}(x_i)=\arg\max_kδ_k(x_i)=\arg\min_k[\n{\tilde{x}_i-\tilde{μ}_k}^2-2\ln(π_k)]\\
\EA$

- **Reduced rank LDA (Fisher)**: $K$ points can form at most $K-1$ linearly independent vectors.
K centroids can span a subspace of at most dimension $K-1$.
When $p≫K$ we can ignore the additional orthogonal dimensions as they would contribute no discriminitive value.
Let $M∈ℝ^{K×p}$ where $μ_k$ is the $k$-th row. Then $M^*=MΣ^{-1/2}$ contains sphered centroids and $S_B^*=\V[M^*]$ is sphered between-class covariance. The eigenvectors of $B^*=V^*D_BV^{*⊤}$ contained in $V^*$ are the discriminant axes.

  - Rank reduction does not alter the result of classification. In practice it offers computational and storage efficiency.

- In binary classification, LDA coefficients $\grass{β=Σ^{-1}(μ_k-μ_l)}$ is proportional to Fisher projection axis $w∝S_W^{-1}(μ_k-μ_l)$ and OLS coefficients.
Coefficients represent the normal of the separating hyperplane, therefore the decision boundaries are strictly parallel.
The following proof uses target $y_i=\{+1,-1\}$ for algebraic simplicity.
$\BA
\dg{⇒}&X∈ℝ^{N×p},β∈ℝ^p\dg{⇒}E(β)=\n{(X-\v{1}\bar{x}^⊤)β-(y-\bar{y}\v{1})}^2\dg{⇒}∇_βE(β)=[2((X-\v{1}\bar{x}^⊤)β-(y-\bar{y}\v{1}))^⊤(X-\v{1}\bar{x}^⊤)]^⊤=0\\
&\dg{⇒}(X-\v{1}\bar{x}^⊤)^⊤(X-\v{1}\bar{x}^⊤)β=(X-\v{1}\bar{x}^⊤)^⊤(y-\bar{y}\v{1})\quad\dg{⇒\t{LHS}}=Σβ\quad\dg{⇒\t{RHS}}=X^⊤y-X^⊤\bar{y}\v{1}-\bar{x}\v{1}^⊤y+\bar{x}\v{1}^⊤\v{1}\bar{y}\\
&=X^⊤y-N\bar{x}\bar{y}-\bar{x}N\bar{y}+\bar{x}N\bar{y}=X^⊤y-N\bar{x}\bar{y}=N_kμ_k-N_lμ_l-N(\/{N_k-N_l}{N})(\/{N_kμ_k+N_lμ_l}{N})=...=\/{2N_kN_l}{N}(μ_k-μ_l)\\
&\dg{⇒}Σβ=\/{2N_kN_l}{N}(μ_k-μ_l)\dg{⇒}\grass{β∝Σ^{-1}(μ_k-μ_l)}
\EA$

  - Recall when target is $y_i=\{+\/{N}{N_k},-\/{N}{N_l}\}$, OLS coefficients are identical to Fisher coefficients.

- **Quadratic discriminant analysis (QDA)**: If we drop the assumption that $Σ_k=Σ$ for all classes, then the quadratic $x_i^⊤Σ_k^{-1}x_i$ term remains and decision boundaries are quadratic curves (parabola, ellipsoid, hyperbola).
$\BA
\dg{⇒}&δ_k(x_i)=-\/{1}{2}\ln\det(Σ_k)-\/{1}{2}(x_i-μ_k)^⊤Σ^{-1}(x_i-μ_k)+\ln(π_k)\dg{ - \t{now we eigendecompose $\hat{Σ}_k$}}\\
\dg{⇒}&\hat{Σ}_k=U_kD_kU_k^⊤\dg{⇒}\hat{Σ}_k^{-1}=U_k^⊤D_k^{-1}U_k\dg{⇒}δ_k(x_i)=-\/{1}{2}\sum_j\ln λ_{kj}-\/{1}{2}[U_k^⊤(x_i-\hat{μ}_k)]^⊤D_k^{-1}[U_k^⊤(x_i-\hat{μ}_k)]+\ln(π_k)
\EA$

  - QDA has $K-1$ additional $Σ_k∈ℝ^{p×p}$ parameters to find from training data.

**Logistic regression**: We model posteriors of $K$ classes using anchored softmax with $β∈ℝ^{p×(K-1)}$
$\BA
\dg{⇒}&k=1,...,K-1\dg{⇒}\ln\/{P(g_i=k|x_i)}{P(g_i=K|x_i)}=β_{k0}+β_k^⊤x_i\dg{⇒}P(g_i=k|x_i)=P(g_i=K|x_i)\e{β_{k0}+β_k^⊤x_i}\\
&\dg{⇒}P(g_i=K|x_i)=\/{1}{1+\sum_{j=1}^{K-1}\e{β_{j0}+β_j^⊤x_i}}\dg{⇒}P(g_i=k|x_i)=\/{\e{β_{k0}+β_k^⊤x_i}}{1+\sum_{j=1}^{K-1}\e{β_{j0}+β_j^⊤x_i}}
\EA$

- Binary classification is trained from MLE $l(θ)=\sum_i\ln p_{g_i}(x_i;θ)$ where notation $p_{g_i}(x_i;θ):=p(g_i|x_i,θ)$.
$\BA
\dg{⇒}&β:=(β_{10},β_1)∈ℝ^{p+1},\ x_i:=(1,x_i)∈ℝ^{p+1}\dg{ - \t{assume intercept is included}}\\
\dg{⇒}&p_i:=p(g_i=1|x_i,β)=\/{\e{β^⊤x_i}}{1+\e{β^⊤x_i}},\ p(g_i=2|x_i,β)=\/{1}{1+\e{β^⊤x_i}}=1-p_i\dg{ - g_i=1⇒y_i=1\t{ and }g_i=2⇒y_i=0}\\
\dg{⇒}&l(β)=\sum_iy_i\ln p_i+(1-y_i)\ln(1-p_i)=\sum_i(y_i\ln \/{\e{β^⊤x_i}}{1+\e{β^⊤x_i}}+\ln \/{1}{1+\e{β^⊤x_i}}-y_i\ln \/{1}{1+\e{β^⊤x_i}})\\
&=\sum_i(y_iβ^⊤x_i-\ln(1+\e{β^⊤x_i}))\dg{⇒}l'(β)=\sum_i(y_ix_i-\/{\e{β^⊤x_i}x_i}{1+\e{β^⊤x_i}})=\sum_i(y_i-p_i)x_i=X^⊤(y-p)\\
&\dg{⇒}l''(β)=-\sum_ip_i(1-p_i)x_ix_i^⊤=-X^⊤RX\dg{ - \t{$p_i'(β)=p_i(1-p_i)x_i$}}\\
\dg{⇒}&\dg{\t{IRLS: }}β^\t{(new)}=β^\t{(old)}-l''(β)^{-1}l'(β)=\red{(X^⊤RX)^{-1}X^⊤Rz}=\arg\min_β(z-Xβ)^⊤R(z-Xβ)\dg{ - \t{MLE of WLS}}\\
&\dg{⇒}\blue{R=\diag[p_i^\t{(old)}(1-p_i^\t{(old)}))],\ z=Xβ^\t{(old)}+R^{-1}(y-p^\t{(old)})\dg{⇒}z_i=\hat{β}^⊤x_i+\/{y_i-\hat{p}_i}{\hat{p}_i(1-\hat{p}_i)}}\\
\dg{⇒}&\dg{\t{RSS: }}\n{R^{1/2}(z-X\hat{β})}^2=\n{R^{1/2}(X\hat{β}+R^{-1}(y-\hat{p})-X\hat{β})}^2=\n{R^{-1/2}(y-\hat{p})}^2=\sum_i\/{(y_i-\hat{p}_i)^2}{\hat{p}_i(1-\hat{p}_i)}=\sum_i\/{(y_i-\E[y_i])^2}{\V[y_i]}∼χ_N^2\\
&\dg{⇒\t{WLS/IRLS residual sum of squares is a Chi-square statistic}}
\EA$

  - $\hat{β}$ is MLE: it is asymptotically consistent $\E[\hat{β}]\arr{p}β$ and efficient $\V[\hat{β}]\arr{p}\/{1}{I_n(β)}=\/{1}{-l''(β)}=(X^⊤RX)^{-1}$.

  - $R=\V[y]$ gives higher weights to points with higher uncertainty. This emphasizes points near the boundary, and de-emphasizes points far from boundary, thereby making it much more robust to outliers.

- **LDA vs logistic regression**: LDA and logistic regression both have linear decision boundaries based on log-odds.

  - LDA assumes Gaussian likelihood $p(x_i|G=k)=\Normal(x_i|μ_k,Σ)$ for each class. Every point contributes equally to the MLE parameters of that likelihood function, which leads to susceptibility to outlier data points.
  Logistic regression is much more robust because it gives higher weight to points near the decision boundary.

# PRML: 5 Neural Networks

**Feed-forward networks** use hidden layers acting as basis functions instead of fixed basis functions $y(x)=f(w^⊤ϕ(x))$.
$\BA
\dg{⇒}&x∈ℝ_i^{D+1}→\block{z_j=h(a_j^{(1)}=w_j^{(1)⊤}x)}→z∈ℝ_j^{M+1}→\block{y_k=f(a_k^{(2)}=w_k^{(2)⊤}z)}→y∈ℝ_k^K\\
&x=\BM 1\\x_1\\\vdots\\x_D\EM∈ℝ^{D+1}\dg{⇒}z(x)=\BM1\\h(w_1^{(1)⊤}x)\\\vdots\\h(w_M^{(1)⊤}x)\EM∈ℝ^{M+1}
\dg{⇒}y(x)=\BM f(w_1^{(2)⊤}z)\\\vdots\\f(w_K^{(2)⊤}z)\EM∈ℝ^K\dg{ - \t{forward propagation}}\\
&\dg{⇒\t{Hidden unit $j=1..M$ has input $x$, activation $a_j^{(1)}=a_j^{(1)}(x)=w_j^{(1)⊤}x$, and output $z_j=z_j(x)=h(a_j^{(1)})$}}\\
&\dg{⇒\t{Output unit $k=1..K$ has input $z=z(x)$, activation $a_k^{(2)}=a_k^{(2)}(z)=w_k^{(2)⊤}z$, and output $y_k=y_k(x)=f(a_k^{(2)})$}}\\
\EA$

- **Weight index convention**: $w_j^{(1)}$ refers to the weight in hidden layer $z_j=z_j(x)=w_j^{(1)⊤}x=\sum_{k=1}^Dw_{jk}^{(1)}x_k$. Here $w_{jk}^{(1)}$ refers to destination $z_j$ from source feature $x_k$.

  - Canonical link activation functions are used. $f(a_k)=a_k$ for linear regression, $f(a_k)=\/{1}{1+\e{-a_k}}$ for binary classification, and $f(a_k)=\/{\e{a_k}}{\sum_j\e{a_j}}$ for multiclass classification.

- **Universal approximation theorem**: We attempt to figure out an unknown underlying rule of the universe that generated outputs from inputs.
A simple input ⟶ hidden ⟶ output two-layer network with enough hidden units can uniformly (no blowup/jump) approximate any continuous function on a limited finite input domain to an arbitrary level of accuracy.

- **Weight-space symmetry**: the choice for weights to achieve the same input ⟶ output mapping is not unique.

  - Sign-flip: activation functions are odd. If we flip input and output sign of hidden unit, then weight also flips sign.

  - Ordering of hidden units don't matter. Indexing of weights change with indexing of hidden units.

**Error backpropagation** refers only to the procedure to calculate $∇E(w)$, before finding $w$ for that layer.
There are $M+K$ units.
**Unit error** $\red{δ_j=\/{∂E}{∂a_j}}$ represents derivative of error wrt activation at every unit.
$\BA
\dg{⇒}&x∈ℝ_i^{D+1}→\block{z_j=h(a_j^{(1)}=w_j^{(1)⊤}x)}→z∈ℝ_j^{M+1}→\block{y_k=a_k^{(2)}=w_k^{(2)⊤}z}→y∈ℝ_k^K\dg{ - \t{regression output layer}}\\
\dg{⇒}&E(w)=\sum_{n=1}^NE_n(w)\dg{ - \t{total error. De-index $n$ for training data $x∈ℝ^{D+1}$, $y∈ℝ^K$, $t∈ℝ^K$, and $E:=E_n$}}\\
\dg{⇒}&a_k=a_k(z)=w_k^⊤z\dg{⇒}\/{∂E}{∂w_{kj}}=\yello{\/{∂E}{∂a_k}}\/{∂a_k}{∂w_{kj}}\dg{⇒}\yello{δ_k:=\/{∂E}{∂a_k}}∈ℝ\dg{⇒}\red{\/{∂E}{∂w_{\green{kj}}}=\yello{δ_\green{k}}z_{\green{j}}}\\
&\dg{⇒\t{let $z$ be input, then all layers and units satisfy $∇_{w_{kj}}E=δ_jz_j$. Once we have all the $δ_j$-s we have $∇_wE(w)$}}\\
\dg{⇒}&\dg{\t{Output layer: }}E=\/{1}{2}\n{y-t}^2=\/{1}{2}\sum_{k=1}^K(y_k-t_k)^2,\ y_k=a_k=w_k^⊤z=\sum_{j=0}^Mw_{kj}z_j\\
&\dg{⇒}∇_{w_k}E=(y_k-t_k)z\dg{⇒}\/{∂E}{∂w_{kj}}=(\yello{y_k-t_k})z_j\dg{⇒}\/{∂E}{∂a_k}=\/{∂E}{∂y_k}=y_k-t_k\dg{⇒}\red{δ_k^{(2)}=y_k-t_k}\dg{ - \t{canonical activation}}\\
\dg{⇒}&\dg{\t{Hidden layer: }}a_k^{(2)}=w_k^{(2)⊤}\BM1\\\\h(a_1^{(1)})\\\vdots\\h(a_M^{(1)})\EM\dg{⇒}\/{∂a_k^{(2)}}{∂a_j^{(1)}}=w_{kj}^{(2)}h'(a_j^{(1)})\dg{⇒}\/{∂E}{∂a_j^{(1)}}=\sum_{k=1}^K\/{∂E}{∂a_k^{(2)}}\/{∂a_k^{(2)}}{∂a_j^{(1)}}=h'(a_j^{(1)})\sum_kδ_k^{(2)}w_{kj}^{(2)}\\
&\dg{⇒}\red{δ_j^{(1)}=h'(a_j^{(1)})\sum_kδ_k^{(2)}w_{kj}^{(2)}}\dg{ - \t{no $t_j$ access for earlier layers. Back-propagate sum of downstream errors}}\\
\dg{⇒}&\dg{\t{Error function is determined by output layer activation function MLE}}\\
&\dg{⇒\t{let: }}x∈ℝ_i^{D+1}→\block{z_j=h(a_j^{(1)}=w_j^{(1)⊤}x)}→z∈ℝ_j^{M+1}→\block{y_k=σ(a_k^{(2)}=w_k^{(2)⊤}z)}→y∈ℝ_k^K\dg{ - \t{logistic output layer}}\\
&\dg{⇒\t{then: }}E=-[t_k\ln y_k+(1-t_k)\ln(1-y_k)]\dg{⇒}δ_k^{(2)}=\/{∂E}{∂a_k}=-\/{∂E}{∂y_k}\/{∂y_k}{∂a_k}=-(\/{t_k}{y_k}-\/{1-t_k}{1-y_k})y_k(1-y_k)=y_k-t_k\\
\EA$

- **Example**: $\BA
x∈ℝ_i^{D+1}→\block{z_j=\tanh(a_j^{(1)}=w_j^{(1)⊤}x)}→z∈ℝ_j^{M+1}→\block{y_k=a_k^{(2)}=w_k^{(2)⊤}z}→y∈ℝ_k^K
\EA$
$\BA
\dg{⇒}h(a)=\tanh(a)=2σ(2a)-1=\/{e^a-e^{-a}}{e^a+e^{-a}}\dg{⇒}h'(a)=1-h(a)^2\quad\dg{⇒}E=\/{1}{2}\sum_k(y_k-t_k)^2\\
\dg{⇒\t{forward propagation: }}a_j^{(1)}=w_j^{(1)⊤}x\dg{⇒}z_j=\tanh(a_j^{(1)})\dg{⇒}a_k^{(2)}=w_k^{(2)⊤}z\dg{⇒}y_k=a_k^{(2)}\\
\dg{⇒\t{error backpropagation: }}δ_k^{(2)}=y_k-t_k\dg{⇒}δ_j^{(1)}=(1-z_j^2)\sum_kδ_k^{(2)}w_{kj}^{(2)}\\
\dg{⇒\t{gradients: }}∇_{w_j^{(1)}}E=δ_j^{(1)}x∈ℝ^{M+1}\dg{⇒}∇_{w_k^{(2)}}E=δ_k^{(2)}z∈ℝ^{K+1}
\EA$

- **Efficiency**: A 2-layer feed-forward network has $\red{W:=(D+1)×M+(M+1)×K}$ weights, $M+K$ units, and $(D+1)+(M+1)+K$ nodes.
There are $M+K$ error deltas to calculate. Most of the compute is on forward pass calculating $y$ and $z$ and backward pass calculating the error deltas from $y$ and $z$. Each pass is $O(W)$.

- **Numerical differences**: adjust each $w_j$ by a tiny bit and see the difference in $E(w)$. This method is used for verifying implementation. Very computationally expensive, uses $O(W^2)$.

- **Jacobian backpropagation**: $x∈ℝ_i^{D+1}\dg{⇒}\red{J_{ki}:=\/{∂y_k}{∂x_i}}$ represents derivative of a system output wrt an input.
$\BA
\dg{⇒}J_{ki}=\/{∂y_k}{∂x_i}=\sum_{j=1}^M\yello{\/{∂y_k}{∂a_j^{(1)}}}\green{\/{∂a_j^{(1)}}{∂x_i}}\dg{ - \t{factor $\green{\/{∂a_j^{(1)}}{∂x_i}}$ is known. We need to recursively expand $\yello{\/{∂y_k}{∂a_j^{(1)}}}=\sum_{l=1}^K\purple{\/{∂y_k}{∂a_l^{(2)}}}\brown{\/{∂a_l^{(2)}}{∂a_j^{(1)}}}$}}\\
\quad=\sum_{j=1}^M\left[\sum_{l=1}^K\purple{\/{∂y_k}{∂a_l^{(2)}}}\brown{\/{∂a_l^{(2)}}{∂a_j^{(1)}}}\right]\green{\/{∂a_j^{(1)}}{∂x_i}}=\sum_{j=1}^M\left[\sum_{l=1}^K\purple{\/{∂y_k}{∂a_l^{(2)}}}\brown{h'(a_j)w_{lj}^{(2)}}\right]\green{\/{∂a_j^{(1)}}{∂x_i}}\dg{⇒}\purple{\/{∂y_k}{∂a_l^{(2)}}}=\BC
I_{kl}σ'(a_l^{(2)})&\t{sigmoid}\\
I_{kl}y_k-y_ky_l&\t{softmax}\EC\\
\EA$

**Hessian**: Parameter matrices are flattened so that $w_{kj}^{(2)}$ and $w_{ji}^{(1)}$ are in one $ℝ^W$ vector.
$\/{∂E}{∂w_{kj}}$ and $\/{∂E}{∂w_{ji}}$ also are in one $ℝ^W$ vector.
Then $H∈ℝ^{W×W}$ is a matrix instead of a 3D tensor.

- **Local quadratic approximation**: network $E(w)$ is non-convex surface with many local minima. We can model $E(w)$ using 2nd order Taylors where $H$ dictates the shape of the surface near optimal point.
$\BA
\dg{⇒}&E(w)≈E(\hat{w})+(w-\hat{w})^⊤∇E(\hat{w})+\/{1}{2}(w-\hat{w})^⊤H(w-\hat{w}),\ H=∇^2E(\hat{w})\dg{ - \t{Taylor 2nd order in neighborhood of $\hat{w}$}}\\
&\dg{⇒}E(w)=E(w^*)+\/{1}{2}(w-w^*)^⊤H(w-w^*)\dg{ - \t{near $\hat{w}=w^*$ optimal point $∇E(w^*)=0$ and shape is dictated by $H$}}\\
&\dg{⇒}Hu_i=λ_iu_i\dg{⇒}w-w^*=\sum_iα_iu_i=U^⊤α\dg{ - \t{$α_i$ represents coordinate of $w-w^*$ along $u_i$}}\\
\dg{⇒}&E(w)=E(w^*)+\/{1}{2}\sum_iα_iu_i^⊤α_iλ_iu_i\dg{⇒}E(w)=E(w^*)+\/{1}{2}\sum_iα_i^2λ_i\dg{ - \t{positive-definite ($λ_i>0$) ⇒ minimum $w^*$}}
\EA$

  - Positive-definite $H$ ensures ellipsoidal level-set $(w-w^*)^⊤H(w-w^*)=c$ dimensional radii $λ_i>0$. If $λ_i=0$ then that dimension collapses into a half-pipe and there is no unique minimum.

- **Exact Hessian** can be found via backpropagation at $O(W^2)$
$\BA
\dg{⇒}&x∈ℝ_i^{D+1}→\block{z_j=h(a_j^{(1)}=w_j^{(1)⊤}x)}→z∈ℝ_j^{M+1}→\block{y_k=a_k^{(2)}=w_k^{(2)⊤}z}→y∈ℝ_k^K\\
\dg{⇒}&\dg{\t{$w_{kj}^{(2)}$ and $w_{k'j'}^{(2)}$: }}E=\/{1}{2}\n{y-t}^2\dg{⇒}\/{∂E}{∂w_{kj}}=\/{∂E}{∂a_k}\/{∂a_k}{∂w_{kj}}=δ_kz_j\\
&\dg{⇒}\green{M_{kk'}:=\/{∂^2E}{∂a_k∂a_{k'}}=\/{∂δ_k}{∂a_{k'}}}\dg{⇒}\/{∂^2E}{∂w_{kj}∂w_{k'j'}}=\/{∂δ_k}{∂w_{k'j'}}z_j=\green{\/{∂δ_k}{∂a_{k'}}}\/{∂a_{k'}}{∂w_{k'j'}}z_j\dg{⇒}\/{∂^2E}{∂w_{kj}^{(2)}∂w_{k'j'}^{(2)}}=\green{M_{kk'}}z_jz_{j'}\\
\dg{⇒}&\dg{\t{$w_{ji}^{(1)}$ and $w_{j'i'}^{(1)}$: }}\/{∂E}{∂w_{ji}}=\/{∂E}{∂a_j}\/{∂a_j}{∂w_{ji}}=\sum_k\/{∂E}{∂a_k}\/{∂a_k}{∂a_j}\/{∂a_j}{∂w_{ji}}=\sum_k\yello{δ_kh'(a_j)}w_{kj}x_i\dg{ - \t{need to account for $\/{∂\yello{δ_k}}{w_{j'i'}}$, and $\/{∂\yello{h'(a_j)}}{∂w_{j'i'}}$ if $\yello{j=j'}$}}\\
&\dg{⇒}\/{∂^2E}{∂w_{ji}^{(1)}w_{j'i'}^{(1)}}=x_ix_{i'}h''(a_{j'})\yello{I_{jj'}}\sum_kw_{kj'}^{(2)}δ_k+x_ix_{i'}h'(a_{j'})h'(a_j)\sum_k\sum_{k'}w_{k'j'}^{(2)}w_{kj}^{(2)}\green{M_{kk'}}\dg{ - \t{product rule}}\\
\dg{⇒}&\dg{\t{$w_{ji}^{(1)}$ and $w_{kj'}^{(2)}$: }}\/{∂E}{∂w_{ji}}=\sum_k\yello{δ_k}h'(a_j)\yello{w_{kj}}x_i\dg{ - \t{need to account for $\/{∂\yello{δ_k}}{∂w_{kj'}}$, and $\/{∂\yello{w_{kj}}}{∂w_{kj'}}$ if $\yello{j=j'}$}}\\
&\dg{⇒}\/{∂^2E}{∂w_{ji}^{(1)}w_{kj'}^{(2)}}=x_ih'(a_j)δ_k\yello{I_{jj'}}+x_iz_jh'(a_j)\sum_{k'}w_{k'j}^{(2)}\green{M_{kk'}}\dg{ - \t{product rule}}
\EA$

- **Pearlmutter trick** evaluates $v^⊤H$ in $O(W)$ without the $O(W^2)$ operation to find $H$ itself.
Operator $\red{\m{R}\{\blue{g}\}:=v^⊤∇_w\blue{g}}$ evaluates how much $g$ changes when the network's weights $w$ change by amount $v$.
$\BA
\dg{⇒}&\BC
1. &\m{R}\{x_i\}=0,\ \m{R}\{t_k\}=0\dg{ - \t{data is treated as a constant wrt change in network's weights}}\\
2. &\m{R}\{w_{ji}\}=v_{ji}\dg{ - v\t{ is the velocity vector corresponding to weights $w$}}\\
3. &\m{R}\{AB\}=\m{R}\{A\}B+A\m{R}\{B\}\dg{ - \t{product rule}}\\
4. &\m{R}\{h(a_j)\}=h'(a_j)\m{R}\{a_j\}\dg{ - \t{chain rule}}
\EC\\
\dg{⇒}&x∈ℝ_i^{D+1}→\block{z_j=h(a_j^{(1)}=w_j^{(1)⊤}x)}→z∈ℝ_j^{M+1}→\block{y_k=a_k^{(2)}=w_k^{(2)⊤}z}→y∈ℝ_k^K\\
&\dg{⇒}\grass{\m{R}\{z_j\}}=h'(a_j)\m{R}\{a_j\}=h'(a_j)\m{R}\{w_j^⊤x\}=h'(a_j)v_j^⊤x\\
&\dg{⇒}\brown{\m{R}\{y_k\}}=\m{R}\{w_k^⊤z\}=\m{R}\{w_k\}^⊤z+w_k^⊤\m{R}\{z\}=v_k^⊤z+w_k^⊤\m{R}\{z\}=\sum_jv_{kj}z_j+\sum_jw_{kj}h'(a_j)v_j^⊤x\\
&\dg{⇒}δ_k^{(2)}=y_k-t_k\dg{⇒}\green{\m{R}\{δ_k\}}=\brown{\m{R}\{y_k\}}\\
&\dg{⇒}δ_j^{(1)}=h'(a_j)\sum_kw_{kj}δ_k^{(2)}\dg{⇒}\yello{\m{R}\{δ_j\}}=h''(a_j)\m{R}\{a_j\}\sum_kw_{kj}δ_k+h'(a_j)\sum_kv_{kj}δ_k+h'(a_j)\sum_kw_{kj}\brown{\m{R}\{y_k\}}\\
\dg{⇒}&\dg{\t{backprop: now $v^⊤H=\m{R}\{\/{∂E}{∂w_{xx}}\}∈ℝ^W$ is a function of $v$. Iterating $v=I_l$ for $l=1,...,W$ returns exact Hessian $H$}}\\
&\dg{⇒}\/{∂E}{∂w_{kj}^{(2)}}=δ_k^{(2)}z_j\dg{⇒}\m{R}\{\/{∂E}{∂w_{kj}}\}=\green{\m{R}\{δ_k\}}z_j+δ_k\grass{\m{R}\{z_j\}}\\
&\dg{⇒}\/{∂E}{∂w_{ji}^{(1)}}=δ_j^{(1)}x_i\dg{⇒}\m{R}\{\/{∂E}{∂w_{ji}}\}=x_i\yello{\m{R}\{δ_j\}}
\EA$


**Hessian inverse approximations**: many algorithms require $H^{-1}$, which is $O(W^3)$.

- **Diagonal approximation**:
Let $w_{ji}$ be a weight in some layer connecting input $z_i$ to activation $a_j$.
True Hessian is strongly non-diagonal, because $w_j$ in one layer could strongly influence downstream $w_k$.
But by assuming different units do not interact, the Hessian is a diagonal matrix and $\/{∂^2E}{∂w_{ji}^2}$ can be solved in $O(W)$ using backpropagation.
Then inverse is also solved in $O(W)$.
$\BA
\dg{⇒}&\/{∂E}{∂w_{ji}}=\/{∂E}{∂a_j}\/{∂a_j}{∂w_{ji}}=\/{∂E}{∂a_j}z_i\dg{⇒}\red{\/{∂^2E}{∂w_{ji}^2}=\yello{\/{∂^2E}{∂a_j^2}}z_i^2}\dg{⇒}\yello{\/{∂^2E}{∂a_j^2}}=h'(a_j)^2\sum_{k=1}^K\sum_{k'=1}^Kw_{kj}w_{k'j}\/{∂^2E}{∂a_k∂a_{k'}}+h''(a_j)\sum_{k=1}^Kw_{kj}\/{∂E}{∂a_k}\\
&\dg{⇒\t{diagonal $\/{∂^2E}{∂w_{ji}^2}$ requires finding $\yello{\/{∂^2E}{∂a_j^2}}$, which involves cross-interactions between downstream units $k$ and $k'$}}\\
&\dg{⇒\t{Now assume zero off-diagonal - i.e., assume different units do not interact, then $\/{∂^2E}{∂a_k∂a_{k'}}→\yello{\/{∂^2E}{∂a_k^2}}$}}\\
\dg{⇒}&\red{\yello{\/{∂^2E}{∂a_j^2}}=h'(a_j)^2\sum_{k=1}^Kw_{kj}^2\yello{\/{∂^2E}{∂a_k^2}}+h''(a_j)\sum_{k=1}^Kw_{kj}\/{∂E}{∂a_k}}\dg{ - \t{Recursive downstream $\yello{\/{∂^2E}{∂a_k^2}}$ allows Hessian backpropagation}}\\
\EA$

- **Outer product approximation**: iteratively build $H$ and $H^{-1}$ by assuming $\E[y_n]=t_n$.
$\BA
\dg{⇒}&E=\/{1}{2}\n{y-t}^2\dg{⇒}∇_wE=\sum_{n=1}^N(y_n-t_n)∇_wy_n\dg{⇒}∇_w^2E=\sum_{n=1}^N∇_wy_n∇_wy_n^⊤+\sum_{n=1}^N(y_n-t_n)∇_w^2y_n\\
&\dg{⇒\t{assume $\E[y_n-t_n]=0$: }}∇_w^2E≈\sum_{n=1}^N∇_wy_n∇_wy_n^⊤\dg{⇒\t{regression $y_n=a_n$: }}\red{H≈\sum_{n=1}^Nb_nb_n^⊤,\ b_n=∇_wa_n}\\
\dg{⇒}&E=-\sum_{n=1}^N[t_n\ln y_n+(1-t_n)\ln(1-y_n)]\dg{⇒}∇_wE=\sum_{n=1}^N\/{y_n-t_n}{y_n(1-y_n)}∇_wy_n=\sum_{n=1}^N(y_n-t_n)∇_wa_n\dg{ - ∇_wy_n=y_n(1-y_n)∇_wa_n}\\
&\dg{⇒}∇_w^2E=\sum_{n=1}^N[∇_wy_n∇_wa_n+(y_n-t_n)∇_w^2a_n]\dg{⇒\t{assume $\E[y_n-t_n]=0$: }}\red{H≈\sum_{n=1}^Ny_n(1-y_n)b_nb_n^⊤,\ b_n=∇_wa_n}\\
\dg{⇒}&\dg{\t{Woodbury inversion lemma: $(Σ_{22}-Σ_{21}Σ_{11}^{-1}Σ_{12})^{-1}=Σ_{22}^{-1}+Σ_{22}^{-1}Σ_{21}Λ_{11}Σ_{12}Σ_{22}^{-1},\ Λ_{11}=(Σ_{11}-Σ_{12}Σ_{22}^{-1}Σ_{21})^{-1}$}}\\
&\dg{⇒\t{Iterative: }}\blue{H_{L+1}=H_L+b_{L+1}b_{L+1}^⊤}\dg{ - \t{$L$ points processed already}}\dg{⇒}\BM 1&b_{L+1}^⊤\\b_{L+1}&H_L\EM\dg{⇒}\blue{H_{L+1}^{-1}=H_L^{-1}+\/{H_L^{-1}b_{L+1}b_{L+1}^⊤H_L^{-1}}{1+b_{L+1}^⊤H_L^{-1}b_{L+1}}}\\
&\dg{⇒\t{Initialize: }}H_0=ϵI\dg{ - \t{for a small $ϵ$ we're in fact building $(H+H_0)^{-1}=(H+ϵI)^{-1}$}}
\EA$

  - Efficiency: finding output activation derivative $b_n=∇_wa_n$ is $O(W)$ by backprop.
  Constructing $b_nb_n^⊤$ is $O(W^2)$.
  Inverse numerator and denominator is each $O(W^2)$.
  Constructing $H$ and $H^{-1}$ is each $O(NW^2)$.

- **Numerical differences**: numerical differences model costs $O(W^3)$. A gradient backprop model costs $O(W^2)$.
Exact Hessian is almost always used instead for verifying implementation, but numerical differences are foolproof.
$\BA
\dg{⇒}\/{∂^2E}{∂w_{ji}∂w_{lk}}=\/{1}{4ϵ^2}[E(w_{ji}+ϵ,w_{lk}+ϵ)+E(w_{ji}-ϵ,w_{lk}-ϵ)-E(w_{ji}+ϵ,w_{lk}-ϵ)-E(w_{ji}-ϵ,w_{lk}+ϵ)]\dg{ - \t{each entry }O(W)}\\
\dg{⇒}\/{∂^2E}{∂w_{ji}∂w_{lk}}=\/{1}{2ϵ}[\/{∂E}{∂w_{ji}}(w_{lk}+ϵ)-\/{∂E}{∂w_{ji}}(w_{lk}-ϵ)]+O(ϵ^2)\dg{- \t{each column of $H$ is filled by 2 backprop runs $\/{∂E}{∂w}(w_{lk}±ϵ)$}}
\EA$

**Mixture density networks**: in **forward problems** features $x∈ℝ^D$ represent a configuration of factors that lead to one outcome represented by unimodal target $p(t|x)$.
In **inverse problems** $x∈ℝ^D$ represents an outcome, and many configurations that may have lead to this outcome are represented by multimodal target $p(t|x)$.
$\BA
\dg{⇒}&p(t|x)=\sum_k\yello{π_k(x)}\Normal(t|\yello{μ_k(x)},\yello{σ_k^2(x)})\dg{ - \t{3 activation function groups, heteroscedastic isotropic variance $Σ_k(x)=σ_k(x)^2I$}}\\
&\dg{⇒}π_k(x)=\/{\e{a_k^π(x)}}{\sum_{l=1}^K\e{a_l^π(x)}}\dg{⇒}\/{∂π_k}{∂a_j}=\/{\e{a_k}}{\sum_{l=1}^K\e{a_l}}I_{kj}-\/{\e{a_k}\e{a_j}}{(\sum_{l=1}^K\e{a_l})^2}\dg{⇒}\blue{\/{∂π_k}{∂a_j}=π_k(I_{kj}-π_j)}\\
&\dg{⇒}μ_k(x)=a_k^μ(x)\dg{ - \t{linear regression activation function output for $K$ means}}\\
&\dg{⇒}σ_k(x)=\e{a_k^σ(x)}∈ℝ_+\dg{ - \t{exponential activation function output. }}\\
\dg{⇒}&\red{γ_k(t|x)=\/{π_k\Normal(t|μ_k(x),σ_k^2(x))}{\sum_{j=1}^Kπ_j\Normal(t|μ_j(x),σ_j^2(x))}}\dg{ - \t{responsibility, posterior $p(C_k|t,x)$}}\\
\dg{⇒}&E(w)=\sum_nE_n(w)\dg{⇒}E_n(w)=-\ln p(t_n|x_n)\dg{⇒}E(w)=-\ln\sum_kπ_k\Normal(t|μ_k,σ_k^2)\dg{ - \t{de-index $n$ for training data}}\\
&\dg{⇒}\/{∂E}{∂a_k^π}=\sum_{l=1}^K\/{∂E}{∂π_l}\/{∂π_l}{∂a_k^π}=-\sum_l\/{\Normal(t|μ_l,σ_l^2)}{\sum_jπ_j\Normal(t|μ_j,σ_j^2)}π_k(I_{kl}-π_l)\blue{\dg{⇒}\/{∂E}{∂a_k^π}=π_k-γ_k}\\
&\dg{⇒}\/{∂E}{∂a_k^μ}=\/{∂E}{∂μ_k}\/{∂μ_k}{∂a_k^μ}=-\/{π_k\Normal(t|μ_k,σ_k^2)}{\sum_jπ_j\Normal(t|μ_j,σ_j^2)}(\/{t-μ_k}{σ_k^2})\dg{⇒}\blue{\/{∂E}{∂a_k^μ}=γ_k\/{μ_k-t}{σ_k^2}}\\
&\dg{⇒}\/{∂E}{∂a_k^σ}=\/{∂E}{∂σ_k}\/{∂σ_k}{∂a_k^σ}=-\/{π_k}{\sum_jπ_j\Normal(t|μ_j,σ_j^2)}(\/{-1}{\sqrt{2π}σ_k^2}\e{-\/{\n{t-μ_k}^2}{2σ_k^2}}+\Normal(t|μ_k,σ_2^2)\/{\n{t-μ_k}^2}{σ_k^3})\dg{⇒}\blue{\/{∂E}{∂a_k^σ}=γ_k(\/{1}{σ_k}-\/{\n{t-μ_k}^2}{σ_k^3})}
\EA$

**Bayesian neural network**: nonlinear network output $y(x,w)+ϵ$ with $x∈ℝ^D$ gives rise to non-convex posterior $p(w|\m{D})$. We use Laplace approximation.
$\BA
\dg{⇒}&\dg{\t{likelihood }}p(t|x,w,β)=\Normal(t|y(x,w),β^{-1})\dg{\t{ and isotropic prior }}p(w|α)=\Normal(w|0,α^{-1}I)\\
&\dg{⇒\t{posterior }}E_T(w)=-\ln p(w|\m{D})=\/{α}{2}\n{w}^2+\/{β}{2}\sum_n(y(x_n,w)-t_n)^2=\/{α}{2}\n{w}^2+βE(w)\dg{ - \t{$E(w)$ is unregularized MLE error}}\\
&\dg{⇒}∇_wE_T(w)=αw+β∇_wE\dg{⇒}∇_w^2E_T(w)=αI+βH\dg{ - \t{find $w_\t{MAP}$ by backprop and optimization}}\\
&\dg{⇒}\red{q(w|\m{D})=\Normal(w|w_\t{MAP},(αI+βH)^{-1})}\dg{ - \t{Laplace approximation of posterior}}\\
\dg{⇒}&\dg{\t{prediction }}p(t|x,\m{D})=∫p(t|x,w)q(w|\m{D})\ dw=∫\Normal(t|y(x,w),β^{-1})\Normal(w|w_\t{MAP},(αI+βH)^{-1})\ dw\dg{-y(x,w)\t{ nonlinear}}\\
&\dg{⇒}y(x,w)≈y(x,w_\t{MAP})+∇_wy(x,w_\t{MAP})^⊤(w-w_\t{MAP})\dg{-\t{now convolution integral can be solved with Adam-Eve}}\\
&\dg{⇒}\V[t|x,\m{D}]=\E[\V[t|x,\m{D}]]+\V[\E[t|x,\m{D}]]=β^{-1}+∇_wy(x,w_\t{MAP})^⊤(αI+βH)^{-1}∇_wy(x,w_\t{MAP})\\
\dg{⇒}&\dg{\t{evidence }}\ln p(\m{D}|α,β)=\/{W}{2}\ln α+\/{N}{2}\ln β-\/{1}{2}\ln\det(A)-E(w_\t{MAP})-\/{N}{2}\ln(2π)\dg{-\t{Bayesian evidence framework}}\\
&\dg{⇒\t{regression: }}E_T(w_\t{MAP})=\/{β}{2}\sum_n(y(x_n,w_\t{MAP})-t_n)^2+\/{α}{2}\n{w_\t{MAP}}^2\\
&\dg{⇒\t{iterative: }}γ=\sum_{j=1}^W\/{λ_j}{α+λ_j}\dg{⇒}α^\t{(new)}=\/{γ}{\n{w_\t{MAP}}^2},\ \/{1}{β^\t{(new)}}=\/{1}{N-γ}\sum_n(y(x_n,w_\t{MAP})-t_n)^2\dg{-\t{$λ_j$ eigenvalues of $βH$}}
\EA$
