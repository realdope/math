**Information theory**: information $\blue{h(x)=-\log_2 p(x)}$ is equivalent to surprise and uncertainty, which holds an inverse relationship with probability.
If $x,y$ are independent, then $h(x,y)=h(x)+h(y)$.

- **Discrete entropy**: $\blue{\H[x]=-\sum_xp(x)\log_2 p(x)}=\E[h(x)]$ is average information. Its unit is bits.

  - **8-state proof**: Suppose $x$ has 8 possible states (A-H).
  If uniform $p(x)=\/{1}{8}$ then $\H[x]=-\sum_x\/{1}{8}(-3)=3$ bits binary coding 001,010,011,100,101,110,111.
  If non-uniform distribution $[\/{1}{2},\/{1}{4},\/{1}{8},\/{1}{16},\/{1}{64},\/{1}{64},\/{1}{64},\/{1}{64}]$ then $\H[x]=\/{1}{2}(1)+\/{1}{4}(2)+\/{1}{8}(3)+\/{1}{16}(4)+\/{4}{64}(6)=2$ bits using variable coding 0,10,110,1110,111110.
  Entropy is the absolute lower bound on average number of bits to transmit the state of a variable.

  - **Thermodynamics**: In Maxwell-Boltzmann statistics, gas molecules are distinguishable.
  Number of ways to distribute $N$ particles into bins of size $n_1+...+n_k=N$ is called multiplicity $W=\/{N!}{\prod_in_i!}$.
  Using Stirling approximation ($\ln n!=n\ln n-n$)
  we have
  $\ln W=\ln(\/{N!}{\prod_in_i!})$
  $=\ln N!-\sum_i\ln n_i!$
  $вүҲN\ln N-N-\sum_in_i\ln n_i+\sum_in_i$
  $=N\ln N-\sum_in_i\ln n_i$
  $=\sum_in_i\ln N-\sum_in_i\ln n_i$
  $=-\sum_in_i\ln(\/{n_i}{N})$.
  Then we divide by $N$ to find entropy per particle
  $\lim\limits_{NвҶ’вҲһ}\/{1}{N}\ln W=-\sum_ip_i\ln(p_i)$
  $\dg{вҮ’}\red{H=-\sum_ip_i\ln(p_i)}$

  - **Maximum entropy** has uniform distribution.
  $\BC\t{minimize}_p&\sum_ip_i\ln(p_i)\\
  \t{subject to}&pвүҪ0,\v{1}^вҠӨp=1
  \EC$
  $\dg{вҮ’}L=\sum_ip_i\ln(p_i)+ОҪ(\v{1}^вҠӨp-1)$
  $\dg{вҮ’}g(ОҪ)=\inf_p(\sum_ip_i\ln(p_i)+ОҪ\sum_ip_i-ОҪ)$
  $=-ОҪ+\inf_p\sum_i(ОҪp_i+p_i\ln(p_i))$.
  Minimize
  $вҲӮ_{p_i}=ОҪ+\ln(p_i)+1=0$
  $\dg{вҮ’}p_i^*=e^{-ОҪ-1}\ вҲҖi$
  is a constant (uniform).

- **Differential entropy**: Mean value theorem says the area under PDF for the $i$-th bin has mean value $p(x_i)$ such that the probability is $p(x_i)О”$ satisfying $\sum_ip(x_i)О”=1$.
We quantize
$H_О”=-\sum_ip(x_i)О”[\ln p(x_i)+\ln О”]$
$=-\sum_ip(x_i)О”\ln p(x_i)-\ln О”$
$\dg{вҮ’}\lim\limits_{О”вҶ’0}H_О”=-вҲ«p(x)\ln p(x)\ dx+вҲһ$.
We discard the infinity term and use the first term as differential entropy $\red{\H[x]=-вҲ«p(x)\ln p(x)\ dx},\ xвҲҲв„қ^D$.

  - **Maximum entropy** for PDF $(Ој,Пғ^2)$:
  $\BC\t{maximize}_{p(x)}&-вҲ«p(x)\ln p(x)\ dx\\
  \t{subject to}&вҲ«p(x)\ dx=1,\ вҲ«xp(x)\ dx=Ој,\ вҲ«(x-Ој)^2p(x)\ dx=Пғ^2
  \EC$
  $L(p)=-вҲ«p\ln(p)\ dx+ОҪ_1(вҲ«p\ dx-1)+ОҪ_2(вҲ«xp\ dx-Ој)+ОҪ_3(вҲ«(x-Ој)^2p\ dx-Пғ^2)$
  $=-вҲ«(p\ln(p)+ОҪ_1p+ОҪ_2xp+ОҪ_3(x-Ој)^2p)\ dx+c_1$
  $\dg{вҮ’}вҲӮ_pL(p)=0$
  $\dg{вҮ’}вҲӮ_p(-p\ln(p)+ОҪ_1p+ОҪ_2xp+ОҪ_3(x-Ој)^2p)=0$
  $\dg{вҮ’}-\ln(p)-1+ОҪ_1+ОҪ_2x+ОҪ_3(x-Ој)^2=0$
  $\dg{вҮ’}p(x)=\e{-1+ОҪ_1+ОҪ_2x+ОҪ_3(x-Ој)^2}$.
  By integrating $p(x)$ against the three constraints, it turns out
  $\blue{p(x)=\/{1}{\sqrt{2ПҖ}Пғ}e^{-(x-Ој)^2/2Пғ^2}}$
  maximizes differential entropy with
  $\blue{\H[\Normal(x|Ој,Пғ^2)]=\/{1}{2}[1+\ln(2ПҖПғ^2)]}$.
  Differential entropy is negative when
  $Пғ^2< \/{1}{2ПҖe}$.

  - **Conditional entropy**: how much uncertainty is left?
  $\H[y|x=x_0]=-вҲ«p(y|x=x_0)\ln p(y|x=x_0)\ dy$
  $\dg{вҮ’}\H[y|x]=-\iint p(y|x=x_0)\ln p(y|x=x_0)\ dy\ p(x_0)\ dx_0$
  $\dg{вҮ’}\red{\H[y|x]=-\iint p(x,y)\ln p(y|x)\ dy\ dx}$

  Proof:
  $\H[x,y]=-\iint p(x,y)\ln p(x,y)\ dy\ dx$
  $=-\iint p(x,y)[\ln p(y|x)+\ln p(x)]\ dy\ dx$
  $=-\iint p(x,y)\ln p(y|x)\ dy\ dx-вҲ« p(x)\ln p(x)\ dx$.

**Kullback-Leibler Divergence**: Suppose $p(x)$ is the true probability of 'A-Z', but $q(x)$ is what the model assumes. KL divergence is the measure of inefficiency that arise from mismatch between the model $q(x)$ and reality $p(x)$.
$\red{\KL(p\|q)=вҲ«p(x)\ln\/{p(x)}{q(x)}\ dx}$
$=-вҲ«p(x)\ln q(x)\ dx-(-вҲ«p(x)\ln p(x)\ dx)$ matches reality $p$ to model $q$.

- $\blue{\KL(p\|q)\neq\KL(q\|p)}$:

  - **Forward KL**: $\KL(p\|q)=вҲ«p(x)\ln\/{p(x)}{q(x)}\ dx$ is expectation wrt true distribution $p(x)$.
  In regions where $q(x)вүҲ0$ and $p(x)>0$, KL blows up. Loss minimization forces $q(x)$ to stretch out and avoid zeros in all regions where true distribution has probability mass.
  This **zero-avoiding** behavior overestimates variance.
  If $p(x)$ is a 2-peak GMM, $q(x)$ tries to position its peak at the mean of the two modes (**mean seeking**).

  - **Reverse KL**: $\KL(q\|p)=вҲ«q(x)\ln\/{q(x)}{p(x)}\ dx$ wants to avoid placing probability mass where true distribution $p(x)вүҲ0$, therefore it is **zero-forcing** for most of its distribution thereby creating a narrow distribution.
  If $p(x)$ is a 2-peak GMM, $q(x)$ tries to position its peak at one mode and zero everywhere else (**mode seeking**).

- $\blue{\KL(p\|q)вүҘ0}$.
Proof: Let $f(y)=-\ln\/{q(x)}{p(x)}$.
Then by Jensen's inequality
$\KL(p\|q)=\E_p[f(y)]вүҘf(\E_p[y])$
$=-\ln вҲ«p(x)\/{q(x)}{p(x)}\ dx$
$=-\ln вҲ«q(x)\ dx$
$=0$.

- **Density estimation**: $\KL(p\|q)=-вҲ«p(x)\ln\/{q(x|Оё)}{p(x)}\ dx$
$=-вҲ«p(x)\ln q(x|Оё)\ dx+вҲ«p(x)\ln p(x)\ dx$.
We want to tweak $Оё$ to make our model $q(x|Оё)$ as close as possible to unknown true distribution $p(x)$.
Using empirical approximation: $вҲ«p(x)f(x)\ dxвүҲ\/{1}{N}\sum_if(x_i)$
$\dg{вҮ’}\red{\KL(p\|q)вүҲ\/{1}{N}\sum_i[-\ln q(x_i|Оё)+\ln p(x_i)]}$.
Therefore optimization
$\arg\min_Оё\KL(p\|q)$
$вүҲ\arg\min_Оё\sum_i-\ln q(x_i|Оё)$
$вүҲ\arg\max_Оё\sum_i\ln q(x_i|Оё)$ is exactly MLE and log-likelihood!

- **Mutual information**:
$\red{\I[x,y]:=\KL(p(x,y)\|p(x)p(y))}$
$=-\iint p(x,y)\ln\/{p(x)p(y)}{p(x,y)}\ dx\ dy$
$\dg{вҮ’}\red{\I[x,y]=\H[x]-\H[x|y]}$.
It is zero if variables are truly independent.
Remove from $\H[x]$ the initial information about $x$
by $\H[x|y]$ the remaining information about $x$ after learning $y$ gives the mutual information.

  - Proof: $\I[x,y]=-\iint p(x,y)\ln\/{p(x)p(y)}{p(x,y)}\ dx\ dy$
  $=-\iint p(x,y)[\ln p(x)+\ln p(y)-\ln p(x,y)]\ dx\ dy$
  $=-вҲ«p(x)\ln p(x)\ dx-вҲ«p(y)\ln p(y)\ dy+\iint p(x,y)\ln p(x,y)\ dx\ dy$
  $=\H[x]+\H[y]-\H[x,y]$

# Book1: 5 Decision Theory; 6 Information Theory

**Fundamentals**: Let $aвҲҲ\m{A}$ be possible actions, $hвҲҲ\m{H}$ be true underlying states, then $l(h,a)$ is the loss function incurred by taking action $a$ when the true underlying state is $h$.

- $\red{ПҒ(a|x):=\E_{p(h|x)}[l(h,a)]=\sum_{hвҲҲ\m{H}}l(h,a)p(h|x)}$ is the expected loss for taking action $a$ given observation $x$ and posterior $p(h|x)$.
The best course of action is **Bayes estimator**
$\red{ПҖ^*(x)=\arg\min_{aвҲҲ\m{A}}\E_{p(h|x)}[l(h,a)]}$.

**Entropy**: If $p$ has high entropy and a data source generates $x_iвҲјp$, then $x_i$ is highly unpredictable. Observing actual data provides a lot of new information. Therefore high entropy = high information content.

- **Discrete**: $\red{\H(X)=-\sum_xp(X=x)\log_2 p(X=x)=-E_X[\log_2 p(X)]}$.
Calculating true entropy of a distribution with a massive number of states (e.g., dictionary) requires immense amounts of data.

  - $\BC

  \t{max entropy}&\H(X)вүӨ\log(K)&\t{uniform }p(X=k)=\/{1}{K}
  \EC$.
  Proof (max): Let $u(x)=\/{1}{K}$ be uniform PDF.
  Then $\KL(p\|u)=\sum_xp(x)\ln\/{p(x)}{u(x)}$
  $=\sum_xp(x)\ln p(x)-\sum_xp(x)\ln u(x)$
  $=-\H(x)-\ln(\/{1}{K})вүҘ0$.

  - $\H(p)$ means entropy of the $p$ probability vector $\red{\H(p)=\sum_kp_k\ln p_k}$

- **Cross entropy**: $\red{\H(p,q)=-\sum_kp_k\ln q_k}$

- **Joint entropy**: $\red{\H(X,Y)=-\sum_{x,y}p(x,y)\ln_2 p(x,y)}$ is simultaneous uncertainty across multiple variables.

  - $\BC
  \t{min joint entropy}&\H(X,Y)вүҘ\max\{\H(X),\H(Y)\}вүҘ0&\t{joint uncertainty вүҘ most uncertain element}\\
  \t{max joint entropy}&\H(X,Y)вүӨ\H(X)+\H(Y)&\t{equal when $X$ and $Y$ are independent}
  \EC$

  - For states $nвҲҲ\{1,...,8\}$ and variables $X=I(n\%2=0)$ and $Y=I(n\t{ prime})$, then $\E[X]=\E[Y]=\/{1}{2}$ and $\H(X)+\H(Y)=2$ bits.

  Variables $X,Y$ are not independent: knowing one gives information about the other. Joint entropy is lower than the sum.

- **Conditional entropy**: $\red{\H(Y|X)=\E_{p(x)}[\H(p(Y|X))]=-\sum_xp(x)\sum_yp(y|x)\ln p(y|x)}$
$=-\sum_{x,y}p(x,y)\ln \/{p(x,y)}{p(x)}$
$=-\sum_{x,y}p(x,y)\ln p(x,y)+\sum_{x,y}p(x,y)\ln p(x)$
$\dg{вҮ’}\red{\H(Y|X)=\H(X,Y)-\H(X)}$

  - **Chain rule** for entropy: $\blue{H(X_1,...,X_n)=\sum_iH(i|X_1,...,X_{i-1})}$

  - $\blue{\H(Y|X)вүӨ\H(Y)}$

- **Perplexity** of LLM with $K$ words is $\t{PP}(p)=2^{\H(p)}$ is the branching factor in predicting the next word in a sentence.
$\t{PP}(p_\m{D},p)=2^{\H(p_\m{D},p)}$ evaluates how well model $p$ predicts distribution of words in dataset $p_\m{D}$.

- **Differential entropy**: $\red{\H(X)=-вҲ«p(x)\ln p(x)\ dx=\E_p[\ln p(x)]}$

  - $\H(\Normal(Ој,ОЈ))=\/{D}{2}\ln(2ПҖ)+\/{1}{2}\ln\det(ОЈ)+\/{1}{2}\E[(x-Ој)^вҠӨОЈ^{-1}(x-Ој)]$
  where
  $\E[(x-Ој)^вҠӨОЈ^{-1}(x-Ој)]$
  $=\E[\tr((x-Ој)^вҠӨОЈ^{-1}(x-Ој))]$
  $=\tr(ОЈ^{-1}\E[(x-Ој)(x-Ој)^вҠӨ])$
  $=\tr(ОЈ^{-1}ОЈ)$
  $=D$.
  Then
  $\H(\Normal(Ој,ОЈ))=\/{D}{2}(\ln(2ПҖ)+1)+\/{1}{2}\ln\det(ОЈ)$
  $\dg{вҮ’}\red{\H(\Normal(Ој,ОЈ))=\/{1}{2}\ln((2ПҖe)^D\det(ОЈ))}$

  - We often calculate differential entropy by binning the distribution into a histogram from $\min(x)$ to $\max(x)$.

**Kullback-Leibler divergence**: $\red{\KL(p\|q)=\sum_kp_k\ln\/{p_k}{q_k}=-\H(p)+\H(p,q)}$ and $\KL(p\|q)=вҲ«p(x)\ln\/{p(x)}{q(x)}\ dx$.
Distance requires Abelian equality and triangle inequality.
Divergence only requires non-negativity and zero with identical operands.

- **Gaussians**: $\KL(\Normal(x|Ој_1,ОЈ_1)\|\Normal(x|Ој_2,ОЈ_2))$
$=\/{1}{2}\E_{Ој_1,ОЈ_1}[\ln\/{\det(ОЈ_2)}{\det(ОЈ_1)}-\tr((x-Ој_1)^вҠӨОЈ_1^{-1}(x-Ој_1))$

- **Minimum $\KL(p_\m{D}\|q)$ = maximum likelihood**: Let $p_\m{D}(x)=\/{1}{N}\sum_iI(x-x_i)$ gives a histogram counter representing empirical distribution.
By delta-dirac convolution,
$q^*=\arg\min_q\KL(p_\m{D}\|q)$

  - **Cross entropy**: $\KL(p\|q)=\sum_yp(y)\ln\/{p(y)}{q(y)}=-\H(p)+\H(p,q)$
  $\dg{вҮ’}\red{\arg\min_q\KL(p\|q)=\arg\min_q\H(p,q)}$ heavily penalizes wrong predictions for rare events. Brier score is an alternative: $\/{1}{N}\sum_y(q(y)-p(y))^2$

  $\H(I(y=c),q)=-\sum_yI(y=c)\ln q(y)$
  $=-\ln q(c)$.

  \EC$

**Non-parametric**: Histogram with $K$ of $N$ points falling into bin $i$ that has bin size $V$ has probability $\/{K}{N}$ and continuous density $p=\/{K}{NV}$.
Suppose points are drawn from unknown true density $p(x)$ we approximate $P=вҲ«_Vp(x)\ dxвүҲp(x)V$ for small $V$ in the neighborhood of $x$ following binomial expectation $K=NP$.
Therefore $\red{p(x)=\/{K}{NV}}$.
Fixing $K$ to determine $V$ is K-nearest-neighbors.
Fixing $V$ to determine $K$ is kernel method.

- **Kernel density estimation** is histogram method with overlapping bins positioned at data points (Parzen window) instead of a fixed grid. Let **kernel function** be unit hypercube
$\red{k(u)=\BC 1&|u_i|вүӨ1/2\\0&\t{otherwise}\EC}$ (Boxcar kernel), and each bin is a cube of size $V=h^D$.
Then
$K=\sum_ik(\/{x-x_i}{h})$ is the number of points inside a bin centered at $x$,
and is also number of bins centered at $x_i$ that contain $x$.
The density estimate is
$\blue{p(x)=\/{1}{N}\sum_i\/{1}{h^D}k(\/{x-x_i}{h})}$.

  - **Gaussian kernel**: $\red{k(u)=\Normal(0,1)}$ then $\blue{p(x)=\/{1}{N}\sum_i\/{1}{(2ПҖh^2)^{1/2}}\e{-\/{\n{x-x_i}^2}{2h^2}}}$

  - Kernel functions $k(u)$ satisfy $k(u)вүҘ0,\ вҲ«k(u)\ du=1$.

  - At dense clusters, small bins stack and describe a contour for the probability density. However sparse regions are overfit to noisy spikes in an ocean of zeros that mathematically assume zero probability of finding new data points. Using large bin size would wash out details of dense clusters.

- **K-nearest-neighbor**: In density estimation, grow a sphere at $x$ until it contains $K$ points, then probability density $p(x)=\/{K}{NV}$ is inversely proportional to sphere size $V$.

  - Suppose $N_k$ of $N$ points are in class $C_k$.
  To classify point $x$, we grow a sphere centered at $x$ until it contains $K$ points irrespective of class and $K_k$ of which belong to class $C_k$.
  $\BC \t{likelihood}&\t{marginal}&\t{prior}\\
  p(x|C_k)=\/{K_k}{N_kV}&p(x)=\/{K}{NV}&p(C_k)=\/{N_k}{N}
  \EC$
  $\dg{вҮ’}p(C_k|x)=\/{p(x|C_k)p(C_k)}{p(x)}=\/{K_k}{K}$ posterior.
  Assign to majority class.

**Bayesian model selection**: Suppose we have models $\{M_1,...,M_L\}$ and training data $\m{D}$. Then posterior distribution over models is
$\purple{p(M_i|\m{D})}вҲқp(M_i)\yello{p(\m{D}|M_i)}$ and prediction distribution is
$p(t_k|x_k,\m{D})=\sum_ip(t_k|x_k,M_i,\m{D})\purple{p(M_i|\m{D})}$.
Model selection chooses the model with highest posterior $\purple{p(M_i|\m{D})}$, which is equivalent to choosing highest evidence $\yello{p(\m{D}|M_i)}$.

- Model evidence is
$\yello{p(\m{D}|M_i)}=вҲ«p(\m{D}|w,M_i)p(w|M_i)\ dw$.
To approximate the model evidence integral, let
$О”w_\t{posterior}< О”w_\t{prior}$ be widths of the modal peaks.

  $p(\m{D})О”w_\t{prior}вүҲp(\m{D}|w_\t{MAP})О”w_\t{posterior}$
  $\dg{вҮ’}p(\m{D})=вҲ«p(\m{D}|w)p(w)\ dwвүҲp(\m{D}|w_\t{MAP})\/{О”w_\t{posterior}}{О”w_\t{prior}}$
  $\dg{вҮ’}\ln p(\m{D})=\ln p(\m{D}|w_\t{MAP})+\ln\/{О”w_\t{posterior}}{О”w_\t{prior}}$.
  Although data fit improves with parameters $\ln p(\m{D})< \ln p(\m{D}|w_\t{MAP})$,
  the second term $\ln\/{О”w_\t{posterior}}{О”w_\t{prior}}< 0$ is a penalty on complexity.

  - Let $wвҲҲв„қ^M$ then $\ln p(\m{D})=\ln p(\m{D}|w_\t{MAP})+M\ln\/{О”w_\t{posterior}}{О”w_\t{prior}}$
  and the complexity penalty is greater.

Assuming $M_1$ is the true distribution and $M_2$ is an incorrect model, then we can average the Bayes factor across all possible datasets $\m{D}$ to get the KL divergence $\KL(\yello{p(\m{D}|M_1)}\|\yello{p(\m{D}|M_2)})=вҲ«\ln\/{\yello{p(\m{D}|M_1)}}{\yello{p(\m{D}|M_2)}}\yello{p(\m{D}|M_1)}\ d\m{D}$

**Empirical Bayes**: Suppose $\BC
p(t|w,ОІ)=\prod_i\Normal(t_i|П•(x_i)^вҠӨw,ОІ^{-1})\\
p(w|Оұ)=\Normal(0,Оұ^{-1}I)\\
p(w|t)=\Normal(w|m_N,S_N)
\EC$
$\dg{вҮ’}\BC
S_N=(ОұI+ОІОҰ^вҠӨОҰ)^{-1}\\
m_N=ОІS_NОҰ^вҠӨt=ОІ(ОұI+ОІОҰ^вҠӨОҰ)^{-1}ОҰ^вҠӨt
\EC$. Full Bayesian treatment requires integration over all hyperparameters
$p(t_k|x_k,t)=\iiint p(t_k|x_k,w,ОІ)p(w|t,Оұ,ОІ)\purple{p(Оұ,ОІ|t)}\ dw\ dОұ\ dОІ$
$вүҲp(t_k|x_k,t,\hat{Оұ},\hat{ОІ})=вҲ«p(t_k|w,\hat{ОІ})p(w|t,\hat{Оұ},\hat{ОІ})\ dw$ where $\hat{Оұ},\hat{ОІ}=\arg\max_{Оұ,ОІ} \purple{p(Оұ,ОІ|t)}$ are MAP estimators.

- **Posterior** $\purple{p(Оұ,ОІ|t)}вҲқ\brown{p(t|Оұ,ОІ)}p(Оұ,ОІ)$. If we assume prior $p(Оұ,ОІ)$ is relatively flat, then $\hat{Оұ},\hat{ОІ}=\arg\max_{Оұ,ОІ}\brown{p(t|Оұ,ОІ)}$ is maximizing the marginal likelihood (evidence).

- **Marginal likelihood (evidence)**: $\green{\ln \brown{p(t|Оұ,ОІ)}=\/{M}{2}\ln Оұ+\/{N}{2}\ln ОІ-\/{1}{2}E(m_N)+\/{1}{2}\ln\det(S_N)-\/{N}{2}\ln(2ПҖ)}$.
Solution by completing squares
$\BA
\dg{вҮ’}&\brown{p(t|Оұ,ОІ)}=вҲ«p(t,w|Оұ,ОІ)\ dw=вҲ«p(t|w,ОІ)p(w|Оұ)\ dw=вҲ«(\/{ОІ}{2ПҖ})^{N/2}(\/{Оұ}{2ПҖ})^{M/2}\e{-\/{1}{2}E(w)}\ dw\\
\dg{вҮ’}&\blue{E(w)=ОІ\n{t-ОҰw}^2+Оұ\n{w}^2}=ОІ(w^вҠӨОҰ^вҠӨОҰw-2w^вҠӨОҰ^вҠӨt+t^вҠӨt)+Оұw^вҠӨw=w^вҠӨS_N^{-1}w-2w^вҠӨS_N^{-1}m_N+ОІt^вҠӨt\\
&=w^вҠӨS_N^{-1}w-2w^вҠӨS_N^{-1}m_N+ОІt^вҠӨt+m_N^вҠӨS_N^{-1}m_N-m_N^вҠӨS_N^{-1}m_N=\yello{(w-m_N)^вҠӨS_N^{-1}(w-m_N)}+\green{ОІt^вҠӨt-m_N^вҠӨS_N^{-1}m_N}\\
\dg{вҮ’}&\green{ОІt^вҠӨt-m_N^вҠӨS_N^{-1}m_N}=ОІt^вҠӨt-2m_N^вҠӨ(S_N^{-1}m_N)+m_N^вҠӨS_N^{-1}m_N=ОІt^вҠӨt-2m_N^вҠӨ(ОІОҰ^вҠӨt)+m_N^вҠӨ(ОұI+ОІОҰ^вҠӨОҰ)m_N\\
&=ОІ(t^вҠӨt-2m_N^вҠӨОҰ^вҠӨt+m_N^вҠӨОҰ^вҠӨОҰm_N)+Оұm_N^вҠӨm_N=ОІ\n{t-ОҰm_N}^2+Оұ\n{m_N}^2=\green{E(m_N)}\\
\dg{вҮ’}&\brown{p(t|Оұ,ОІ)}=вҲ«p(t,w|Оұ,ОІ)\ dw=(\/{ОІ}{2ПҖ})^{N/2}(\/{Оұ}{2ПҖ})^{M/2}\e{-\/{1}{2}\green{E(m_N)}}вҲ«\e{-\/{1}{2}\yello{(w-m_N)^вҠӨS_N^{-1}(w-m_N)}}\ dw\\
&=(\/{ОІ}{2ПҖ})^{N/2}(\/{Оұ}{2ПҖ})^{M/2}(2ПҖ)^{M/2}\det(S_N)^{1/2}\e{-\/{1}{2}\green{E(m_N)}}
\EA$
Solution by Adam-Eve convolution
$\BA
\dg{вҮ’}&\brown{p(t|Оұ,ОІ)}=вҲ«p(t,w|Оұ,ОІ)\ dw=вҲ«p(t|w,ОІ)p(w|Оұ)\ dw=вҲ«\Normal(t|ОҰw,ОІ^{-1}I)\Normal(w|0,Оұ^{-1}I)\ dw=\Normal(t|\E[t],\yello{\V[t]})\\
&\dg{вҮ’}\E[t]=\E[\E[t|w]]=\E[ОҰw]=0\quad\dg{вҮ’}\V[t]=\E[\V[t|w]]+\V[\E[t|w]]=\E[ОІ^{-1}I]+\V[ОҰw]\dg{вҮ’}\yello{\V[t]=ОІ^{-1}I+Оұ^{-1}ОҰОҰ^вҠӨ}\\
\dg{вҮ’}&\t{Find $\yello{\V[t]}^{-1}$ by carefully mapping to Woodbury identity }\yello{\V[t]}=A_{22}+A_{21}A_{11}^{-1}A_{12}\dg{вҮ’}A=\BMОұI&ОҰ^вҠӨ\\ОҰ&ОІ^{-1}I \EM\\
&\dg{вҮ’}(A_{22}+A_{21}A_{11}^{-1}A_{12})^{-1}=A_{22}^{-1}-A_{22}^{-1}A_{21}(A_{11}+A_{12}A_{22}^{-1}A_{21})^{-1}A_{12}A_{22}^{-1}\dg{вҮ’}\yello{\V[t]}^{-1}=ОІI-ОІ^2ОҰ(ОұI+ОІОҰ^вҠӨОҰ)ОҰ^вҠӨ\\
\dg{вҮ’}&t^вҠӨ\yello{\V[t]}^{-1}t=t^вҠӨ[ОІI-ОІ^2ОҰ(ОұI+ОІОҰ^вҠӨОҰ)ОҰ^вҠӨ]t=ОІt^вҠӨt-ОІ^2t^вҠӨОҰS_N^{-1}ОҰ^вҠӨt=\green{ОІt^вҠӨt-m_N^вҠӨS_N^{-1}m_N}\\
&=ОІt^вҠӨt-2m_N^вҠӨ(S_N^{-1}m_N)+m_N^вҠӨS_N^{-1}m_N=ОІt^вҠӨt-2m_N^вҠӨ(ОІОҰ^вҠӨt)+m_N^вҠӨ(ОұI+ОІОҰ^вҠӨОҰ)m_N\\
&=ОІ(t^вҠӨt-2m_N^вҠӨОҰ^вҠӨt+m_N^вҠӨОҰ^вҠӨОҰm_N)+Оұm_N^вҠӨm_N=ОІ\n{t-ОҰm_N}^2+Оұ\n{m_N}^2=\green{E(m_N)}\\
\dg{вҮ’}&\det(\yello{\V[t]})=\det(ОІ^{-1}I+Оұ^{-1}ОҰОҰ^вҠӨ)=\det(ОІ^{-1}I_N)\det(Оұ^{-1}I_M)\det(ОұI_N+ОІОҰОҰ^вҠӨ)=ОІ^{-N}Оұ^{-M}\det(ОұI_M+ОІОҰ^вҠӨОҰ)\\
&=ОІ^{-N}Оұ^{-M}\det(S_N^{-1})\\
\dg{вҮ’}&\brown{p(t|Оұ,ОІ)}=\Normal(t|0,\yello{\V[t]})=\/{1}{(2ПҖ)^{N/2}}\det(\yello{\V[t]})^{-1/2}\e{-\/{1}{2}\green{E(m_N)}}=\/{1}{(2ПҖ)^{N/2}}Оұ^{M/2}ОІ^{N/2}\det(S_N)^{1/2}\e{-\/{1}{2}\green{E(m_N)}}
\EA$

- **Maximizing evidence**: Let $(ОІОҰ^вҠӨОҰ)u_i=О»_iu_i$.
$\BA
\dg{вҮ’}&\ln\det(S_N)=-\ln\det(S_N^{-1})=-\ln\prod_i(Оұ+О»_i)=-\sum_i\ln(Оұ+О»_i)\dg{вҮ’}вҲӮ_Оұ\ln\det(S_N)=-\sum_i\/{1}{Оұ+О»_i}\\
\dg{вҮ’}&вҲҮ_{m_N}E(m_N)=0\dg{вҮ’}E(m_N)=ОІ\n{t-ОҰm_N}^2+Оұ\n{m_N}^2\dg{вҮ’}вҲӮ_ОұE(m_N)=вҲӮ_ОұE+вҲҮ_{m_N}E(m_N)^вҠӨвҲӮ_Оұm_N=\n{m_N}^2\\
\dg{вҮ’}&вҲӮ_Оұ\ln \brown{p(t|Оұ,ОІ)}=\/{M}{2Оұ}-\/{1}{2}\n{m_N}^2-\/{1}{2}\sum_i\/{1}{Оұ+О»_i}\dg{вҮ’}Оұm_N^вҠӨm_N=M-\sum_i\/{Оұ}{Оұ+О»_i}=\sum_i(1-\/{Оұ}{Оұ+О»_i})=\sum_i\/{О»_i}{Оұ+О»_i}\\
\dg{вҮ’}&\blue{Оівүң\sum_i\/{О»_i}{Оұ+О»_i}=Оұm_N^вҠӨm_N}\dg{вҮ’}\green{Оұ^{(k+1)}=\/{Оі^{(k)}}{m_N^{(k)вҠӨ}m_N^{(k)}}}\\
\dg{вҮ’}&(ОІОҰ^вҠӨОҰ)u_i=О»_iu_i\dg{вҮ’}(ОҰ^вҠӨОҰ)u_i=Оҙ_iu_i\dg{вҮ’}вҲӮ_ОІО»_i=Оҙ_i=\/{О»_i}{ОІ}\dg{вҮ’}вҲӮ_ОІ\ln\det(S_N)=-\sum_i\/{1}{Оұ+О»_i}\/{О»_i}{ОІ}=-\/{Оі}{ОІ}\\
\dg{вҮ’}&вҲӮ_ОІE(m_N)=вҲӮ_ОІE+вҲҮ_{m_N}E(m_N)^вҠӨвҲӮ_ОІm_N=\n{t-ОҰm_N}^2\\
\dg{вҮ’}&вҲӮ_ОІ\ln \brown{p(t|Оұ,ОІ)}=\/{N}{2ОІ}-\/{1}{2}\n{t-ОҰm_N}^2-\/{Оі}{2ОІ}=0\dg{вҮ’}\green{\/{1}{ОІ^{(k+1)}}=\/{1}{N-Оі^{(k)}}\n{t-ОҰm_N^{(k)}}^2}
\EA$

- **Bayesian regularization**: $\red{w_\t{MAP}=\arg\max_wp(w|t)=m_N=ОІ(ОұI+ОІОҰ^вҠӨОҰ)^{-1}ОҰ^вҠӨt}$ the mode peaks at its mean
$\BC
\dg{вҮ’}&вҲҮ_w\ln p(w|t)=вҲҮ_w[-\/{Оұ}{2}\n{w}^2-\/{ОІ}{2}\n{t-ОҰw}^2]=-Оұw^вҠӨ-ОІ(t-ОҰw)^вҠӨ(-ОҰ)=-Оұw^вҠӨ+ОІt^вҠӨОҰ-ОІw^вҠӨОҰ^вҠӨОҰ=0\\
\dg{вҮ’}&w^вҠӨ(ОұI+ОІОҰ^вҠӨОҰ)=ОІt^вҠӨОҰ\dg{вҮ’}w^*=ОІ(ОұI+ОІОҰ^вҠӨОҰ)^{-1}ОҰ^вҠӨt
\EC$
  
  - Non-Bayesian regularized $(ОҰ^вҠӨОҰ+О»I)^{-1}ОҰ^вҠӨt$ finds adhoc $О»=\/{Оұ}{ОІ}$ by cross-validation, but Bayesian regularized $w_\t{MAP}$ finds $Оұ,ОІ$ from training data by maximizing model evidence.

- **Well-determined parameters**: PCA of empirical covariance $\purple{ОІОҰ^вҠӨОҰ}=UОӣU^вҠӨ$ gives anisotropic shrinkage along each principal axis
$\red{u_i^вҠӨw_\t{MAP}=\/{О»_i}{Оұ+О»_i}u_i^вҠӨw_\t{ML}}$.
Well determined parameters have high $О»_i$ to pull $u_i^вҠӨw_\t{MAP}вҶ’u_i^вҠӨw_\t{ML}$ vs $u_i^вҠӨw_\t{MAP}вҶ’0$.
$\BC
\dg{вҮ’}&\yello{w_\t{ML}=(ОҰ^вҠӨОҰ)^{-1}ОҰ^вҠӨt}\quad\dg{вҮ’}\green{w_\t{MAP}=ОІ(ОұI+ОІОҰ^вҠӨОҰ)^{-1}ОҰ^вҠӨt}=(ОұI+ОІОҰ^вҠӨОҰ)^{-1}(\purple{ОІОҰ^вҠӨОҰ})\yello{w_\t{ML}}\\
\dg{вҮ’}&u_i^вҠӨ(ОұI+ОІОҰ^вҠӨОҰ)^{-1}(ОІОҰ^вҠӨОҰ)=[(ОұI+ОІОҰ^вҠӨОҰ)^{-1}(\purple{ОІОҰ^вҠӨОҰ})]^вҠӨu_i=(\purple{ОІОҰ^вҠӨОҰ})\/{1}{Оұ+О»_i}u_i=\/{О»_i}{Оұ+О»_i}u_i\\
\dg{вҮ’}&u_i^вҠӨ\green{w_\t{MAP}}=u_i^вҠӨ(ОұI+ОІОҰ^вҠӨОҰ)^{-1}(\purple{ОІОҰ^вҠӨОҰ})w_\t{ML}=\/{О»_i}{Оұ+О»_i}u_i^вҠӨ\yello{w_\t{ML}}\dg{вҮ’}u_i^вҠӨ\green{w_\t{MAP}}=\red{\/{О»_i}{Оұ+О»_i}}u_i^вҠӨ\yello{w_\t{ML}}
\EC$

  - **Effective number of parameters** $\red{Оі=\sum_i\/{О»_i}{Оұ+О»_i}}$ is a count of well-determined parameters, or degrees of freedom used to determine parameters.
  The data has enough strong eigenvalues to determine $Оі$ of $M$ parameters.

  - $Nвү«M\dg{вҮ’}ОҰ^вҠӨОҰ=\sum_iП•(x_i)П•(x_i)^вҠӨ$ has high eigenvalues: $Оі=M,\ Оұ=\/{M}{\n{m_N}^2},\ ОІ=\/{N}{\n{t-ОҰm_N}^2)}$

# ESL: 3 Linear Models for Regression

**Setup**: $y_i=\E[y_i|x_i]+Пө=ОІ_0+x_i^вҠӨОІ_{1..p}+Пө$ from training data $xвҲҲв„қ^{NГ—p},yвҲҲв„қ^N$ and $X=\BM\v{1}&x\EMвҲҲв„қ^{NГ—(p+1)}$.
OLS estimate $\BA
\hat{ОІ}=\arg\min_ОІ\t{RSS}(ОІ)
=\arg\min_ОІ\n{y-ОІ_0\v{1}-xОІ_{1..p}}^2
=\arg\min_ОІ\n{y-XОІ}^2
\dg{вҮ’}\red{\hat{ОІ}=(X^вҠӨX)^{-1}X^вҠӨy}вҲҲв„қ^{p+1}
\EA$.

- **Degrees of freedom**: OLS estimates $\hat{ОІ}вҲҲв„қ^{p+1}$ with $p+1$ degrees of freedom.
Statistical definition of degrees of freedom is how sensitive the fitted values $\hat{y}_i$ are wrt changes in observed values $y_i$.
Here $\hat{y}=Hy\dg{вҮ’}\hat{y}_i=H_iy$ where $H=X(X^вҠӨX)^{-1}X^вҠӨвҲҲв„қ^{NГ—N}$ is the hat matrix.

  - $\BA
  \red{\t{df}(\hat{y})=\sum_{i=1}^N\/{\V[\hat{y}_i,y_i]}{\V[y_i]}}
  =\sum_i\/{\V[\sum_jH_{ij}y_j,y_i]}{\V[y_i]}
  =\sum_i\/{\V[H_{ii}y_i,y_i]}{\V[y_i]}
  =\sum_iH_{ii}
  \dg{вҮ’}\red{\t{df}(\hat{y})=\tr(H)}
  \EA$.

- Assume $\V[y_i]=Пғ^2$ and uncorrelated.
Then $\red{\hat{ОІ}вҲј\Normal(ОІ,(X^вҠӨX)^{-1}Пғ^2)}$ and $\V[\hat{ОІ}_j]=v_j\hat{Пғ}^2$ where $v_j=[(X^вҠӨX)^{-1}]_{jj}$.
$\BA
\dg{вҮ’}&\V[\hat{ОІ}]=\V[(X^вҠӨX)^{-1}X^вҠӨy]=[(X^вҠӨX)^{-1}X^вҠӨ]\V[y][(X^вҠӨX)^{-1}X^вҠӨ]^вҠӨ=(X^вҠӨX)^{-1}X^вҠӨПғ^2IX(X^вҠӨX)^{-1}=(X^вҠӨX)^{-1}Пғ^2
\EA$

  - $\hat{Пғ}^2=\/{1}{N-p-1}\sum_i(y_i-\hat{y}_i)^2$
  $\dg{вҮ’}\blue{\/{(N-p-1)\hat{Пғ}^2}{Пғ^2}вҲјПҮ_{N-p-1}^2}$.

  - Quadratic form: $Z^2=\/{(x-Ој)^2}{Пғ^2}вҲјПҮ_1^2$
  $\dg{вҮ’}\blue{(\hat{ОІ}-ОІ)^вҠӨОЈ^{-1}(\hat{ОІ}-ОІ)вҲјПҮ_{p+1}^2}$

  - $H_0:ОІ_j=0$ Z-test $\blue{z_j=\/{\hat{ОІ}}{\hat{Пғ}\sqrt{v_j}}}$.

- **Gauss-Markov theorem**: Linear estimator $\hat{Оё}=c^вҠӨy$ is a linear combination of observed response $y$.  LS estimator $\hat{Оё}=a^вҠӨ\hat{ОІ}=a^вҠӨ(x^вҠӨx)^{-1}x^вҠӨy$ has the smallest variance among linear unbiased estimators (BLUE).

  - Quality of estimator should be measured by estimator MSE, which measures how close estimates are to true parameter values
  $\t{MSE}(\hat{Оё})=\E[(\hat{Оё}-Оё)^2]$
  $=\V[\hat{Оё}]+(\E[\hat{Оё}]-Оё)^2$
  even if biased $\E[\hat{Оё}]-Оё\neq0$.

**Gram-Schmidt multiple regression**: $p=1$ no intercept regressing $y$ on $x$ to get $\hat{ОІ}=\/{x^вҠӨy}{x^вҠӨx}$ with residual $r=y-x\hat{ОІ}$.

- Consider $\BM 1&x_1\\\vdots&\vdots\\1&x_N\EM$. Regressing $x$ on $z_0=\v{1}$ to get $\bar{x}=\/{\v{1}^вҠӨx}{\v{1}^вҠӨ\v{1}}$ with residual $z=x-\v{1}\bar{x}$ centering $x$.
Now $x=z+\v{1}\bar{x}$ is the sum of two orthogonal components satisfying $(x-\bar{x}\v{1})^вҠӨ\v{1}=0$.
Then $\hat{ОІ}_1=\/{z^вҠӨy}{z^вҠӨz}$

- Consider $\BM 1&x_{11}&x_{21}\\\vdots&\vdots&\vdots\\1&x_{1N}&x_{2N}\EM$.
$\BC
\t{Regress $x_1$ on $\yello{z_0}=\green{v_0}=\v{1}$}&\hat{Оі}_{01}=\/{\yello{z_0}^вҠӨx_1}{\yello{z_0}^вҠӨ\yello{z_0}}=\bar{x}_1&\yello{z_1}=x_1-\hat{Оі}_{01}\yello{z_0}\\
\t{Regress $x_2$ on $\yello{z_0},\yello{z_1}$}&\hat{Оі}_{02}=\/{\yello{z_0}^вҠӨx_2}{\yello{z_0}^вҠӨ\yello{z_0}}=\bar{x}_2&\green{v_1}=x_2-\hat{Оі}_{02}\green{v_0}\\
&\hat{Оі}_{12}=\/{\yello{z_1}^вҠӨx_2}{\yello{z_1}^вҠӨ\yello{z_1}}&\yello{z_2}=x_2-\hat{Оі}_{02}\yello{z_0}-\hat{Оі}_{12}\yello{z_1}&\dg{вҮ’}\hat{ОІ}_2=\/{\yello{z_2}^вҠӨy}{\yello{z_2}^вҠӨ\yello{z_2}}\\
\t{Regress $x_1$ on $\green{v_1}$}&\hat{Оі}_{11}=\/{\green{v_1}^вҠӨx_1}{\green{v_1}^вҠӨ\green{v_1}}&\green{v_2}=x_1-\hat{Оі}_{01}\green{v_0}-\hat{Оі}_{11}\green{v_1}&\dg{вҮ’}\hat{ОІ}_1=\/{\green{v_2}^вҠӨy}{\green{v_2}^вҠӨ\green{v_2}}
\EC$

  - Rearrange to have $x_1$ as the last column to find $\hat{ОІ}_1$.
  Similarly have $\v{1}$ as the last column to find $\hat{ОІ}_0$.
  
  - **Deflation**: Projecting $y$ onto $z_p$ gives $\hat{ОІ}_p$ because $z_p$ represents the only unique independent information in $x_p$ after removing influence of $x_0,...,x_{p-1}$ by deflating away $z_0,...,z_{p-1}$ from $x_p$.

- Consider $xвҲҲв„қ^p$. $\BC
1.&z_0=x_0=\v{1}\\
2.&\t{for }j=1,2,...,p\\
&\t{Regress }x_j\t{ on }z_0,...,z_{j-1}\t{ to produce }\hat{Оі}_{lj}=\/{z_l^вҠӨx_j}{z_l^вҠӨz_l}, z_j=x_j-\sum_{k=0}^{j-1}\hat{Оі}_{kj}z_k\\
3.&\red{\hat{ОІ}_p=\/{z_p^вҠӨy}{z_p^вҠӨz_p}}\t{ where }\V[\hat{ОІ}_p]=\/{Пғ^2}{z_p^вҠӨz_p}
\EC$

- Gram-Schmidt orthogonalization of $X=\BM\v{1}&x\EMвҲҲв„қ^{NГ—(p+1)}$ finds $Z=\BM z_0&...&z_p\EMвҲҲв„қ^{NГ—(p+1)}$ as orthogonal basis for the column space of $X=ZО“$.
Normalizing $Z$ to have unit length columns using diagonal $D$ yields QR decomposition $X=(ZD^{-1})(DО“)=QR$ where $QвҲҲв„қ^{NГ—(p+1)}$ and $Q^вҠӨQ=I$.
Then $\red{\hat{ОІ}=R^{-1}Q^вҠӨy}$ and $\red{\hat{y}=QQ^вҠӨy}$.

**Shrinkage**: shrinkage methods reduce variance by shrinking the coefficients toward zero. Because regression is not unit invariant, the model will pull intercept toward zero if it is included in ridge penalty thus resulting in different slopes for Celsius vs Kelvin. Standard practice is to center $x$ and $y$ so that $\bar{x}=\bar{y}=0$ and $\hat{ОІ}_0=0$. Let $XвҲҲв„қ^{NГ—p}$ and $\hat{ОІ}вҲҲв„қ^p$.

- **Ridge**: $\red{\hat{ОІ}^\t{ridge}=\arg\min_ОІ(\n{y-XОІ}^2+О»\n{ОІ}^2)}$
$=\blue{(X^вҠӨX+О»I)^{-1}X^вҠӨy}$. Let $X=UОЈV^вҠӨ$ with thin $UвҲҲв„қ^{NГ—D}$. Then $X^вҠӨX=VОЈ^2V^вҠӨ$
$\dg{вҮ’}\BC
X\hat{ОІ}^\t{ls}=X(X^вҠӨX)^{-1}X^вҠӨy=UОЈV^вҠӨ(VОЈ^2V^вҠӨ)^{-1}VОЈ^вҠӨU^вҠӨy=UU^вҠӨy\dg{вҮ’}\red{\hat{y}^\t{ls}=\sum_{j=1}^pu_ju_j^вҠӨy}\\
X\hat{ОІ}^\t{ridge}=X(X^вҠӨX+О»I)^{-1}X^вҠӨy=UОЈ(ОЈ^2+О»I)^{-1}ОЈ^вҠӨU^вҠӨy\dg{вҮ’}\red{\hat{y}^\t{ridge}=\sum_{j=1}^pu_j\/{Пғ_j^2}{Пғ_j^2+О»}u_j^вҠӨy}
\EC$

  - **Anisotropic**: OLS regression is isotropic and uniform along all axes. $UU^вҠӨ$ is projection matrix onto the column space of $X$. Along each axis $u_ju_j^вҠӨy$ is projection of $y$ onto unit vector $u_j$. Ridge regression is anisotropic and axis-dependent. Along each axis $\/{Пғ_j^2}{Пғ_j^2+О»}u_ju_j^вҠӨy$ applies shrinkage according to singular values.

  - **PCA**: Let $X^вҠӨX=VОЈ^2V^вҠӨ$, then code $Z_M=XV_M=U_MОЈ_MвҲҲв„қ^{NГ—M}$. The first principal component is the first column $z_1=Xv_1вҲҲв„қ^N$ with sample variance $\V[z_1]=\/{1}{N}z_1^вҠӨz_1$
  $=\/{1}{N}v_1^вҠӨX^вҠӨXv_1$
  $=\/{Пғ_1^2}{N}v_1^вҠӨv_1$
  $\dg{вҮ’}\blue{\V[z_1]=\/{Пғ_1^2}{N}}$.
  Then $ОЈ^2$ in $\/{Пғ_j^2}{Пғ_j^2+О»}$ are just variances of the data along principal axes.
  Higher variance axes are easy to estimate slopes - ridge leaves this alone. Low variance slopes can flip wildly to effect of noise - ridge aggressively attunates the coordinates along this axis toward zero (robust).

  - **Degrees of freedom**: Anisotropic shrinkage pulls each parameter toward zero and reduces freedom.
  $\BA
  \t{df}(О»)=\tr[H_О»]
  =\tr[X(X^вҠӨX+О»I)^{-1}X^вҠӨ]
  =\tr[UОЈ(ОЈ^2+О»I)^{-1}ОЈ^вҠӨU^вҠӨ]
  =\tr[ОЈ^2(ОЈ^2+О»I)^{-1}]
  \dg{вҮ’}\blue{\t{df}(О»)=\sum_j\/{Пғ_j^2}{Пғ_j^2+О»}}
  \EA$
  is the effective degrees of freedom of the ridge regression fit.

- **Lasso**: $\red{\hat{ОІ}^\t{lasso}=\arg\min_ОІ(\n{y-XОІ}_2^2+О»\n{ОІ}_1)}$ uses penalty $\sum_j|ОІ_j|$ instead of $\sum_jОІ_j^2$. It has no closed-form solution, and requires convex optimization to solve. In 2004 LAR was introduced to compute lasso. Then in 2007 coordinate descent became the gold standard algorithm for lasso and elastic net.

- **Elastic net**: $О»\sum_{j=1}^p(ОұОІ_j^2+(1-Оұ)|ОІ_j|)$ has sharp corners while $\sum_j|ОІ_j|^q$ with q=1.x has more rounded corners.

- **3.6**: Under prior $ОІвҲј\Normal(0,П„I)$ and likelihood $yвҲј\Normal(XОІ,Пғ^2I)$, the mean/mode of the posterior distribution is $\hat{ОІ}^\t{ridge}$.
$\BA
\dg{вҮ’}&p(ОІ|y,X)вҲқp(y|X,ОІ)p(ОІ)=\Normal(y|XОІ,Пғ^2I)\Normal(ОІ|0,П„I)
\dg{вҮ’}E=-\/{1}{2Пғ^2}\n{y-XОІ}^2-\/{1}{2П„}\n{ОІ}^2\\
&=-\/{1}{2}[ОІ^вҠӨ(\/{1}{Пғ^2}X^вҠӨX+\/{1}{П„}I)ОІ-2ОІ^вҠӨX^вҠӨy]+c_1
\dg{вҮ’}p(ОІ|y,X)=\Normal(ОІ|m_N,S_N),\BC
S_N^{-1}=\/{1}{Пғ^2}X^вҠӨX+\/{1}{П„}I\\
m_N=S_NX^вҠӨy=(\/{1}{Пғ^2}X^вҠӨX+\/{1}{П„}I)^{-1}X^вҠӨy
\EC
\EA$
$\BA
\dg{вҮ’}&p(ОІ,y|X)=p(y|X,ОІ)p(ОІ)=\Normal(y|XОІ,Пғ^2I)\Normal(ОІ|0,П„I)
\dg{вҮ’}E=-\/{1}{2Пғ^2}\n{y-XОІ}^2-\/{1}{2П„}\n{ОІ}^2\\
&=-\/{1}{2}[ОІ^вҠӨ(\/{1}{Пғ^2}X^вҠӨX+\/{1}{П„}I)ОІ-2ОІ^вҠӨX^вҠӨy+\/{1}{Пғ^2}y^2]=-\/{1}{2}\BM ОІ\\y\EM^вҠӨ\BM\/{1}{Пғ^2}X^вҠӨX+\/{1}{П„}I&-X^вҠӨ\\-X&\/{1}{Пғ^2}I\EM\BM ОІ\\y\EM
\dg{вҮ’}\BC
ОЈ_{ОІ|y}=(\/{1}{Пғ^2}X^вҠӨX+\/{1}{П„}I)^{-1}\\
Ој_{ОІ|y}=(\/{1}{Пғ^2}X^вҠӨX+\/{1}{П„}I)^{-1}X^вҠӨy
\EC
\EA$

**Least angle regression (LAR)**: Standardize $XвҲҲв„қ^{NГ—p}$ (zero mean, unit norm). Center $y$ (zero mean). Set $ОІ=0$. Calculate residual $r_0=y-\bar{y}$. Find column $x_j$ most correlated with $r_0$ and add to active predictor set $\m{A}_1=\{x_j\}$.

- **Step-$k$** has prediction $\blue{\hat{y}_k=X_{\m{A}_k}ОІ_{\m{A}_k}}$, residual $\red{r_k=y-X_{\m{A}_k}ОІ_{\m{A}_k}}$, and direction $\red{Оҙ_k=(X_{\m{A}_k}^вҠӨX_{\m{A}_k})^{-1}X_{\m{A}_k}^вҠӨr_k}$.

  - $\BA
  \Corr(x_j,r_k)=\/{\V[x_j,r_k]}{\sqrt{\V[x_j]\V[r_k]}}=\/{\/{1}{N}x_j^вҠӨr_k}{\sqrt{\/{1}{N}\V[r_k]}}=\/{1}{\sqrt{N\V[r_k]}}x_j^вҠӨr_k
  \dg{вҮ’}\blue{c_k=X_{\m{A}_k}^вҠӨr_k}
  \EA$
  represents $\Corr(X_{\m{A}_k},r_k)$.

  - $Оҙ_kвҲҲв„қ^{|\m{A}_k|}$ is the direction of coefficient update.
  As $x_j$ enters $A_k$, coefficient $ОІ_j$ has the same sign as its correlation with residual $x_j^вҠӨr_k$.
  $\BA
  \dg{вҮ’}&Оҙ_k=(X_{\m{A}_k}^вҠӨX_{\m{A}_k})^{-1}\brown{X_{\m{A}_k}^вҠӨr_k}\dg{вҮ’}Оҙ_{kj}=\/{\brown{x_j^вҠӨr_k}}{x_j^вҠӨx_j}\dg{вҮ’}ОІ_j(Оұ)=ОұОҙ_{kj}=Оұ\/{\brown{x_j^вҠӨr_k}}{x_j^вҠӨx_j}
  \EA$

- **Updates**: During step-$k$ gradually increase step size $Оұ=0вҶ’1$ and grow active coefficients $\red{ОІ_{\m{A}_k}(Оұ)=ОІ_{\m{A}_k}+ОұОҙ_k}$ on top of the model states at the start of step-$k$.
$\BA
\dg{вҮ’}&r(Оұ)=y-X_{\m{A}_k}ОІ_{\m{A}_k}(Оұ)=y-X_{\m{A}_k}(ОІ_{\m{A}_k}+ОұОҙ_k)=r_k-ОұX_{\m{A}_k}Оҙ_k
\dg{вҮ’}\brown{r(Оұ)=r_k-ОұX_{\m{A}_k}Оҙ_k}\\
\dg{вҮ’}&c(Оұ)=X_{\m{A}_k}^вҠӨr(Оұ)=X_{\m{A}_k}^вҠӨ(r_k-ОұX_{\m{A}_k}Оҙ_k)=c_k-ОұX_{\m{A}_k}^вҠӨX_{\m{A}_k}Оҙ_k=c_k-ОұX_{\m{A}_k}^вҠӨX_{\m{A}_k}(X_{\m{A}_k}^вҠӨX_{\m{A}_k})^{-1}X_{\m{A}_k}^вҠӨr_k\dg{вҮ’}\brown{c(Оұ)=(1-Оұ)c_k}\\
\dg{вҮ’}&\hat{y}(Оұ)=X_{\m{A}_k}ОІ_{\m{A}_k}(Оұ)=X_{\m{A}_k}(ОІ_{\m{A}_k}+ОұОҙ_k)=\hat{y}_k+ОұX_{\m{A}_k}Оҙ_k\dg{вҮ’}\brown{\hat{y}(Оұ)=\hat{y}_k+Оұu_k}
\EA$

  - As coefficients move away from zero, the residual shrinks by $ОұX_{\m{A}_k}Оҙ_k$ and the correlation of all active predictors with the residual are tied to $c(Оұ)$ and decrease at the same rate $(1-Оұ)$.

  - Equiangular direction $\blue{u_k=X_{\m{A}_k}Оҙ_k}вҲҲв„қ^N$ is the direction of prediction update that ensures the correlation of all active predictors with the residual decrease at the same rate. Geometrically it sits exactly in the middle of all active predictors, and creates the least possible angle that is equal among all of them. For example if k=3, then $u_k$ is the vector right down the center of the cone formed by columns $x_1,x_2,x_3$.

  - $ОұвҲҲ[0,1]$. When $Оұ=1$ we have the OLS coefficients on the active set $X_{\m{A}_k}$
  $\BA
  ОІ_{\m{A}_k}+Оҙ_k=ОІ_{\m{A}_k}+(X_{\m{A}_k}^вҠӨX_{\m{A}_k})^{-1}X_{\m{A}_k}^вҠӨr_k=ОІ_{\m{A}_k}+(X_{\m{A}_k}^вҠӨX_{\m{A}_k})^{-1}X_{\m{A}_k}^вҠӨ(y-X_{\m{A}_k}ОІ_{\m{A}_k})=(X_{\m{A}_k}^вҠӨX_{\m{A}_k})^{-1}X_{\m{A}_k}^вҠӨy\dg{вҮ’}\green{ОІ_{\m{A}_k}(1)=\hat{ОІ}_{\m{A}_k}^\t{ls}}
  \EA$

- **Next step**: As $Оұ$ increases and $c(Оұ)$ decreases, another column $x_j'$ satisfying $|x_j'^вҠӨr(Оұ)|=c(Оұ)$ is added to $\m{A}_{k+1}$ to start next step. As such LAR always takes $p$ steps to get to the full OLS coefficients.

  - Along the way LAR gives a sparse model with 1, 2, ..., p predictors with non-zero coefficients smoothly growing toward their OLS values.

- **LAR-Lasso** adds an additional rule: when $ОІ_j$ shrinks to exact zero, $x_j$ is removed from $\m{A}_{k+1}$ to start the next step.
LAR-Lasso takes more than $p$ steps to get the full OLS coefficients, and along the way traces the entire continuous path of lasso-regularized coefficients with $О»=вҲһвҶ’0$. Any $О»>\max_j|x_j^вҠӨy|$ just gives $ОІ=0$.

  - Lasso ojective $\min_ОІ(\/{1}{2}\n{y-XОІ}_2^2+О»\n{ОІ}_1)$ using subgradient $вҲҮ\n{ОІ}_1$ has KKT stationarity conditions
  $\BA
  вҲҮ_ОІ=(y-XОІ)^вҠӨ(-X)+О»вҲҮ\n{ОІ}_1^вҠӨ=0\dg{вҮ’}X^вҠӨr=О»вҲҮ\n{ОІ}_1
  \dg{вҮ’}\BC
  ОІ_j\neq0&вҲӮ_j\n{ОІ}_1=\t{sign}(ОІ_j)&\brown{x_j^вҠӨr=О»вӢ…\t{sign}(ОІ_j)}\\
  ОІ_j=0&вҲӮ_j\n{ОІ}_1вҲҲ[-1,1]&\brown{|x_j^вҠӨr|вүӨО»}
  \EC\\
  \EA$.
  LAR-lasso active predictors ($ОІ_j\neq0$) have tied correlation $\brown{x_j^вҠӨr(Оұ)=c(Оұ)вӢ…\t{sign}_j}$ while inactive predictors ($ОІ_j=0$) satisfy $\brown{|x_j'^вҠӨr(Оұ)|вүӨc(Оұ)}$.
  LAR-lasso satisfies KKT stationarity as long as $\t{sign}_j=\t{sign}(ОІ_j)$, which only fails during $ОІ_j$ zero-crossing.
  Kicking out $x_j$ from $\m{A}_k$ prevents the failure of alignment with KKT stationarity.

  - LAR-lasso has same order of computation as a least squares fit with QR decomposition of $X$.

- **Degrees of freedom**: At step-$k$ the degrees of freedom is exactly $k$.

**Principal components regression**: Let $X=UОЈV^вҠӨвҲҲв„қ^{NГ—p}$ with thin $UвҲҲв„қ^{NГ—p}$, then principal components $Z_M=XV_M$ where $MвүӨp$ already has orthogonal columns and doesn't need deflation.
Coefficients along principal axes $\red{\hat{Оё}_j=\/{z_j^вҠӨy}{z_j^вҠӨz_j}}$ can be translated to standard basis
$\blue{\hat{ОІ}_M^\t{pcr}=V_M\hat{Оё}}=\sum_{j=1}^M\hat{Оё}_jv_j$.
Predictions are $\red{\hat{y}_M^\t{pcr}=\bar{y}\v{1}+Z_M\hat{Оё}}=\bar{y}\v{1}+\sum_{j=1}^M\hat{Оё}_jz_j$.

- Ridge regression scales principal components of $X$ by scaling factor $\/{Пғ_j^2}{Пғ_j^2+О»}$ while principal components regression truncates lower $p-M$ components of $X$ using scaling factor $\{1,0\}$.


# PRML: 4 Linear Models for Classification

**Discriminant function** takes an input $xвҲҲв„қ^D$ and assigns to a class $C_1,...,C_K$.
Binary problem has target $tвҲҲ\{1_{C_1},0_{C_2}\}$ while $K>2$ use one-hot $tвҲҲ\{0,1\}^K$.
Linear regression has prediction function $y(x)=w^вҠӨx+w_0$.
Classification has **activation function** $y(x)=f(w^вҠӨx+w_0)$ mapping linear output to a probability (sigmoid).


- **Binary linear discriminant** has a decision hyperplane $y(x)=w^вҠӨx+w_0=0$ as class boundary. A hyperplane is defined to be
$\red{\{x|w^вҠӨx=-w_0\}=\{x|xвҲҲx_0+w^\perp\}}$ where
$x_0=-\/{w_0}{\n{w}}\/{w}{\n{w}}$ is the projection of origin. Distance from hyperplane to any point $x_1$ is the distance from $x_0$ to projection of $x_1$ onto $w$, which is
$\n{\/{w^вҠӨx_1}{\n{w}}\/{w}{\n{w}}-x_0}=\/{y(x_1)}{\n{w}}$.

  - Naively extending to $K>2$ multiclass by combining multiple binary classifiers.
  **One-vs-rest** trains $K$ classifiers independently.
  **One-vs-one** trains $K(K-1)/2$ classifiers independently.
  There will be ambiguous regions in the input space with points claimed by multiple classifiers and points claimed by no classifier.

- **$K$-class linear discriminant** trains $K$ discriminant functions simultaneously
$\red{y_k(x)=w_k^вҠӨx+w_{k0}}$
with decision region
$\red{R_k=\{x|y_k(x)вүҘy_j(x)\ вҲҖj\}}$ and decision boundary
$y_k(x)=y_j(x)$
$\dg{вҮ’}(w_k-w_j)^вҠӨx+(w_{k0}-w_{j0})=0$.

  - The decision region is convex.
  Proof: $R_k=вӢӮ_j\{x|(w_j-w_k)^вҠӨxвүӨw_{k0}-w_{j0}\}$ is an intersection of halfspaces.

**OLS linear discriminant**: $\BA
W=\BM w_{10}&...&w_{K0}\\w_1&...&w_K\EMвҲҲв„қ^{(D+1)Г—K},X=\BM\v{1}&x\EMвҲҲв„қ^{NГ—(D+1)}\dg{вҮ’}XW=\BM y_1(x_1)&...&y_K(x_1)\\\vdots&\ddots&\vdots\\y_1(x_N)&...&y_K(x_N)\EMвҲҲв„қ^{NГ—K}
\EA$ is trained against $TвҲҲ\{0,1\}^{NГ—K}$ by minimizing RSS error $\red{\/{1}{2}\n{XW-T}_F^2}=\/{1}{2}\tr[(XW-T)^вҠӨ(XW-T)]$.

- $\red{W^*=(X^вҠӨX)^{-1}X^вҠӨT}$ where every row $y(x_i)^вҠӨ=(XW)_iвҲҲв„қ^K$ guarantees $\v{1}^вҠӨy(x_i)=1$ but not $y(x_i)вүҪ0$ and is therefore not a probability distribution.

- **Too correct problem** (Figure 4.4): On a 2D scatterplot of $x_1,x_2$ axes, regression predicts $x_2$ based on $x_1$.
Binary classification uses $x_1$ and $x_2$ to predict a third axis where $t=0$ is lower and $t=1$ is higher.
The boundary is the waterline at half-height.
OLS tilts decision boundary to outlier training points placed far from boundary because it assumes $t_iвҲј\Normal(y(x_i),Пғ^2)вҲҲ(-вҲһ,вҲһ)$ full range against actual $t_iвҲҲ\{0,1\}$ discrete range.
Outlier $x_i$ makes large predictions far from $\{0,1\}$.
For example $y(x_i)=5$ has penalty $\n{5-1}^2$ (too correct), thus tilts the boundary.

- **Masking problem** (Figure 4.5): $K=3$ parallel clusters of data has targets (1,0,0), (0,1,0), and (0,0,1). $y_1(x)$ and $y_3(x)$ easily predicts the two outer classes. Middle column goes from 0 to 1 to 0, but affine $y_2(x)$ cannot bend and therefore fits to a middling constant, say, $y_2(x)=0.33$, and is easily masked by $y_1(x)$ and $y_2(x)$ in $\arg\max_ky_k(x)$.

**Fisher linear discriminant**: $\red{y(x_i)=w^вҠӨx_i+w_0}$ projects points onto axis $w$ against a threshold placed at $-w_0$.

- **Scatter matrices**: Let $\bar{x}_k=\/{1}{N_k}\sum_{iвҲҲC_k}x_i$ be the class-$k$ mean, then along $w$ we have projected variances
$\BA
\dg{вҮ’}\t{between-class scatter}&[w^вҠӨ(\bar{x}_1-\bar{x}_2)]^2=w^вҠӨ[\yello{(\bar{x}_1-\bar{x}_2)(\bar{x}_1-\bar{x}_2)^вҠӨ}]w=w^вҠӨ\yello{S_{12}}w\\
\dg{вҮ’}\t{within-class scatter}&s_k^2=\sum_{iвҲҲC_k}[w^вҠӨ(x_i-\bar{x}_k)]^2=w^вҠӨ[\green{\sum_{iвҲҲC_k}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^вҠӨ}]w=w^вҠӨ\green{S_{k}}w
\EA$.

  - Covariance matrix is a sum of within-class and between-class scatters
  $\BA
  \dg{вҮ’}&\V[X]=(X-\v{1}\bar{x}^вҠӨ)^вҠӨ(X-\v{1}\bar{x}^вҠӨ)=\sum_i(x_i-\bar{x})(x_i-\bar{x})^вҠӨ=\sum_{iвҲҲC_1}(x_i-\bar{x})(x_i-\bar{x})^вҠӨ+\sum_{iвҲҲC_2}(x_i-\bar{x})(x_i-\bar{x})^вҠӨ\\
  \dg{вҮ’}&(x_i-\bar{x})(x_i-\bar{x})^вҠӨ=(x_i-\bar{x}_k+\bar{x}_k-\bar{x})(x_i-\bar{x}_k+\bar{x}_k-\bar{x})^вҠӨ=(x_i-\bar{x}_k)(x_i-\bar{x}_k)^вҠӨ+(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^вҠӨ\\
  &+(\bar{x}_k-\bar{x})(x_i-\bar{x}_k)^вҠӨ+(x_i-\bar{x}_k)(\bar{x}_k-\bar{x})^вҠӨ\dg{вҮҗ}\sum_{iвҲҲC_k}(x_i-\bar{x}_k)=0\\
  \dg{вҮ’}&\sum_{iвҲҲC_k}(x_i-\bar{x})(x_i-\bar{x})^вҠӨ=\sum_{iвҲҲC_k}(x_i-\bar{x}_k)(x_i-\bar{x}_k)^вҠӨ+N_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^вҠӨ=\green{S_k}+N_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^вҠӨ\\
  \dg{вҮ’}&\V[X]=\green{S_W}+\yello{S_B}\dg{вҮ’}\BC
  \green{S_W=\sum_k\sum_{iвҲҲC_k}(x_i-\bar{x})(x_i-\bar{x})^вҠӨ}\\
  \yello{S_B=\sum_kN_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^вҠӨ}
  \EC
  \EA$

  - In the binary case $\V[X]=\green{S_1+S_2}+\yello{\/{N_1N_2}{N}S_{12}}$ where $\green{S_W=S_1+S_2}$ and $\yello{S_B=\/{N_1N_2}{N}S_{12}}$.
  $\BA
  \dg{вҮ’}&\bar{x}=\/{N_1}{N}\bar{x}_1+\/{N_2}{N}\bar{x}_2
  \dg{вҮ’}\bar{x}_1-\bar{x}=\bar{x}_1-(\/{N_1}{N}\bar{x}_1+\/{N_2}{N}\bar{x}_2)
  \dg{вҮ’}\bar{x}_1-\bar{x}=\/{N_2}{N}(\bar{x}_1-\bar{x}_2)
  \dg{вҮ’}\bar{x}_2-\bar{x}=-\/{N_1}{N}(\bar{x}_1-\bar{x}_2)\\
  \dg{вҮ’}&\yello{S_B}=N_1(\bar{x}_1-\bar{x})(\bar{x}_1-\bar{x})^вҠӨ+N_2(\bar{x}_2-\bar{x})(\bar{x}_2-\bar{x})^вҠӨ=(N_1\/{N_2^2}{N^2}+N_2\/{N_1^2}{N^2})(\bar{x}_1-\bar{x}_2)(\bar{x}_1-\bar{x}_2)^вҠӨ\dg{вҮ’}\yello{S_B=\/{N_1N_2}{N}S_{12}}
  \EA$

- **Fisher criterion** is $J(w)=\/{[w^вҠӨ(\bar{x}_2-\bar{x}_1)]^2}{s_1^2+s_2^2}$
$=\/{w^вҠӨ\yello{S_{12}}w}{w^вҠӨ[\green{S_1+S_2}]w}$.
We pick axis $w^*=\arg\sup_w\red{\/{w^вҠӨ\yello{S_B}w}{w^вҠӨ\green{S_W}w}}$ that maximizes between-class scatter and minimizes within-class scatter to find the axis with minimum overlap.
  
  - **Generalized Rayleigh Quotient**: $О»_\min(\green{S_W}^{-1}\yello{S_B})вүӨ\red{\/{w^вҠӨ\yello{S_B}w}{w^вҠӨ\green{S_W}w}}вүӨО»_\max(\green{S_W}^{-1}\yello{S_B})$
  $\BA
  J(w)=(w^вҠӨ\yello{S_B}w)(w^вҠӨ\green{S_W}w)^{-1}
  \dg{вҮ’}вҲҮ_wJ(w)=[вҲҮ_w(w^вҠӨ\yello{S_B}w)](w^вҠӨ\green{S_W}w)^{-1}+(w^вҠӨ\yello{S_B}w)[вҲҮ_w(w^вҠӨ\green{S_W}w)^{-1}]\\
  =2w^вҠӨ\yello{S_B}(w^вҠӨ\green{S_W}w)^{-1}-(w^вҠӨ\yello{S_B}w)(w^вҠӨ\green{S_W}w)^{-2}2w^вҠӨ\green{S_W}=\/{2w^вҠӨ\yello{S_B}(w^вҠӨ\green{S_W}w)-2w^вҠӨ\green{S_W}(w^вҠӨ\yello{S_B}w)}{(w^вҠӨ\green{S_W}w)^2}=0\\
  \dg{вҮ’}w^вҠӨ\yello{S_B}(w^вҠӨ\green{S_W}w)-w^вҠӨ\green{S_W}(w^вҠӨ\yello{S_B}w)=0
  \dg{вҮ’}(w^вҠӨ\green{S_W}w)\yello{S_B}w=(w^вҠӨ\yello{S_B}w)\green{S_W}w
  \dg{вҮ’}\green{S_W}^{-1}\yello{S_B}w=\red{\/{w^вҠӨ\yello{S_B}w}{w^вҠӨ\green{S_W}w}}w\\
  \dg{вҮ’}\green{S_W}^{-1}\yello{S_B}w=\green{S_W}^{-1}(\bar{x}_2-\bar{x}_1)[(\bar{x}_2-\bar{x}_1)^вҠӨw]
  \dg{вҮ’}\red{w^*вҲқ\green{S_W}^{-1}(\bar{x}_2-\bar{x}_1)}
  \EA$

  - $w$ is eigenvector of $\green{S_W}^{-1}\yello{S_B}$ corresponding to $О»=\sup_w\red{\/{w^вҠӨ\yello{S_B}w}{w^вҠӨ\green{S_W}w}}$

- Modify OLS for binary classification with $t_i=\/{N}{N_1}\ вҲҖx_iвҲҲC_1$ and $t_i=-\/{N}{N_2}\ вҲҖx_iвҲҲC_2$ where $XвҲҲв„қ^{NГ—D}$
$\BA
\dg{вҮ’}&\v{1}^вҠӨt=0\dg{вҮ’}w_0=-w^вҠӨ\bar{x}\\
\dg{вҮ’}&вҲҮ_w\n{Xw+w_0\v{1}-t}^2=2(Xw+w_0\v{1}-t)^вҠӨX=0
\dg{вҮ’}X^вҠӨ(Xw-\v{1}\bar{x}^вҠӨw-t)=0
\dg{вҮ’}X^вҠӨ(X-\v{1}\bar{x}^вҠӨ)w=X^вҠӨt\\
&\t{RHS: }X^вҠӨt=\sum_ix_i^вҠӨt_i=\/{N}{N_1}\sum_{iвҲҲC_1}x_i-\/{N}{N_2}\sum_{iвҲҲC_2}x_i=N(\bar{x}_1-\bar{x}_2)\\
&\t{LHS: }(\v{1}\bar{x}^вҠӨ)^вҠӨ(X-\v{1}\bar{x}^вҠӨ)=\bar{x}\v{1}^вҠӨX-\bar{x}\v{1}^вҠӨ\v{1}\bar{x}^вҠӨ=\bar{x}N\bar{x}^вҠӨ-\bar{x}N\bar{x}^вҠӨ=0\dg{вҮ’}X^вҠӨ(X-\v{1}\bar{x}^вҠӨ)=(X-\v{1}\bar{x}^вҠӨ)^вҠӨ(X-\v{1}\bar{x})\\
\dg{вҮ’}&(\green{S_W}+\yello{S_B})w=N(\bar{x}_1-\bar{x}_2)\dg{вҮ’}\red{w^*вҲқ\green{S_W}^{-1}(\bar{x}_2-\bar{x}_1)}
\EA$

- For $K>2$, we use $WвҲҲв„қ^{DГ—D'}$ to project $y_i=W^вҠӨx_iвҲҲв„қ^{D'}$
$\BC s_W=\sum_k\sum_{iвҲҲC_k}(y_i-\bar{y}_k)(y_i-\bar{y}_k)^вҠӨ=WS_WW^вҠӨ\\s_B=\sum_kN_k(\bar{y}_k-\bar{y})(\bar{y}_k-\bar{y})^вҠӨ=WS_BW^вҠӨ\EC$
$\dg{вҮ’}J(W)=\tr[s_W^{-1}s_B]=\tr[(WS_WW^вҠӨ)^{-1}(WS_BW^вҠӨ)]$.

  - $W^*$ are the eigenvectors corresponding to $D'$ largest eigenvalues of $S_W^{-1}S_B$.

  - $S_B=\sum_kN_k(\bar{x}_k-\bar{x})(\bar{x}_k-\bar{x})^вҠӨвҲҲв„қ^{DГ—D}$ is a sum of $K$ rank-1 matrices.
  $\rk(S_B)=K$ but $\bar{x}$ used 1 degree of freedom. Only $K-1$ of those matrices are linearly independent.
  There are at most $K-1$ non-zero eigenvalues.
  Therefore $1< D'вүӨK-1$.

**Perceptron discriminant**: binary linear discriminant $y(x_i)=\t{sign}(w^вҠӨП•(x_i))$ with activation function $\t{sign}(вӢ…)$.
We seek $w$ predicting $\BC
w^вҠӨП•(x_i)> 0\dg{вҮ’}iвҲҲC_1\\
w^вҠӨП•(x_i)< 0\dg{вҮ’}iвҲҲC_2
\EC$ against $t_iвҲҲ\{+1_{C_1},-1_{C_2}\}$
with cost $\BC
-w^вҠӨП•(x_i)t_i> 0&\t{if misclassified }iвҲҲ\m{M}\\
-w^вҠӨП•(x_i)t_i=0&\t{if correctly classified}
\EC$.

- **Cost** $E_P(w)=-\sum_{iвҲҲ\m{M}}w^вҠӨП•(x_i)t_i=\sum_{i=1}^N\max(0,-w^вҠӨП•(x_i)t_i)$ is piecewise linear sum of $N$ hockey sticks.
SGD $w^{(П„+1)}=w^{(П„)}-О·вҲҮE_P(w)$ uses $вҲҮE_P(w)=-П•(x_i)t_i$ for random $iвҲҲ\m{M}$.
Then $\red{w^{(П„+1)}=w^{(П„)}+О·П•(x_i)t_i}$.

  - Because magnitude is ignored by the activation function and $w^{(0)}=0$, we can assume $О·=1$ without loss of generality as long as it stays constant.

  - Naive solution to the problem by simply counting $|\m{M}|$ creates a piecewise constant error, which does not have gradients.
  Perceptron algorithm has piecewise linear error, which allows gradient descent.
  The non-differentiable joints between pieces are not a problem for SGD because $вҲҮE_P(w)=-П•(x_i)t_i$ is always defined.

- **Heart of algorithm**: $w:=w+П•(x_i)$ for misclassified $t_i=+1$ to align $w$ closer to $П•(x_i)$ and $w^вҠӨП•(x_i)вҶ’+1$.
Update $w:=w-П•(x_i)$ for misclassified $t_i=-1$ to align $w$ farther from $П•(x_i)$ and $w^вҠӨП•(x_i)вҶ’-1$.

- **Convergence**: If the data points are linearly separable into two classes neatly by a separating hyperplane, then the algorithm is guaranteed to find an exact solution in finite steps.

  - If the data points are not linearly separable then the algorithm will never converge.

  - The algorithm does not generalize to $K>2$.

**Probabilistic generative methods** model the underlying distributions and calculate the probability that input $xвҲҲв„қ^D$ belongs to a class $C_1,...,C_K$.

- **Logistic and logit**: $p(C_1|x_i)=\/{p(x_i|C_1)p(C_1)}{p(x_i|C_1)p(C_1)+p(x_i|C_2)p(C_2)}=\/{1}{1+\e{-a}}=Пғ(a)$ is logistic sigmoid function
where $\red{a=\ln\/{p(x_i|C_1)p(C_1)}{p(x_i|C_2)p(C_2)}}$ is logit (log-odds) function.
$\red{Пғ(a)=\/{1}{1+\e{-a}}}=\/{1}{1+\/{p(x_i|C_2)p(C_2)}{p(x_i|C_1)p(C_1)}}=\/{p(x_i|C_1)p(C_1)}{p(x_i|C_1)p(C_1)+p(x_i|C_2)p(C_2)}$.

  - Logistic $p(C_k|x_i)=Пғ(a)$ and logit $a=\ln\/{Пғ}{1-Пғ}$ only works for $K=2$. It is one-vs-rest in $K>2$ and cannot guarantee $\sum_kp(C_k|x_i)=1$ because it creates ambiguities in the input space.
  Let $\blue{a_k=\ln p(x|C_k)p(C_k)}$ then
  $\BA
  \dg{вҮ’}&a=\ln\/{p(x_i|C_1)p(C_1)}{p(x_i|C_2)p(C_2)}=\ln\/{p(C_1|x_i)}{p(C_2|x_i)}\t{ log-odds in posterior probabilities = log-odds in joint probabilities}\\
  \dg{вҮ’}&a=\ln p(x_i|C_1)p(C_1)-\ln p(x_i|C_2)p(C_2)=a_1-a_2 \t{ is the difference in log joint probabilities}\\
  \dg{вҮ’}&Пғ=\/{1}{1+\e{-a}}\dg{вҮ’}1=Пғ+Пғ\e{-a}\dg{вҮ’}\e{-a}=\/{1-Пғ}{Пғ}\dg{вҮ’}a=\ln\/{Пғ}{1-Пғ}\t{ log-odds in posterior probabilities}\\
  \dg{вҮ’}&Пғ(a)=\/{1}{1+\e{-a}}=\/{\e{a}}{1+\e{a}}\t{ logistic equivalent form}\\
  \dg{вҮ’}&Пғ(a)=\/{\e{a_1-a_2}}{1+\e{a_1-a_2}}=\/{\e{a_1}}{\e{a_1}+\e{a_2}}\t{ logistic is binary softmax of log joint probabilities}\\
  \dg{вҮ’}&Пғ(-a)=1-Пғ(a)\t{ logistic is symmetric}\\
  \EA$

  - $K>2$: $\BA \red{p(C_k|x_i)=\/{p(x_i|C_k)p(C_k)}{\sum_jp(x_i|C_j)p(C_j)}=\/{\e{a_k}}{\sum_j\e{a_j}}}\EA$ is softmax (normalized exponential) function.

- **Shared covariance**: If we assume $ОЈ$ is shared among all $p(x_i|C_k)$ we arrive at a linear discriminant model.
$\BA
\dg{вҮ’}&p(x_i|C_k)=\Normal(x_i|Ој_k,ОЈ)=\/{1}{(2ПҖ)^{D/2}\det(ОЈ)^{1/2}}\e{-\/{1}{2}(x_i-Ој_k)^вҠӨОЈ^{-1}(x_i-Ој_k)}\\
\dg{вҮ’}&a=\ln p(x_i|C_1)-\ln p(x_i|C_2)+\ln\/{p(C_1)}{p(C_2)}
=-\/{1}{2}(x_i-Ој_1)^вҠӨОЈ^{-1}(x_i-Ој_1)+\/{1}{2}(x_i-Ој_2)^вҠӨОЈ^{-1}(x_i-Ој_2)+\ln\/{p(C_1)}{p(C_2)}\\
&=\green{(Ој_1-Ој_2)^вҠӨОЈ^{-1}}x_i\yello{-\/{1}{2}[Ој_1^вҠӨОЈ^{-1}Ој_1-Ој_2^вҠӨОЈ^{-1}Ој_2]+\ln\/{p(C_1)}{p(C_2)}}\dg{вҮ’}a=\green{w^вҠӨ}x_i+\yello{w_0}\dg{вҮ’}\red{p(C_1|x_i)=Пғ(\green{w^вҠӨ}x_i+\yello{w_0})}
\EA$
$\BA
\dg{вҮ’}&a_k=\purple{c_1}-\/{1}{2}(x_i-Ој_k)^вҠӨОЈ^{-1}(x_i-Ој_k)+\ln p(C_k)=\purple{c_1-\/{1}{2}x_i^вҠӨОЈ^{-1}x_i}-\/{1}{2}(-2Ој_k^вҠӨОЈ^{-1}x_i+Ој_k^вҠӨОЈ^{-1}Ој_k)+\ln p(C_k)\\
&=\purple{c_1-\/{1}{2}x_i^вҠӨОЈ^{-1}x_i}+\green{Ој_k^вҠӨОЈ^{-1}}x_i\yello{-\/{1}{2}Ој_k^вҠӨОЈ^{-1}Ој_k+\ln p(C_k)}\dg{вҮ’}a_k=\green{w_k^вҠӨ}x_i+\yello{w_{k0}}\dg{вҮ’}\red{p(C_k|x_i)=\/{\e{\green{w_k^вҠӨ}x_i+\yello{w_{k0}}}}{\sum_j\e{\green{w_j^вҠӨ}x_i+\yello{w_{j0}}}}}
\EA$

  - $\purple{c_1-\/{1}{2}x_i^вҠӨОЈ^{-1}x_i}$ gets canceled out against denominator in softmax, and we're left with same decision region as K-class linear discriminant $R_k=\{x|y_k(x)вүҘy_j(x)\ вҲҖj\}$ but in softmax form.

  - $\BA
  p(C_k|x_i)=p(C_j|x_i)\dg{вҮ’}a_k(x_i)=a_j(x_i)\dg{вҮ’}(w_k-w_j)^вҠӨx_i+(w_{k0}-w_{j0})=0
  \EA$
  Linearity of the discriminant derives from linearity of $a_k$.
  Relaxing shared covariance condition gives rise to quadratic $a_k$ and quadratic discriminant.

  - To use $p(C_k|x_i)$ we need to find $Ој_k$ and $ОЈ$ from training data through likelihood function $p(t)$.
  In practice we skip all these steps and go straight to linear discriminant model: logistic regression.

- **MLE**: we find $Ој_k$ and $ОЈ$ from likelihood $p(t)=\prod_ip(x_i,C_1)^{t_i}p(x_i,C_2)^{1-t_i}$ where $x_iвҲҲв„қ^D,t_iвҲҲ\{1_{C_1},0_{C_2}\}$.
Then $\BA
\dg{вҮ’}&\BC
p(C_1)=ПҖ\dg{вҮ’}p(x_i,C_1)=ПҖ\Normal(x_i|Ој_1,ОЈ)\\
p(C_2)=1-ПҖ\dg{вҮ’}p(x_i,C_2)=(1-ПҖ)\Normal(x_i|Ој_2,ОЈ)
\EC
\dg{вҮ’}p(t|ПҖ,Ој_1,Ој_2,ОЈ)=\prod_i(ПҖ\Normal(x_i|Ој_1,ОЈ))^{t_i}((1-ПҖ)\Normal(x_i|Ој_2,ОЈ))^{1-t_i}\\
\dg{вҮ’}&\ln p(t|ПҖ)=\sum_i[t_i\lnПҖ+(1-t_i)\ln(1-ПҖ)]+c_1\dg{вҮ’}вҲӮ_ПҖ=\/{\sum_it_i}{ПҖ}-\/{N-\sum_it_i}{1-ПҖ}=0\dg{вҮ’}N_1-ПҖN_1=ПҖN-ПҖN_1\dg{вҮ’}\green{ПҖ_\t{ML}=\/{N_1}{N}}\\
\dg{вҮ’}&\ln p(t|Ој_1)=\sum_i-\/{1}{2}t_i(x_i-Ој_1)^вҠӨОЈ^{-1}(x_i-Ој_1)+c_2\dg{вҮ’}вҲӮ_{Ој_1}=\sum_it_i(x_i-Ој_1)^вҠӨОЈ^{-1}=0\dg{вҮ’}\sum_it_ix_i=\sum_it_iОј_1\\
&\dg{вҮ’}\green{Ој_{1\t{ML}}=\/{\sum_it_ix_i}{N_1}}\quad\dg{вҮ’}\sum_i(1-t_i)x_i=\sum_i(1-t_i)Ој_2\dg{вҮ’}\green{Ој_{2\t{ML}}=\/{\sum_i(1-t_i)x_i}{N_2}}\\
\dg{вҮ’}&\ln p(t|ОЈ)=\sum_it_i[-\/{1}{2}\ln\det(ОЈ)-\/{1}{2}(x_i-Ој_1)^вҠӨОЈ^{-1}(x_i-Ој_1)]+\sum_i(1-t_i)[-\/{1}{2}\ln\det(ОЈ)-\/{1}{2}(x_i-Ој_2)^вҠӨОЈ^{-1}(x_i-Ој_2)]\\
&=-\/{N}{2}\ln\det(ОЈ)-\sum_i\/{t_i}{2}\tr[ОЈ^{-1}(x_i-Ој_1)(x_i-Ој_1)^вҠӨ]-\sum_i\/{1-t_i}{2}\tr[ОЈ^{-1}(x_i-Ој_2)(x_i-Ој_2)^вҠӨ]\\
&=-\/{N}{2}\ln\det(ОЈ)-\/{1}{2}\tr[ОЈ^{-1}(\sum_it_i(x_i-Ој_1)(x_i-Ој_1)^вҠӨ+\sum_i(1-t_i)(x_i-Ој_2)(x_i-Ој_2)^вҠӨ)]\\
&=-\/{N}{2}\ln\det(ОЈ)-\/{1}{2}\tr[ОЈ^{-1}(N_1ОЈ_1+N_2ОЈ_2)]\dg{вҮ’}вҲӮ_ОЈ=-\/{N}{2}ОЈ^{-1}+\/{1}{2}(N_1ОЈ_1+N_2ОЈ_2)ОЈ^{-2}=0\dg{вҮ’}\green{ОЈ_\t{ML}=\/{N_1}{N}ОЈ_1+\/{N_2}{N}ОЈ_2}
\EA$
